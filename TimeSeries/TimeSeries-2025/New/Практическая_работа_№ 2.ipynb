### **2. Теоретические предпосылки**

Классические модели временных рядов опираются на строгие статистические предположения, выполнение которых является необходимым условием их корректного применения. Нарушение этих предпосылок — например, наличие мультиколлинеарности в экзогенных признаках, автокорреляции или нестационарности остатков — приводит к смещённым оценкам, неадекватным прогнозам и ошибочным интерпретациям.

Ключевым аспектом сравнительного анализа является использование **единых, предварительно обработанных данных**, что исключает влияние различий в предобработке на итоговые результаты. В рамках настоящей работы такие данные поступают из Практической работы № 1, где уже выполнены: унификация временной метки, обработка пропусков и выбросов, инженерия признаков (включая лаги, скользящие окна и циклические кодировки), а также приведение ряда к стационарности с помощью лог-трансформации, преобразования Бокса–Кокса и дифференцирования.

При включении экзогенных переменных (например, в SARIMAX или Prophet с регрессорами) необходимо проверять отсутствие мультиколлинеарности с помощью коэффициента инфляции дисперсии (VIF). Значения VIF > 10 указывают на чрезмерную корреляцию между признаками и требуют коррекции, так как искажают оценки коэффициентов и снижают интерпретируемость.

Многопшаговое прогнозирование на горизонт \( h > 1 \) требует выбора стратегии. Рекурсивная стратегия экономична, но подвержена накоплению ошибки; прямая — более точна, но вычислительно затратна; гибридная — представляет компромисс. Корректная оценка качества возможна только при использовании временной кросс-валидации (TimeSeriesSplit), исключающей утечку будущего.

Интерпретируемость модели — не менее важный критерий, чем точность. В SARIMAX анализируются p-значения коэффициентов экзогенных переменных; в Prophet — вклад компонентов тренда и сезонности; в моделях Хольта — значения параметров сглаживания \( \alpha \) и \( \beta \). Эти аспекты позволяют не только прогнозировать, но и объяснять динамику системы.

### **3. Порядок выполнения работы**

Работа выполняется в несколько последовательных этапов, каждый из которых направлен на достижение конкретных образовательных и исследовательских результатов.

#### **3.1. Использование данных из Практической работы № 1**

**Задача:** Обеспечить методологическую преемственность и воспроизводимость.  
**Требования к выполнению:**  
– В качестве входных данных используется финальный обработанный датасет из Практической работы № 1:  
  • Объём: не менее 720 последовательных наблюдений;  
  • Наличие целевой переменной и не менее двух дополнительных признаков;  
  • Подтверждённая стационарность (результаты ADF/KPSS);  
  • Сохранённые параметры преобразований (например, \( \lambda \) для Бокса–Кокса).  
– Все экзогенные признаки (лаги, скользящие статистики, циклические кодировки) доступны для моделей, поддерживающих внешние факторы.  
– Обратное преобразование прогнозов осуществляется с использованием сохранённых параметров.

#### **3.2. Анализ мультиколлинеарности (VIF)**

**Задача:** Обеспечить устойчивость и интерпретируемость коэффициентов моделей с экзогенными переменными.  
**Требования к выполнению:**  
– Расчёт VIF для всех экзогенных признаков с использованием `statsmodels.stats.outliers_influence.variance_inflation_factor`;  
– При обнаружении VIF > 10 — исключение одного из коррелирующих признаков с сохранением наиболее содержательного или лучше коррелирующего с целевой переменной;  
– Фиксация финального набора признаков и их VIF в отчёте;  
– Использование скорректированного набора только для моделей SARIMAX, Prophet (с регрессорами) и VAR.

#### **3.3. Стратегии многопшагового прогнозирования**

**Задача:** Реализовать и сравнить подходы к прогнозу на горизонт \( h \geq 7 \).  
**Требования к выполнению:**  
– **Рекурсивная стратегия:** одна модель; прогноз на \( t+1 \) используется как вход для \( t+2 \);  
– **Прямая стратегия:** отдельная модель для каждого шага \( t+1, \dots, t+h \);  
– **Гибридная стратегия:** например, рекурсивная для \( h \leq 3 \), прямая — для \( h > 3 \);  
– Для каждой стратегии и каждого горизонта (\( h = 1, 7, 30 \)):  
  • Расчёт MAE, RMSE на каждом шаге;  
  • Фиксация времени вычислений;  
  • Построение графика накопления ошибки.

#### **3.4. Построение и настройка моделей**

**Задача:** Реализовать единообразный интерфейс для всех классических моделей.  
**Требования к выполнению:**  
– **Базовые модели:** AR, MA, ARMA, ARIMA, SARIMA — с подбором порядков через AIC/BIC или `pmdarima.auto_arima`;  
– **SARIMAX:** с использованием скорректированного набора признаков после VIF-анализа;  
– **Модели экспоненциального сглаживания:**  
  • SES: `ExponentialSmoothing(..., trend=None, seasonal=None, optimized=True)`;  
  • Хольт (аддитивный): `trend='add'`;  
  • Хольт (мультипликативный): `trend='mul'` (только при положительных данных);  
– **GARCH(1,1):** обучение на остатках ARIMA с использованием библиотеки `arch`;  
– **Сложные модели:** TBATS (для сложной сезонности), Prophet (с поддержкой праздников и регрессоров);  
– **Многомерные модели:** VAR, VECM (при наличии ≥2 рядов в исходных данных);  
– **Бенчмарки:** Naive (\( \hat{y}_{t+h} = y_t \)), Seasonal Naive (\( \hat{y}_{t+h} = y_{t+h-s} \)).  
– Для всех моделей: прогноз, доверительные интервалы (если поддерживается), обратное преобразование.

#### **3.5. Диагностика адекватности моделей**

**Задача:** Убедиться, что модель корректно аппроксимирует динамику ряда.  
**Требования к выполнению:**  
– **Тест Льюнга–Бокса** на остатках с использованием `statsmodels.stats.diagnostic.acorr_ljungbox` (H₀: автокорреляция отсутствует; p > 0.05 — адекватность);  
– **Гомоскедастичность:** визуальный анализ графика остатков против прогнозов;  
– **Нормальность остатков:**  
  • Q-Q график;  
  • Тест Шапиро–Уилка (`scipy.stats.shapiro`); p > 0.05 — нормальность не отвергается;  
– **ACF/PACF остатков** — должны лежать в пределах доверительного интервала.  
– Диагностика обязательна для всех моделей экспоненциального сглаживания и топ-5 моделей по метрикам.

#### **3.6. Оценка качества и статистическое сравнение**

**Задача:** Обоснованно сравнить модели по объективным критериям.  
**Требования к выполнению:**  
– Расчёт метрик на тестовой выборке для каждого горизонта: MAE, RMSE, MAPE, MASE, SMAPE, R², RMSLE;  
– Применение **теста Дикболда–Мариано** (`statsmodels.stats.diagnostic.dm_test`) для попарного сравнения прогнозов;  
– Составление **сравнительных таблиц**:  
  • Модели × стратегии × горизонты (метрики, время);  
  • Интерпретируемость (VIF, значимые коэффициенты, p-значения);  
  • Диагностические p-значения (Ljung–Box, Shapiro–Wilk).  
– Ранжирование моделей по взвешенной оценке, учитывающей точность, адекватность и интерпретируемость.

#### **3.7. Анализ интерпретируемости моделей**

**Задача:** Оценить способность модели объяснять динамику системы.  
**Требования к выполнению:**  
– Для **SARIMAX**: анализ коэффициентов экзогенных переменных, их p-значений и доверительных интервалов;  
– Для **Prophet**: визуализация и количественная оценка вклада компонентов (тренд, сезонность, праздники);  
– Для **моделей Хольта**: интерпретация параметров \( \alpha \) (уровень) и \( \beta \) (тренд) как мер адаптивности;  
– Для **VAR**: анализ матрицы коэффициентов и взаимного влияния рядов;  
– Результаты интерпретации включаются в сравнительные таблицы и веб-интерфейс.

#### **3.8. Разработка веб-интерфейса для анализа результатов**
