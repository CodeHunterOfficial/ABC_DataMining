### **2. Теоретические предпосылки**

Современные задачи анализа временных рядов всё чаще сталкиваются с требованиями к скорости, масштабируемости и минимизации участия эксперта. В этом контексте AutoML-платформы выступают не просто как инструменты автоматизации, а как **методологические среды**, в которых реализуются целые философии моделирования: от «всё автоматически» до «автоматизация под контролем».

Ключевое различие между AutoML-системами для временных рядов заключается в **парадигме поиска**:  
– **Метаобучение и эвристики** (AutoGluon) — выбор моделей на основе внутренней базы знаний;  
– **Генетические алгоритмы и эволюционный поиск** (AutoTS) — итеративная оптимизация конфигурации модели;  
– **Адаптивный бюджетированный поиск** (FLAML) — минимизация функции потерь при фиксированном времени;  
– **Унифицированные глубокие архитектуры** (Darts) — использование нейросетевых моделей с автоматической настройкой через Optuna или Ray Tune.

Важнейшей характеристикой AutoML-решения является **прозрачность**: возможность заглянуть внутрь процесса — какие модели сравнивались, какие гиперпараметры выбирались, как строилось ансамбль. В условиях регулируемых отраслей (финансы, здравоохранение) «чёрный ящик» недопустим, даже если он точен.

Кроме того, не все AutoML-системы одинаково корректно обрабатывают **временную структуру данных**. Некоторые фреймворки (например, H2O AutoML) изначально разрабатывались для задач классификации/регрессии и применяются к временным рядам через преобразование в табличную форму, что требует от пользователя явного контроля за утечкой будущего. Другие (AutoGluon-TimeSeries, Darts) встроены с пониманием временной оси и автоматически применяют временные разбиения.

Настоящая работа ставит задачу не просто «запустить AutoML», а **оценить, насколько автоматизация заменяет или дополняет методологическую строгость**, заложенную в классический ML-пайплайн (Практическая работа № 3).

### **3. Порядок выполнения работы**

Работа выполняется в несколько последовательных этапов, каждый из которых направлен на достижение конкретных образовательных и исследовательских результатов.

#### **3.1. Использование данных из Практической работы № 1**

**Задача:** Обеспечить методологическую преемственность и корректный формат для всех платформ.  
**Требования к выполнению:**  
– В качестве входных данных используется финальный обработанный датасет из Практической работы № 1:  
  • Объём: не менее 720 последовательных наблюдений;  
  • Наличие целевой переменной и не менее пяти дополнительных признаков (включая лаги \( y_{t-1}, y_{t-7}, y_{t-30} \), скользящие средние, циклические кодировки времени);  
  • Подтверждённая стационарность (результаты ADF/KPSS);  
  • Сохранённые параметры преобразований (например, \( \lambda \) для Бокса–Кокса).  
– Форматирование под требования платформ:  
  • Единый формат: колонка `timestamp` (datetime), `target` и, при наличии, `item_id` (для многорядных сценариев);  
  • Отсутствие пропусков в целевой переменной;  
  • Для AutoGluon: данные в формате `TimeSeriesDataFrame`;  
  • Для Darts: преобразование в `TimeSeries` объект.  
– Все платформы получают **один и тот же тестовый период** (последние 90 наблюдений) — **только для итоговой оценки, не участвует в настройке**.

#### **3.2. Формализация задачи и разделение данных**

**Задача:** Корректно сформулировать задачу машинного обучения с учётом временной структуры.  
**Требования к выполнению:**  
– Целевая переменная: \( y_t \); признаки \( X_t \): все доступные лаги, статистики и экзогенные факторы.  
– Разделение данных:  
  • Тренировочная + валидационная часть: всё до \( T - 90 \);  
  • Финальный тест: последние 90 наблюдений — **только для итоговой оценки**.  
– Для кросс-валидации в платформах, поддерживающих валидацию: `TimeSeriesSplit(n_splits=5, test_size=60)` — без перемешивания, с сохранением порядка.

#### **3.3. Определение конфигураций и настройка платформ**

**Задача:** Привести каждую AutoML-платформу к её рабочей конфигурации в честных условиях.  
**Требования к выполнению:**  
– Для каждой платформы задаются:  
  • **Временной бюджет**: 30 минут (1800 секунд) на задачу;  
  • **Метрика оптимизации**: MAE или MASE (масштаб-инвариантные);  
  • **Горизонт прогнозирования**: \( h = 1, 7, 30 \);  
  • **Разрешённые модели**: при необходимости — ограничение для честного сравнения;  
  • **Ковариаты**: использование инженерных признаков — только если платформа поддерживает.  
– Примеры конфигураций:  
  • **AutoGluon**: `predictor = TimeSeriesPredictor(..., eval_metric="MASE").fit(df, presets="high_quality")`;  
  • **AutoTS**: `model = AutoTS(..., forecast_length=30, model_list="superfast")`;  
  • **FLAML**: `automl = AutoML(); automl.fit(X_train, y_train, task="ts_forecast", time_budget=1800)`;  
  • **Darts**: `model = NBEATSModel(input_chunk_length=30, output_chunk_length=30); model.fit(series)`;  
  • **H2O**: `aml = H2OAutoML(max_runtime_secs=1800, sort_metric="mae")`.  
– Все процедуры выполняются **только на тренировочной части**.  
– Фиксация: лучшая модель, гиперпараметры, значение метрики на валидации, время настройки.

#### **3.4. Стратегии многопшагового прогнозирования**

**Задача:** Реализовать и сравнить подходы к прогнозу на горизонт \( h \geq 7 \).  
**Требования к выполнению:**  
– Для платформ, поддерживающих стратегии:  
  • **Рекурсивная стратегия**: одна настроенная модель; прогноз на \( t+1 \) используется как вход для \( t+2 \);  
  • **Прямая стратегия**: отдельная настройка и обучение модели для каждого шага \( t+1, \dots, t+h \);  
  • **Автоматический выбор**: если платформа сама выбирает стратегию — фиксация этого решения.  
– Для каждой платформы и каждого горизонта (\( h = 1, 7, 30 \)):  
  • Расчёт MAE, RMSE на каждом шаге;  
  • Фиксация общего времени вычислений;  
  • Построение графика накопления ошибки и времени.

#### **3.5. Построение и настройка моделей**

**Задача:** Реализовать единообразный процесс запуска всех AutoML-платформ.  
**Требования к выполнению:**  
– Все платформы запускаются в изолированных окружениях с фиксированным seed;  
– Для платформ без встроенной валидации — ручное разделение с `TimeSeriesSplit`;  
– Для H2O AutoML — явный контроль за утечкой будущего через ручное формирование лагов;  
– Для каждой платформы: прогноз на тесте, обратное преобразование (если применялось), сохранение времени и метаданных.

#### **3.6. Диагностика адекватности моделей**

**Задача:** Убедиться, что AutoML-системы корректно обрабатывают временную структуру.  
**Требования к выполнению:**  
– **Анализ утечки будущего**: проверка, что признаки не содержат информации из будущего;  
– **Детекция аномалий**: оценка, использует ли платформа встроенные механизмы обнаружения выбросов;  
– **Автоматическая обработка пропусков**: фиксация, как платформа реагирует на missing values;  
– **Сезонность**: проверка, корректно ли определён период (например, 7 для недельного цикла).  
– Диагностика обязательна для всех платформ.

#### **3.7. Оценка качества и статистическое сравнение**

**Задача:** Обоснованно сравнить платформы по объективным критериям.  
**Требования к выполнению:**  
– Расчёт метрик **на финальной тестовой выборке** для каждого горизонта: MAE, RMSE, MAPE, MASE, SMAPE, R²;  
– Применение **теста Дикболда–Мариано** (`statsmodels.stats.diagnostic.dm_test`) для попарного сравнения прогнозов топ-5 моделей из AutoML и топ-3 моделей из работы № 3;  
– Составление **сравнительных таблиц**:  
  • Платформы × стратегии × горизонты (метрики, время настройки, время прогноза);  
  • Интерпретируемость (SHAP, важность признаков, прозрачность Leaderboard);  
  • Поддержка функций (ковариаты, многорядность, аномалии).  
– Ранжирование платформ по взвешенной оценке, учитывающей точность (50%), адекватность (30%) и интерпретируемость (20%).

#### **3.8. Анализ интерпретируемости моделей**

**Задача:** Оценить способность платформы объяснять прогноз.  
**Требования к выполнению:**  
– Для **AutoGluon**: анализ встроенных SHAP-значений для глубоких моделей;  
– Для **H2O**: переменная важность по Джини или SHAP;  
– Для **Darts**: attention-мапы (если модель — TFT);  
– Для **AutoTS/FLAML**: фиксация отсутствия интерпретации как ограничения;  
– Проведение **локальной интерпретации**: SHAP-водопадные диаграммы для отдельных точек прогноза;  
– Результаты интерпретации включаются в сравнительные таблицы и веб-интерфейс.

#### **3.9. Анализ вычислительной эффективности**

**Задача:** Оценить практическую применимость платформ в условиях ограниченных ресурсов.  
**Требования к выполнению:**  
– Фиксация времени:  
  • Настройка (в секундах),

**Задача:** Создать интерактивное приложение для визуального и функционального сравнения платформ.

### **6. Критерии оценивания**

| Оценка | Критерии |
|--------|----------|
| **Отлично** | Полное выполнение всех этапов; настройка всех AutoML-платформ через временную кросс-валидацию; реализация всех трёх стратегий прогнозирования; углублённая диагностика утечки и аномалий; тест Дикболда–Мариано; развёрнутый анализ интерпретируемости (SHAP, Leaderboard, устойчивость); оценка вычислительной сложности; функциональный веб-интерфейс с отображением гиперпараметров и объяснений; публикация всех материалов; отчёт, соответствующий ГОСТ. |
| **Хорошо** | Выполнение основных этапов; настройка 3–4 платформ; наличие сравнительных таблиц и базовой диагностики; реализация основных моделей и стратегий; отсутствие лишь отдельных элементов (например, SHAP или анализа времени). |
| **Удовлетворительно** | Протестирована одна платформа (например, только AutoGluon); есть сравнение по метрикам; отсутствует диагностика утечки, интерпретируемость или стратегии прогнозирования. |
| **Неудовлетворительно** | Отсутствуют ключевые компоненты работы (данные из работы № 1 не использованы, отсутствует настройка платформ, нет сравнительного анализа, отчёт не представлен). |
