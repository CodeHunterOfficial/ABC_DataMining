## 1. –û—Å–Ω–æ–≤–Ω–∞—è –∏–¥–µ—è PCA

PCA —Å—Ç—Ä–æ–∏—Ç –Ω–æ–≤—É—é —Å–∏—Å—Ç–µ–º—É –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –≤ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ –∏—Å—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö —Ç–∞–∫, —á—Ç–æ–±—ã –ø–µ—Ä–≤–∞—è –æ—Å—å (–≥–ª–∞–≤–Ω–∞—è –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–∞) —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–æ–≤–∞–ª–∞ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—é –Ω–∞–∏–±–æ–ª—å—à–µ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –¥–∞–Ω–Ω—ã—Ö, –≤—Ç–æ—Ä–∞—è ‚Äî –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—é –≤—Ç–æ—Ä–æ–π –ø–æ –≤–µ–ª–∏—á–∏–Ω–µ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –∏ —Ç.–¥. –≠—Ç–∏ –Ω–æ–≤—ã–µ –æ—Å–∏ –æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω—ã –¥—Ä—É–≥ –¥—Ä—É–≥—É.

## 2. –§–æ—Ä–º–∞–ª—å–Ω–∞—è –ø–æ—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–¥–∞—á–∏

–ü—É—Å—Ç—å —É –Ω–∞—Å –µ—Å—Ç—å –º–∞—Ç—Ä–∏—Ü–∞ –¥–∞–Ω–Ω—ã—Ö $ X \in \mathbb{R}^{n \times p} $, –≥–¥–µ:
- $ n $ ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±—ä–µ–∫—Ç–æ–≤,
- $ p $ ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤.

–ö–∞–∂–¥–∞—è —Å—Ç—Ä–æ–∫–∞ –º–∞—Ç—Ä–∏—Ü—ã $ X $ ‚Äî —ç—Ç–æ –≤–µ–∫—Ç–æ—Ä –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –æ–¥–Ω–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞.

**–ó–∞–¥–∞—á–∞ PCA:**
–ù–∞–π—Ç–∏ —Ç–∞–∫–∏–µ –µ–¥–∏–Ω–∏—á–Ω—ã–µ –≤–µ–∫—Ç–æ—Ä—ã $ w_1, w_2, ..., w_k $, —á—Ç–æ –ø—Ä–æ–µ–∫—Ü–∏–∏ –¥–∞–Ω–Ω—ã—Ö –Ω–∞ —ç—Ç–∏ –≤–µ–∫—Ç–æ—Ä—ã –∏–º–µ—é—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é –¥–∏—Å–ø–µ—Ä—Å–∏—é, –∏ –ø—Ä–∏ —ç—Ç–æ–º –≤–µ–∫—Ç–æ—Ä—ã $ w_i $ –æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω—ã –¥—Ä—É–≥ –¥—Ä—É–≥—É.


----

## üîπ –ß—Ç–æ —Ç–∞–∫–æ–µ –µ–¥–∏–Ω–∏—á–Ω—ã–π –≤–µ–∫—Ç–æ—Ä?

–≠—Ç–æ –≤–µ–∫—Ç–æ—Ä –¥–ª–∏–Ω—ã 1:  
$$

$$

–ó–∞—á–µ–º —ç—Ç–æ –Ω—É–∂–Ω–æ? –ß—Ç–æ–±—ã —Å—Ä–∞–≤–Ω–∏–≤–∞—Ç—å –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—è "–Ω–∞ —Ä–∞–≤–Ω—ã—Ö —É—Å–ª–æ–≤–∏—è—Ö". –ò–Ω–∞—á–µ –º–æ–∂–Ω–æ –±—ã–ª–æ –±—ã –ø—Ä–æ—Å—Ç–æ —Ä–∞—Å—Ç—è–≥–∏–≤–∞—Ç—å –≤–µ–∫—Ç–æ—Ä ‚Äî –∏ –¥–∏—Å–ø–µ—Ä—Å–∏—è —Ä–æ—Å–ª–∞ –±—ã —Ç–æ–ª—å–∫–æ –∏–∑-–∑–∞ –µ–≥–æ –¥–ª–∏–Ω—ã, –∞ –Ω–µ –∏–∑-–∑–∞ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—è.


## üî∏ –ß—Ç–æ —Ç–∞–∫–æ–µ –ø—Ä–æ–µ–∫—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –≤–µ–∫—Ç–æ—Ä?

–ü—Ä–µ–¥—Å—Ç–∞–≤—å, —É —Ç–µ–±—è –µ—Å—Ç—å —Ç–æ—á–∫–∞ $ x \in \mathbb{R}^p $ (–≤–µ–∫—Ç–æ—Ä –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –æ–±—ä–µ–∫—Ç–∞), –∏ —Ç—ã —Ö–æ—á–µ—à—å —Å–ø—Ä–æ–µ—Ü–∏—Ä–æ–≤–∞—Ç—å –µ—ë –Ω–∞ –≤–µ–∫—Ç–æ—Ä $ w \in \mathbb{R}^p $. –¢–æ–≥–¥–∞ —Å–∫–∞–ª—è—Ä–Ω–∞—è –ø—Ä–æ–µ–∫—Ü–∏—è –±—É–¥–µ—Ç:

$$
y = x \cdot w
$$

–ï—Å–ª–∏ —É —Ç–µ–±—è –º–Ω–æ–≥–æ —Ç–∞–∫–∏—Ö —Ç–æ—á–µ–∫ (–Ω–∞–ø—Ä–∏–º–µ—Ä, –º–∞—Ç—Ä–∏—Ü–∞ –¥–∞–Ω–Ω—ã—Ö $ X \in \mathbb{R}^{n \times p} $), —Ç–æ –ø—Ä–æ–µ–∫—Ü–∏–∏ –≤—Å–µ—Ö –æ–±—ä–µ–∫—Ç–æ–≤ –Ω–∞ $ w $ ‚Äî —ç—Ç–æ –ø—Ä–æ—Å—Ç–æ –≤–µ–∫—Ç–æ—Ä $ Y = X \cdot w $ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ $ n \times 1 $.

–¢–µ–ø–µ—Ä—å –º–æ–∂–µ–º –ø–æ—Å—á–∏—Ç–∞—Ç—å **–¥–∏—Å–ø–µ—Ä—Å–∏—é —ç—Ç–∏—Ö –ø—Ä–æ–µ–∫—Ü–∏–π** ‚Äî —Ç–æ –µ—Å—Ç—å, –Ω–∞—Å–∫–æ–ª—å–∫–æ —Ç–æ—á–∫–∏ "—Ä–∞–∑–±—Ä–æ—Å–∞–Ω—ã" –≤–¥–æ–ª—å —ç—Ç–æ–≥–æ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—è.


## üî∏ –ü–æ—á–µ–º—É –≤–∞–∂–Ω–∞ **–º–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è**?

–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –º—ã –≤—ã–±–∏—Ä–∞–µ–º —Ç–∞–∫–æ–µ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ, –≤–¥–æ–ª—å –∫–æ—Ç–æ—Ä–æ–≥–æ –¥–∞–Ω–Ω—ã–µ –Ω–∞–∏–±–æ–ª–µ–µ "—Ä–∞–∑–±—Ä–æ—Å–∞–Ω—ã", —Ç.–µ. **–∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–≤–Ω—ã**.

üí° –ß–µ–º –±–æ–ª—å—à–µ –¥–∏—Å–ø–µ—Ä—Å–∏—è ‚Äî —Ç–µ–º –ª—É—á—à–µ —Ä–∞–∑–ª–∏—á–∞—é—Ç—Å—è –æ–±—ä–µ–∫—Ç—ã –≤–¥–æ–ª—å —ç—Ç–æ–≥–æ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—è, –∏ —Ç–µ–º –º–µ–Ω—å—à–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ —Ç–µ—Ä—è–µ—Ç—Å—è –ø—Ä–∏ —Å–Ω–∏–∂–µ–Ω–∏–∏ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏.


## üî∏ –ß—Ç–æ –∑–Ω–∞—á–∏—Ç "–≤–µ–∫—Ç–æ—Ä—ã –æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω—ã –¥—Ä—É–≥ –¥—Ä—É–≥—É"?

–î–≤–∞ –≤–µ–∫—Ç–æ—Ä–∞ $ w_i $ –∏ $ w_j $ –æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω—ã, –µ—Å–ª–∏ –∏—Ö —Å–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —Ä–∞–≤–Ω–æ –Ω—É–ª—é:

$$
w_i^T w_j = 0, \quad \text{–ø—Ä–∏ } i \ne j
$$

–ì–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏ —ç—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –≤–µ–∫—Ç–æ—Ä—ã –ø–µ—Ä–ø–µ–Ω–¥–∏–∫—É–ª—è—Ä–Ω—ã –¥—Ä—É–≥ –¥—Ä—É–≥—É.

### –ó–∞—á–µ–º –Ω—É–∂–Ω–∞ –æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω–æ—Å—Ç—å?

- –≠—Ç–æ –≥–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç, —á—Ç–æ –Ω–æ–≤—ã–µ –æ—Å–∏ (–≥–ª–∞–≤–Ω—ã–µ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã) **–Ω–µ–∑–∞–≤–∏—Å–∏–º—ã** –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞.
- –ü—Ä–æ–µ–∫—Ü–∏–∏ –Ω–∞ —Ä–∞–∑–Ω—ã–µ –≥–ª–∞–≤–Ω—ã–µ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã –Ω–µ –∫–æ—Ä—Ä–µ–ª–∏—Ä—É—é—Ç –º–µ–∂–¥—É —Å–æ–±–æ–π ‚Äî —ç—Ç–æ —É–¥–æ–±–Ω–æ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –∏ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏–∏.
- –û—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω–æ—Å—Ç—å –ø–æ–∑–≤–æ–ª—è–µ—Ç —Å—Ç—Ä–æ–∏—Ç—å –Ω–æ–≤—ã–π –æ—Ä—Ç–æ–Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –±–∞–∑–∏—Å –≤ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ –¥–∞–Ω–Ω—ã—Ö.

## 3. –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö: —Ü–µ–Ω—Ç—Ä–∏—Ä–æ–≤–∞–Ω–∏–µ

–ü–µ—Ä–≤—ã–π —à–∞–≥ ‚Äî **—Ü–µ–Ω—Ç—Ä–∏—Ä–æ–≤–∞–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö**:

$$
\bar{x}_j = \frac{1}{n} \sum_{i=1}^n x_{ij}, \quad j = 1,...,p
$$

–î–∞–ª–µ–µ –≤—ã—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∏–∑ –∫–∞–∂–¥–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞:

$$

–ò–ª–∏ –≤ –º–∞—Ç—Ä–∏—á–Ω–æ–π —Ñ–æ—Ä–º–µ:

$$

$$

–≥–¥–µ $ \mathbf{1}_n $ ‚Äî –≤–µ–∫—Ç–æ—Ä –∏–∑ –µ–¥–∏–Ω–∏—Ü –¥–ª–∏–Ω—ã $ n $, $ \bar{x} $ ‚Äî –≤–µ–∫—Ç–æ—Ä —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø—Ä–∏–∑–Ω–∞–∫–æ–≤.



## 4. –ö–æ–≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞

–ü–æ—Å–ª–µ —Ü–µ–Ω—Ç—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö –≤—ã—á–∏—Å–ª—è–µ–º **–∫–æ–≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É**:

$$
\Sigma = \frac{1}{n - 1} X_{\text{centered}}^T X_{\text{centered}}
$$

–≠—Ç–æ –º–∞—Ç—Ä–∏—Ü–∞ —Ä–∞–∑–º–µ—Ä–∞ $ p \times p $, –≥–¥–µ –∫–∞–∂–¥—ã–π —ç–ª–µ–º–µ–Ω—Ç $ \Sigma_{ij} $ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –∫–æ–≤–∞—Ä–∏–∞—Ü–∏—é –º–µ–∂–¥—É i-–º –∏ j-–º –ø—Ä–∏–∑–Ω–∞–∫–æ–º.

### –°–≤–æ–π—Å—Ç–≤–∞ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã:
- –û–Ω–∞ —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–∞—è ($ \Sigma = \Sigma^T $).
- –í—Å–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –Ω–µ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã (–º–∞—Ç—Ä–∏—Ü–∞ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ –ø–æ–ª—É–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–∞—è).

----


## üìå –ß—Ç–æ —Ç–∞–∫–æ–µ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏—è?

**–ö–æ–≤–∞—Ä–∏–∞—Ü–∏—è** ‚Äî —ç—Ç–æ –º–µ—Ä–∞ **–ª–∏–Ω–µ–π–Ω–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏** –º–µ–∂–¥—É –¥–≤—É–º—è —Å–ª—É—á–∞–π–Ω—ã–º–∏ –≤–µ–ª–∏—á–∏–Ω–∞–º–∏ (–≤ –Ω–∞—à–µ–º —Å–ª—É—á–∞–µ ‚Äî –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏).

–ï—Å–ª–∏ —É –Ω–∞—Å –µ—Å—Ç—å –º–∞—Ç—Ä–∏—Ü–∞ –¥–∞–Ω–Ω—ã—Ö $ X \in \mathbb{R}^{n \times p} $, –≥–¥–µ:
- —Å—Ç—Ä–æ–∫–∏ ‚Äî –æ–±—ä–µ–∫—Ç—ã,
- —Å—Ç–æ–ª–±—Ü—ã ‚Äî –ø—Ä–∏–∑–Ω–∞–∫–∏ (—Ñ–∏—á–∏),

—Ç–æ **–∫–æ–≤–∞—Ä–∏–∞—Ü–∏—è –º–µ–∂–¥—É i-–º –∏ j-–º –ø—Ä–∏–∑–Ω–∞–∫–æ–º** –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, **–Ω–∞—Å–∫–æ–ª—å–∫–æ —ç—Ç–∏ –¥–≤–∞ –ø—Ä–∏–∑–Ω–∞–∫–∞ –ª–∏–Ω–µ–π–Ω–æ —Å–≤—è–∑–∞–Ω—ã –º–µ–∂–¥—É —Å–æ–±–æ–π**.



## üßÆ –§–æ—Ä–º—É–ª–∞ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–∏

–î–ª—è –¥–≤—É—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ $ x_i $ –∏ $ x_j $ (—ç—Ç–æ –¥–≤–∞ —Å—Ç–æ–ª–±—Ü–∞ –º–∞—Ç—Ä–∏—Ü—ã $ X $), –∏—Ö –∫–æ–≤–∞—Ä–∏–∞—Ü–∏—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è —Ç–∞–∫:

$$
\text{Cov}(x_i, x_j) = \frac{1}{n - 1} \sum_{k=1}^n (x_{ki} - \bar{x}_i)(x_{kj} - \bar{x}_j)
$$

–ì–¥–µ:

## 9. –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ã–π –ø–æ–¥—Ö–æ–¥: SVD

## 10. –ü—Ä–∏–º–µ—Ä —à–∞–≥ –∑–∞ —à–∞–≥–æ–º
