### –®–∞–≥ 1: –ù–∞–∑–Ω–∞—á–µ–Ω–∏–µ —Ç–æ—á–µ–∫ –∫ –±–ª–∏–∂–∞–π—à–µ–º—É —Ü–µ–Ω—Ç—Ä–æ–∏–¥—É

### –®–∞–≥ 2: –ü–µ—Ä–µ—Å—á—ë—Ç —Ü–µ–Ω—Ç—Ä–æ–∏–¥–æ–≤

> ‚úÖ –ú—ã —Ö–æ—Ç–∏–º **–º–∏–Ω–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å $ J $** ‚Äî —Å–¥–µ–ª–∞—Ç—å –∫–ª–∞—Å—Ç–µ—Ä—ã –∫–∞–∫ –º–æ–∂–Ω–æ –±–æ–ª–µ–µ –ø–ª–æ—Ç–Ω—ã–º–∏ –∏ –∫–æ–º–ø–∞–∫—Ç–Ω—ã–º–∏.



## üîÅ –ö–∞–∫ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –º–∏–Ω–∏–º–∏–∑–∞—Ü–∏—è?

K-Means –∏—Å–ø–æ–ª—å–∑—É–µ—Ç **–∏—Ç–µ—Ä–∞—Ç–∏–≤–Ω—ã–π –ø—Ä–æ—Ü–µ—Å—Å –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏–∏**, —Å–æ—Å—Ç–æ—è—â–∏–π –∏–∑ –¥–≤—É—Ö —á–µ—Ä–µ–¥—É—é—â–∏—Ö—Å—è —à–∞–≥–æ–≤:



### 1. **–®–∞–≥ –Ω–∞–∑–Ω–∞—á–µ–Ω–∏—è (Assignment Step)**

–ù–∞ —ç—Ç–æ–º —à–∞–≥–µ –º—ã —Ñ–∏–∫—Å–∏—Ä—É–µ–º —Ç–µ–∫—É—â–∏–µ —Ü–µ–Ω—Ç—Ä–æ–∏–¥—ã $ c_1, ..., c_k $ –∏ –ø–µ—Ä–µ—Ä–∞—Å–ø—Ä–µ–¥–µ–ª—è–µ–º —Ç–æ—á–∫–∏ –ø–æ –∫–ª–∞—Å—Ç–µ—Ä–∞–º.

–ö–∞–∂–¥–∞—è —Ç–æ—á–∫–∞ $ x_i $ –Ω–∞–∑–Ω–∞—á–∞–µ—Ç—Å—è –∫ —Ç–æ–º—É –∫–ª–∞—Å—Ç–µ—Ä—É $ j $, –¥–ª—è –∫–æ—Ç–æ—Ä–æ–≥–æ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è:
$$
j = \arg\min_{j'} \|x_i - c_{j'}\|^2
$$

–≠—Ç–æ—Ç —à–∞–≥ **—É–º–µ–Ω—å—à–∞–µ—Ç –∏–Ω–µ—Ä—Ü–∏—é $ J $**, —Ç–∞–∫ –∫–∞–∫ —Ç–µ–ø–µ—Ä—å —Ç–æ—á–∫–∏ –Ω–∞—Ö–æ–¥—è—Ç—Å—è –±–ª–∏–∂–µ –∫ —Å–≤–æ–∏–º —Ü–µ–Ω—Ç—Ä–æ–∏–¥–∞–º.



### 2. **–®–∞–≥ –æ–±–Ω–æ–≤–ª–µ–Ω–∏—è (Update Step)**

–ù–∞ —ç—Ç–æ–º —à–∞–≥–µ –º—ã —Ñ–∏–∫—Å–∏—Ä—É–µ–º –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç—å —Ç–æ—á–µ–∫ –∫ –∫–ª–∞—Å—Ç–µ—Ä–∞–º –∏ –ø–µ—Ä–µ—Å—á–∏—Ç—ã–≤–∞–µ–º –ø–æ–ª–æ–∂–µ–Ω–∏–µ —Ü–µ–Ω—Ç—Ä–æ–∏–¥–æ–≤.

–ù–æ–≤—ã–π —Ü–µ–Ω—Ç—Ä–æ–∏–¥ —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ—Ç—Å—è –∫–∞–∫ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≤—Å–µ—Ö —Ç–æ—á–µ–∫ –≤ –∫–ª–∞—Å—Ç–µ—Ä–µ:
$$
c_j = \frac{1}{|C_j|} \sum_{x_i \in C_j} x_i
$$

–≠—Ç–æ—Ç —à–∞–≥ —Ç–∞–∫–∂–µ **—É–º–µ–Ω—å—à–∞–µ—Ç –∏–Ω–µ—Ä—Ü–∏—é $ J $**, –ø–æ—Ç–æ–º—É —á—Ç–æ –Ω–æ–≤–æ–µ –ø–æ–ª–æ–∂–µ–Ω–∏–µ —Ü–µ–Ω—Ç—Ä–æ–∏–¥–∞ –º–∏–Ω–∏–º–∏–∑–∏—Ä—É–µ—Ç —Å—É–º–º—É –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π –¥–æ –≤—Å–µ—Ö —Ç–æ—á–µ–∫ –∫–ª–∞—Å—Ç–µ—Ä–∞.



## üìê –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–±–æ—Å–Ω–æ–≤–∞–Ω–∏–µ: –ø–æ—á–µ–º—É —ç—Ç–æ —Ä–∞–±–æ—Ç–∞–µ—Ç

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –æ–¥–∏–Ω –∫–ª–∞—Å—Ç–µ—Ä $ C_j $ —Å —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–º–∏ —Ç–æ—á–∫–∞–º–∏ $ x_1, ..., x_n $. –ù–∞–π–¥—ë–º —Ç–∞–∫–æ–µ $ c $, –∫–æ—Ç–æ—Ä–æ–µ –º–∏–Ω–∏–º–∏–∑–∏—Ä—É–µ—Ç:

$$
\sum_{i=1}^{n} \|x_i - c\|^2
$$

–†–∞—Å–∫—Ä—ã–≤–∞–µ–º –∫–≤–∞–¥—Ä–∞—Ç –Ω–æ—Ä–º—ã:
$$
\|x_i - c\|^2 = (x_i - c)^T(x_i - c) = x_i^T x_i - 2 x_i^T c + c^T c
$$

–°—É–º–º–∏—Ä—É–µ–º –ø–æ –≤—Å–µ–º $ i $:
$$
\sum_{i=1}^{n} \|x_i - c\|^2 = \sum_{i=1}^{n} (x_i^T x_i - 2 x_i^T c + c^T c)
$$

–†–∞–∑–±–∏–≤–∞–µ–º –Ω–∞ —á–∞—Å—Ç–∏:
$$
= \sum_{i=1}^{n} x_i^T x_i - 2 c^T \sum_{i=1}^{n} x_i + n c^T c
$$

–ß—Ç–æ–±—ã –Ω–∞–π—Ç–∏ –º–∏–Ω–∏–º—É–º, –≤–æ–∑—å–º—ë–º –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é –ø–æ $ c $ –∏ –ø—Ä–∏—Ä–∞–≤–Ω—è–µ–º –∫ –Ω—É–ª—é:
$$
\frac{\partial}{\partial c} \left( -2 c^T \sum x_i + n c^T c \right) = -2 \sum x_i + 2 n c = 0
$$

–†–µ—à–∞–µ–º:
$$
2 n c = 2 \sum x_i \quad \Rightarrow \quad c = \frac{1}{n} \sum x_i
$$

‚úÖ –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, **–æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–π —Ü–µ–Ω—Ç—Ä–æ–∏–¥ –¥–ª—è —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –∫–ª–∞—Å—Ç–µ—Ä–∞ ‚Äî —ç—Ç–æ –µ–≥–æ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ**.



## üîÑ –ü–æ—á–µ–º—É K-Means —Å—Ö–æ–¥–∏—Ç—Å—è?

–ö–∞–∂–¥—ã–π —à–∞–≥ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –≥–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç, —á—Ç–æ **—Ü–µ–ª–µ–≤–∞—è —Ñ—É–Ω–∫—Ü–∏—è $ J $ –Ω–µ –≤–æ–∑—Ä–∞—Å—Ç–∞–µ—Ç**:

1. –ü—Ä–∏ –Ω–∞–∑–Ω–∞—á–µ–Ω–∏–∏ —Ç–æ—á–µ–∫ –∫ –∫–ª–∞—Å—Ç–µ—Ä–∞–º $ J $ **—É–º–µ–Ω—å—à–∞–µ—Ç—Å—è –∏–ª–∏ –æ—Å—Ç–∞—ë—Ç—Å—è –ø—Ä–µ–∂–Ω–µ–π**
2. –ü—Ä–∏ –æ–±–Ω–æ–≤–ª–µ–Ω–∏–∏ —Ü–µ–Ω—Ç—Ä–æ–∏–¥–æ–≤ $ J $ **—Ç–∞–∫–∂–µ —É–º–µ–Ω—å—à–∞–µ—Ç—Å—è –∏–ª–∏ –æ—Å—Ç–∞—ë—Ç—Å—è –ø—Ä–µ–∂–Ω–µ–π**

–ü–æ—Å–∫–æ–ª—å–∫—É $ J \geq 0 $, –æ–Ω–∞ –Ω–µ –º–æ–∂–µ—Ç —É–º–µ–Ω—å—à–∞—Ç—å—Å—è –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ ‚Äî –∑–Ω–∞—á–∏—Ç, **–∞–ª–≥–æ—Ä–∏—Ç–º –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ —Å–æ–π–¥—ë—Ç—Å—è –∑–∞ –∫–æ–Ω–µ—á–Ω–æ–µ —á–∏—Å–ª–æ —à–∞–≥–æ–≤**.

‚ö†Ô∏è –û–¥–Ω–∞–∫–æ, K-Means –º–æ–∂–µ—Ç —Å–æ–π—Ç–∏—Å—å –∫ **–ª–æ–∫–∞–ª—å–Ω–æ–º—É –º–∏–Ω–∏–º—É–º—É**, –∞ –Ω–µ –∫ –≥–ª–æ–±–∞–ª—å–Ω–æ–º—É. –ü–æ—ç—Ç–æ–º—É —Ä–µ–∫–æ–º–µ–Ω–¥—É–µ—Ç—Å—è –∑–∞–ø—É—Å–∫–∞—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ä–∞–∑ —Å —Ä–∞–∑–Ω—ã–º–∏ –Ω–∞—á–∞–ª—å–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ —Ü–µ–Ω—Ç—Ä–æ–∏–¥–æ–≤ –∏ –≤—ã–±–∏—Ä–∞—Ç—å –ª—É—á—à–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç.



## üß† –í—ã–≤–æ–¥: –∫–∞–∫ –º–∏–Ω–∏–º–∏–∑–∏—Ä—É–µ—Ç—Å—è —Ü–µ–ª–µ–≤–∞—è —Ñ—É–Ω–∫—Ü–∏—è?

–ê–ª–≥–æ—Ä–∏—Ç–º K-Means —Ä–∞–±–æ—Ç–∞–µ—Ç –ø–æ —Å–ª–µ–¥—É—é—â–µ–π —Å—Ö–µ–º–µ:

2. **–ü–æ–≤—Ç–æ—Ä—è–µ–º –¥–æ —Å—Ö–æ–¥–∏–º–æ—Å—Ç–∏**:
   - **–®–∞–≥ –Ω–∞–∑–Ω–∞—á–µ–Ω–∏—è**: –∫–∞–∂–¥–∞—è —Ç–æ—á–∫–∞ –Ω–∞–∑–Ω–∞—á–∞–µ—Ç—Å—è –∫ –±–ª–∏–∂–∞–π—à–µ–º—É —Ü–µ–Ω—Ç—Ä–æ–∏–¥—É
   - **–®–∞–≥ –æ–±–Ω–æ–≤–ª–µ–Ω–∏—è**: —Ü–µ–Ω—Ç—Ä–æ–∏–¥—ã –æ–±–Ω–æ–≤–ª—è—é—Ç—Å—è –∫–∞–∫ —Å—Ä–µ–¥–Ω–µ–µ –ø–æ –∫–ª–∞—Å—Ç–µ—Ä—É
3. **–û—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º—Å—è**, –∫–æ–≥–¥–∞ –∏–∑–º–µ–Ω–µ–Ω–∏—è —Ü–µ–Ω—Ç—Ä–æ–∏–¥–æ–≤ —Å—Ç–∞–Ω–æ–≤—è—Ç—Å—è –æ—á–µ–Ω—å –º–∞–ª—ã–º–∏ –∏–ª–∏ –¥–æ—Å—Ç–∏–≥–Ω—É—Ç –ª–∏–º–∏—Ç –∏—Ç–µ—Ä–∞—Ü–∏–π

–ö–∞–∂–¥—ã–π —à–∞–≥ –≥–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç, —á—Ç–æ **$ J $ –Ω–µ –≤–æ–∑—Ä–∞—Å—Ç–∞–µ—Ç**, –∏ –≤ –∫–æ–Ω–µ—á–Ω–æ–º –∏—Ç–æ–≥–µ –º—ã –ø–æ–ª—É—á–∞–µ–º –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫—É, –≥–¥–µ —Ç–æ—á–∫–∏ –≤–Ω—É—Ç—Ä–∏ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –±–ª–∏–∑–∫–∏ –¥—Ä—É–≥ –∫ –¥—Ä—É–≥—É.



## üìù –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –∑–∞–º–µ—á–∞–Ω–∏—è

### 1. **–í—ã–±–æ—Ä –Ω–∞—á–∞–ª—å–Ω—ã—Ö —Ü–µ–Ω—Ç—Ä–æ–∏–¥–æ–≤**
- –°–ª—É—á–∞–π–Ω–∞—è –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –º–æ–∂–µ—Ç –ø—Ä–∏–≤–µ—Å—Ç–∏ –∫ –ø–ª–æ—Ö–æ–º—É —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É
- –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ **K-Means++** –¥–ª—è —É–ª—É—á—à–µ–Ω–∏—è –∫–∞—á–µ—Å—Ç–≤–∞ –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏

### 2. **–ß—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∫ –≤—ã–±—Ä–æ—Å–∞–º**
- K-Means —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª–µ–Ω –∫ –≤—ã–±—Ä–æ—Å–∞–º, —Ç–∞–∫ –∫–∞–∫ –æ–Ω–∏ —Å–∏–ª—å–Ω–æ –≤–ª–∏—è—é—Ç –Ω–∞ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
- –†–µ–∫–æ–º–µ–Ω–¥—É–µ—Ç—Å—è –ø—Ä–µ–¥–≤–∞—Ä–∏—Ç–µ–ª—å–Ω–æ –æ—á–∏—â–∞—Ç—å –¥–∞–Ω–Ω—ã–µ –∏–ª–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å **K-Medoids (PAM)**

### 3. **–ú–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤**
- –ê–ª–≥–æ—Ä–∏—Ç–º —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª–µ–Ω –∫ –º–∞—Å—à—Ç–∞–±—É –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
- –ü–µ—Ä–µ–¥ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ–º **–≤—Å–µ–≥–¥–∞ –Ω–æ—Ä–º–∞–ª–∏–∑—É–π—Ç–µ –∏–ª–∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∏—Ä—É–π—Ç–µ –¥–∞–Ω–Ω—ã–µ**

### 4. **–§–æ—Ä–º–∞ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤**
- K-Means —Ö–æ—Ä–æ—à–æ —Ä–∞–±–æ—Ç–∞–µ—Ç —Ç–æ–ª—å–∫–æ —Å **–≤—ã–ø—É–∫–ª—ã–º–∏ –∏ —Å—Ñ–µ—Ä–∏—á–µ—Å–∫–∏–º–∏ –∫–ª–∞—Å—Ç–µ—Ä–∞–º–∏**
- –î–ª—è –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω—ã—Ö —Ñ–æ—Ä–º –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ –ª—É—á—à–µ –ø–æ–¥–æ–π–¥—É—Ç –º–µ—Ç–æ–¥—ã –≤—Ä–æ–¥–µ **DBSCAN**, **Spectral Clustering**



## üìà –ü—Ä–∏–º–µ—Ä—ã –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è

### –ü—Ä–∏–º–µ—Ä 1: –ö–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö Iris

```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data

model = KMeans(n_clusters=3)
model.fit(X)
labels = model.predict(X)
```

### –ü—Ä–∏–º–µ—Ä 2: –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –∫–ª–∞—Å—Ç–µ—Ä–æ–≤

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s=300, c='red', label='Centroids')
plt.legend()
plt.show()
```
