#–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–µ –æ—Å–Ω–æ–≤—ã –º–µ—Ç–æ–¥–∞ –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç (Principal Component Analysis, PCA)



## –í–≤–µ–¥–µ–Ω–∏–µ

–ú–µ—Ç–æ–¥ –ì–ª–∞–≤–Ω—ã—Ö –ö–æ–º–ø–æ–Ω–µ–Ω—Ç (PCA ‚Äî Principal Component Analysis) —è–≤–ª—è–µ—Ç—Å—è –æ–¥–Ω–∏–º –∏–∑ –∫–ª—é—á–µ–≤—ã—Ö –º–µ—Ç–æ–¥–æ–≤ –ª–∏–Ω–µ–π–Ω–æ–≥–æ –ø–æ–Ω–∏–∂–µ–Ω–∏—è —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ –≤ –∞–Ω–∞–ª–∏–∑–µ –¥–∞–Ω–Ω—ã—Ö. –û–Ω –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –¥–ª—è —É–º–µ–Ω—å—à–µ–Ω–∏—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Å–ª—É—á–∞–π–Ω—ã—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö, —É—á–∞—Å—Ç–≤—É—é—â–∏—Ö –≤ –º–æ–¥–µ–ª–∏, –±–µ–∑ –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ–π –ø–æ—Ç–µ—Ä–∏ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏. PCA –ø–æ–∑–≤–æ–ª—è–µ—Ç –Ω–∞—Ö–æ–¥–∏—Ç—å –Ω–æ–≤—ã–µ –æ—Å–∏ (–≥–ª–∞–≤–Ω—ã–µ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã), –∫–æ—Ç–æ—Ä—ã–µ –æ–ø–∏—Å—ã–≤–∞—é—Ç –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –≤ –¥–∞–Ω–Ω—ã—Ö.

–í –¥–∞–Ω–Ω–æ–π –ª–µ–∫—Ü–∏–∏ –º—ã –ø–æ–¥—Ä–æ–±–Ω–æ —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º **–º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–µ –æ—Å–Ω–æ–≤–∞–Ω–∏—è** –º–µ—Ç–æ–¥–∞ PCA, –Ω–∞—á–∏–Ω–∞—è —Å –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π –∏ –∑–∞–∫–∞–Ω—á–∏–≤–∞—è —Ñ–æ—Ä–º–∞–ª—å–Ω—ã–º–∏ –≤—ã–≤–æ–¥–∞–º–∏.



## 1. –§–æ—Ä–º–∞–ª—å–Ω–∞—è –ø–æ—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–¥–∞—á–∏

–ü—É—Å—Ç—å –¥–∞–Ω –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö $ X = \{x_1, x_2, ..., x_n\} $, –≥–¥–µ –∫–∞–∂–¥—ã–π –æ–±—ä–µ–∫—Ç $ x_i \in \mathbb{R}^p $ ‚Äî —ç—Ç–æ $ p $-–º–µ—Ä–Ω—ã–π –≤–µ–∫—Ç–æ—Ä –ø—Ä–∏–∑–Ω–∞–∫–æ–≤. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –¥–∞–Ω–Ω—ã–µ –º–æ–∂–Ω–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç—å –∫–∞–∫ –º–∞—Ç—Ä–∏—Ü—É $ X \in \mathbb{R}^{n \times p} $, –≥–¥–µ:

$$
X =
\begin{bmatrix}
x_{11} & x_{12} & \cdots & x_{1p} \\
x_{21} & x_{22} & \cdots & x_{2p} \\
\vdots & \vdots & \ddots & \vdots \\
x_{n1} & x_{n2} & \cdots & x_{np}
\end{bmatrix}
$$

–ó–∞–¥–∞—á–∞ PCA —Å–æ—Å—Ç–æ–∏—Ç –≤ —Ç–æ–º, —á—Ç–æ–±—ã –Ω–∞–π—Ç–∏ —Ç–∞–∫–æ–µ –æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω–æ–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ –æ—Ç–æ–±—Ä–∞–∂–∞–µ—Ç –∏—Å—Ö–æ–¥–Ω—ã–µ $ p $-–º–µ—Ä–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –≤ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ –º–µ–Ω—å—à–µ–π —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ $ k < p $, —Å–æ—Ö—Ä–∞–Ω–∏–≤ –ø—Ä–∏ —ç—Ç–æ–º –∫–∞–∫ –º–æ–∂–Ω–æ –±–æ–ª—å—à–µ –≤–∞—Ä–∏–∞—Ü–∏–∏ (–¥–∏—Å–ø–µ—Ä—Å–∏–∏) –¥–∞–Ω–Ω—ã—Ö.



## 2. –¶–µ–Ω—Ç—Ä–∏—Ä–æ–≤–∞–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö

–ü–µ—Ä–µ–¥ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ–º PCA –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–æ–≤–µ—Å—Ç–∏ **—Ü–µ–Ω—Ç—Ä–∏—Ä–æ–≤–∞–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö**, —Ç–æ –µ—Å—Ç—å –≤—ã—á–µ—Å—Ç—å —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∫–∞–∂–¥–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞. –≠—Ç–æ –¥–µ–ª–∞–µ—Ç—Å—è –¥–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∞—è –º–æ–¥–µ–ª—å –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ —Ä–∞–±–æ—Ç–∞–ª–∞ —Å –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω–æ–π –º–∞—Ç—Ä–∏—Ü–µ–π.

–û–±–æ–∑–Ω–∞—á–∏–º:
$$
\bar{x}_j = \frac{1}{n} \sum_{i=1}^n x_{ij}, \quad j = 1, ..., p
$$

–¢–æ–≥–¥–∞ —Ü–µ–Ω—Ç—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –±—É–¥—É—Ç –∏–º–µ—Ç—å –≤–∏–¥:
$$
\tilde{x}_{ij} = x_{ij} - \bar{x}_j
$$

–ò–ª–∏ –≤ –º–∞—Ç—Ä–∏—á–Ω–æ–º –≤–∏–¥–µ:
$$
\tilde{X} = X - \mathbf{1}_n \cdot \bar{x}^\top
$$
–≥–¥–µ $ \mathbf{1}_n $ ‚Äî –≤–µ–∫—Ç–æ—Ä –∏–∑ $ n $ –µ–¥–∏–Ω–∏—Ü, –∞ $ \bar{x} $ ‚Äî –≤–µ–∫—Ç–æ—Ä —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø—Ä–∏–∑–Ω–∞–∫–æ–≤.



## 3. –ö–æ–≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞

–ö–æ–≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ –∏–≥—Ä–∞–µ—Ç —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω—É—é —Ä–æ–ª—å –≤ PCA. –û–Ω–∞ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –∫–∞–∫ —Å–∏–ª—å–Ω–æ –∏–∑–º–µ–Ω—è—é—Ç—Å—è –º–µ–∂–¥—É —Å–æ–±–æ–π —Ä–∞–∑–ª–∏—á–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏.

–î–ª—è —Ü–µ–Ω—Ç—Ä–∏—Ä–æ–≤–∞–Ω–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã –¥–∞–Ω–Ω—ã—Ö $ \tilde{X} $ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ $ C \in \mathbb{R}^{p \times p} $ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –∫–∞–∫:

$$
C = \frac{1}{n - 1} \tilde{X}^\top \tilde{X}
$$

–≠—Ç–∞ –º–∞—Ç—Ä–∏—Ü–∞:

- —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–∞—è,
- –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ –ø–æ–ª—É–æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–∞—è,
- –∏–º–µ–µ—Ç –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∏ –æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω—ã–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –≤–µ–∫—Ç–æ—Ä—ã.


## 4. –ò–¥–µ—è –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç

–ú–µ—Ç–æ–¥ –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç (Principal Component Analysis, PCA) –æ—Å–Ω–æ–≤–∞–Ω –Ω–∞ –ø–æ–∏—Å–∫–µ —Ç–∞–∫–∏—Ö –æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω—ã—Ö –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–π (–≤–µ–∫—Ç–æ—Ä–æ–≤) –≤ –∏—Å—Ö–æ–¥–Ω–æ–º $ p $-–º–µ—Ä–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –≤–¥–æ–ª—å –∫–æ—Ç–æ—Ä—ã—Ö –ø—Ä–æ–µ–∫—Ü–∏–∏ –æ–±—ä–µ–∫—Ç–æ–≤ –¥–∞–Ω–Ω—ã—Ö –∏–º–µ—é—Ç **–º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é –≤—ã–±–æ—Ä–æ—á–Ω—É—é –¥–∏—Å–ø–µ—Ä—Å–∏—é**. –≠—Ç–∏ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—è –Ω–∞–∑—ã–≤–∞—é—Ç—Å—è **–≥–ª–∞–≤–Ω—ã–º–∏ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–∞–º–∏**, –∏ –æ–Ω–∏ –ø–æ–∑–≤–æ–ª—è—é—Ç –≤—ã–¥–µ–ª–∏—Ç—å –Ω–∞–∏–±–æ–ª–µ–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–≤–Ω—ã–µ –ª–∏–Ω–µ–π–Ω—ã–µ –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏ –∏—Å—Ö–æ–¥–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤.

–ü—É—Å—Ç—å $ X = \{x_1, x_2, \dots, x_n\} $ ‚Äî –º–Ω–æ–∂–µ—Å—Ç–≤–æ —Ü–µ–Ω—Ç—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –Ω–∞–±–ª—é–¥–µ–Ω–∏–π, –≥–¥–µ –∫–∞–∂–¥—ã–π –æ–±—ä–µ–∫—Ç $ x_i \in \mathbb{R}^p $. –†–∞—Å—Å–º–æ—Ç—Ä–∏–º –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω—ã–π –µ–¥–∏–Ω–∏—á–Ω—ã–π –≤–µ–∫—Ç–æ—Ä $ v \in \mathbb{R}^p $, —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—é—â–∏–π —É—Å–ª–æ–≤–∏—é –Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω–æ—Å—Ç–∏:  
$$
\|v\| = 1.
$$

–ü—Ä–æ–µ–∫—Ü–∏—è –∫–∞–∂–¥–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞ $ x_i $ –Ω–∞ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ $ v $ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –∫–∞–∫ —Å–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ:
$$
z_i = x_i^\top v.
$$

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –º—ã –ø–æ–ª—É—á–∞–µ–º –æ–¥–Ω–æ–º–µ—Ä–Ω–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –≤–¥–æ–ª—å –∑–∞–¥–∞–Ω–Ω–æ–≥–æ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—è $ v $. –ß—Ç–æ–±—ã –≤—ã–±—Ä–∞—Ç—å —Ç–∞–∫–æ–µ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ –±—ã –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ —Ö–æ—Ä–æ—à–æ –æ—Ç—Ä–∞–∂–∞–ª–æ –≤–∞—Ä–∏–∞—Ü–∏—é –¥–∞–Ω–Ω—ã—Ö, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –º–∞–∫—Å–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å –≤—ã–±–æ—Ä–æ—á–Ω—É—é –¥–∏—Å–ø–µ—Ä—Å–∏—é –ø—Ä–æ–µ–∫—Ü–∏–π $ z_i $.

–§–æ—Ä–º–∞–ª—å–Ω–æ –∑–∞–¥–∞—á–∞ —Ñ–æ—Ä–º—É–ª–∏—Ä—É–µ—Ç—Å—è —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º:

$$
\max_{v \in \mathbb{R}^p,\; \|v\|=1} \left( \frac{1}{n - 1} \sum_{i=1}^{n} (x_i^\top v)^2 \right).
$$

–≠—Ç–∞ –∑–∞–¥–∞—á–∞ –º–æ–∂–µ—Ç –±—ã—Ç—å –ø–µ—Ä–µ–ø–∏—Å–∞–Ω–∞ –≤ –º–∞—Ç—Ä–∏—á–Ω–æ–π —Ñ–æ—Ä–º–µ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã $ C $, –∫–æ—Ç–æ—Ä–∞—è –¥–ª—è —Ü–µ–Ω—Ç—Ä–∏—Ä–æ–≤–∞–Ω–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã –¥–∞–Ω–Ω—ã—Ö $ \tilde{X} \in \mathbb{R}^{n \times p} $ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –∫–∞–∫:
$$
C = \frac{1}{n - 1} \tilde{X}^\top \tilde{X}.
$$

–° —É—á–µ—Ç–æ–º —ç—Ç–æ–≥–æ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ –¥–ª—è –≤—ã–±–æ—Ä–æ—á–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –≤–∏–¥:
$$
\frac{1}{n - 1} \sum_{i=1}^{n} (x_i^\top v)^2 = v^\top C v.
$$

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –∑–∞–¥–∞—á–∞ PCA —Å–≤–æ–¥–∏—Ç—Å—è –∫ –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—é —Ç–∞–∫–æ–≥–æ –µ–¥–∏–Ω–∏—á–Ω–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞ $ v $, –∫–æ—Ç–æ—Ä—ã–π –æ–±–µ—Å–ø–µ—á–∏–≤–∞–µ—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–π —Ñ–æ—Ä–º—ã $ v^\top C v $, —Ç–æ –µ—Å—Ç—å:

$$
\max_{v:\; \|v\|=1} v^\top C v.
$$



## 5. –†–µ—à–µ–Ω–∏–µ –∑–∞–¥–∞—á–∏ —á–µ—Ä–µ–∑ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ —á–∏—Å–ª–∞ –∏ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –≤–µ–∫—Ç–æ—Ä—ã

–ó–∞–¥–∞—á–∞ –º–∞–∫—Å–∏–º–∏–∑–∞—Ü–∏–∏ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–π —Ñ–æ—Ä–º—ã $ v^\top C v $ –ø—Ä–∏ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–∏ $ v^\top v = 1 $ –æ—Ç–Ω–æ—Å–∏—Ç—Å—è –∫ –∫–ª–∞—Å—Å—É –∑–∞–¥–∞—á –Ω–∞ **—É—Å–ª–æ–≤–Ω—ã–π —ç–∫—Å—Ç—Ä–µ–º—É–º**. –î–ª—è –µ—ë —Ä–µ—à–µ–Ω–∏—è –ø—Ä–∏–º–µ–Ω—è–µ—Ç—Å—è –º–µ—Ç–æ–¥ –º–Ω–æ–∂–∏—Ç–µ–ª–µ–π –õ–∞–≥—Ä–∞–Ω–∂–∞.

–í–≤–µ–¥—ë–º —Ñ—É–Ω–∫—Ü–∏—é –õ–∞–≥—Ä–∞–Ω–∂–∞:


## üßÆ –§–æ—Ä–º–∞–ª—å–Ω–æ–µ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –õ–∞–≥—Ä–∞–Ω–∂–∞

–¶–µ–ª–µ–≤–∞—è —Ñ—É–Ω–∫—Ü–∏—è:  
$$
J(v) = v^\top C v
$$

–û–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ:  
$$
g(v) = v^\top v - 1 = 0
$$

–í–≤–æ–¥–∏–º **—Ñ—É–Ω–∫—Ü–∏—é –õ–∞–≥—Ä–∞–Ω–∂–∞**:

$$
L(v, \lambda) = v^\top C v - \lambda (v^\top v - 1)
$$

–ó–¥–µ—Å—å:
- $ v $ ‚Äî –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–π –≤–µ–∫—Ç–æ—Ä (–≥–ª–∞–≤–Ω–∞—è –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–∞),
- $ \lambda $ ‚Äî —á–∏—Å–ª–æ–≤–æ–π –ø–∞—Ä–∞–º–µ—Ç—Ä, –Ω–∞–∑—ã–≤–∞–µ–º—ã–π **–º–Ω–æ–∂–∏—Ç–µ–ª–µ–º –õ–∞–≥—Ä–∞–Ω–∂–∞**,
- –í—ã—á–∏—Ç–∞–Ω–∏–µ –Ω—É–∂–Ω–æ, —á—Ç–æ–±—ã –ø—Ä–∏ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–º –º–Ω–æ–∂–∏—Ç–µ–ª–µ —à—Ç—Ä–∞—Ñ–æ–≤–∞—Ç—å –∑–∞ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –æ—Ç —É—Å–ª–æ–≤–∏—è.


–î–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è —ç–∫—Å—Ç—Ä–µ–º—É–º–∞ –≤–æ–∑—å–º—ë–º —á–∞—Å—Ç–Ω—É—é –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é —Ñ—É–Ω–∫—Ü–∏–∏ $ L $ –ø–æ $ v $ –∏ –ø—Ä–∏—Ä–∞–≤–Ω—è–µ–º –µ—ë –∫ –Ω—É–ª—é:
$$
\frac{\partial L}{\partial v} = 2Cv - 2\lambda v = 0.
$$

–£–ø—Ä–æ—â–∞—è, –ø–æ–ª—É—á–∞–µ–º:
$$

$$

–≠—Ç–æ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ—Ç —Å–æ–±–æ–π **–∑–∞–¥–∞—á—É –Ω–∞ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∏ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –≤–µ–∫—Ç–æ—Ä—ã –º–∞—Ç—Ä–∏—Ü—ã $ C $**. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, —Ä–µ—à–µ–Ω–∏–µ –∑–∞–¥–∞—á–∏ PCA –∑–∞–∫–ª—é—á–∞–µ—Ç—Å—è –≤ –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏–∏ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã $ C $, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö **–Ω–∞–∏–±–æ–ª—å—à–∏–º —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏—è–º**.

- –ü–µ—Ä–≤—ã–π —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–π –≤–µ–∫—Ç–æ—Ä $ v_1 $, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–π –Ω–∞–∏–±–æ–ª—å—à–µ–º—É —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é $ \lambda_1 $, –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç **–ø–µ—Ä–≤—É—é –≥–ª–∞–≤–Ω—É—é –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—É** ‚Äî –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ –≤ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –≤–¥–æ–ª—å –∫–æ—Ç–æ—Ä–æ–≥–æ –¥–∞–Ω–Ω—ã–µ –∏–º–µ—é—Ç –Ω–∞–∏–±–æ–ª—å—à—É—é –¥–∏—Å–ø–µ—Ä—Å–∏—é.
- –í—Ç–æ—Ä–æ–π —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–π –≤–µ–∫—Ç–æ—Ä $ v_2 $, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–π —Å–ª–µ–¥—É—é—â–µ–º—É –ø–æ –≤–µ–ª–∏—á–∏–Ω–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é $ \lambda_2 $, –∑–∞–¥–∞—ë—Ç **–≤—Ç–æ—Ä—É—é –≥–ª–∞–≤–Ω—É—é –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—É**, –æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω—É—é –ø–µ—Ä–≤–æ–π, –∏ —Ç–∞–∫ –¥–∞–ª–µ–µ.

–ú–∞—Ç—Ä–∏—Ü–∞ $ C $ —è–≤–ª—è–µ—Ç—Å—è —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–æ–π –∏ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ –ø–æ–ª—É–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ–π, —á—Ç–æ –≥–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏–µ –ø–æ–ª–Ω–æ–≥–æ –Ω–∞–±–æ—Ä–∞ –æ—Ä—Ç–æ–Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ –∏ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –Ω–µ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π. –≠—Ç–æ –ø–æ–∑–≤–æ–ª—è–µ—Ç –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω—ã–π –±–∞–∑–∏—Å, –∞–¥–∞–ø—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –∫ —Å—Ç—Ä—É–∫—Ç—É—Ä–µ –¥–∞–Ω–Ω—ã—Ö.


## 6. –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –Ω–æ–≤—ã—Ö –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç

–ü—É—Å—Ç—å –≤—ã–±—Ä–∞–Ω–æ $ k < p $ –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç $ v_1, v_2, ..., v_k $. –°–æ—Å—Ç–∞–≤–∏–º –º–∞—Ç—Ä–∏—Ü—É $ V_k = [v_1\; v_2\; \cdots \; v_k] \in \mathbb{R}^{p \times k} $.

–¢–æ–≥–¥–∞ –ø—Ä–æ–µ–∫—Ü–∏—è –∏—Å—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –Ω–æ–≤–æ–µ –ø–æ–¥–ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ:

$$
Z = \tilde{X} V_k \in \mathbb{R}^{n \times k}
$$

–ö–∞–∂–¥–∞—è —Å—Ç—Ä–æ–∫–∞ –º–∞—Ç—Ä–∏—Ü—ã $ Z $ ‚Äî —ç—Ç–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ –∏—Å—Ö–æ–¥–Ω–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞ –≤ –Ω–æ–≤–æ–º $ k $-–º–µ—Ä–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ.



## 7. –û–±—ä—è—Å–Ω—ë–Ω–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è

–°–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –º–∞—Ç—Ä–∏—Ü—ã $ C $ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä—É—é—Ç—Å—è –∫–∞–∫ –æ–±—ä—è—Å–Ω—ë–Ω–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è –ø–æ –∫–∞–∂–¥–æ–º—É –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—é. –ß–µ–º –±–æ–ª—å—à–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ, —Ç–µ–º –±–æ–ª—å—à–µ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ "–Ω–µ—Å—ë—Ç" —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∞—è –≥–ª–∞–≤–Ω–∞—è –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–∞.

–û–±—â–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è —Ä–∞–≤–Ω–∞ —Å–ª–µ–¥—É –º–∞—Ç—Ä–∏—Ü—ã $ C $:

$$
\text{Tr}(C) = \sum_{i=1}^p \lambda_i
$$

–ü—Ä–æ—Ü–µ–Ω—Ç –æ–±—ä—è—Å–Ω—ë–Ω–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –¥–ª—è –ø–µ—Ä–≤—ã—Ö $ k $ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç:

$$
\frac{\sum_{i=1}^k \lambda_i}{\sum_{i=1}^p \lambda_i}
$$

–≠—Ç–æ –≤–∞–∂–Ω–∞—è –º–µ—Ç—Ä–∏–∫–∞ –ø—Ä–∏ –≤—ã–±–æ—Ä–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç $ k $.



## 8. –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ã–π –ø–æ–¥—Ö–æ–¥: —Å–∏–Ω–≥—É–ª—è—Ä–Ω–æ–µ —Ä–∞–∑–ª–æ–∂–µ–Ω–∏–µ (SVD)

PCA –º–æ–∂–µ—Ç –±—ã—Ç—å —Ç–∞–∫–∂–µ —Ä–µ–∞–ª–∏–∑–æ–≤–∞–Ω —á–µ—Ä–µ–∑ **—Å–∏–Ω–≥—É–ª—è—Ä–Ω–æ–µ —Ä–∞–∑–ª–æ–∂–µ–Ω–∏–µ** –º–∞—Ç—Ä–∏—Ü—ã –¥–∞–Ω–Ω—ã—Ö $ \tilde{X} $.

–ü—É—Å—Ç—å $ \tilde{X} \in \mathbb{R}^{n \times p} $ ‚Äî —Ü–µ–Ω—Ç—Ä–∏—Ä–æ–≤–∞–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ –¥–∞–Ω–Ω—ã—Ö. –¢–æ–≥–¥–∞ –µ—ë SVD –∏–º–µ–µ—Ç –≤–∏–¥:

$$
\tilde{X} = U \Sigma V^\top
$$

–≥–¥–µ:

- $ U \in \mathbb{R}^{n \times n} $ ‚Äî –æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ –ª–µ–≤—ã—Ö —Å–∏–Ω–≥—É–ª—è—Ä–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤,
- $ \Sigma \in \mathbb{R}^{n \times p} $ ‚Äî –¥–∏–∞–≥–æ–Ω–∞–ª—å–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ —Å–∏–Ω–≥—É–ª—è—Ä–Ω—ã—Ö —á–∏—Å–µ–ª $ \sigma_i $,
- $ V \in \mathbb{R}^{p \times p} $ ‚Äî –æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ –ø—Ä–∞–≤—ã—Ö —Å–∏–Ω–≥—É–ª—è—Ä–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤.

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –≥–ª–∞–≤–Ω—ã–µ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã ‚Äî —ç—Ç–æ —Å—Ç–æ–ª–±—Ü—ã –º–∞—Ç—Ä–∏—Ü—ã $ V $, –∞ –ø—Ä–æ–µ–∫—Ü–∏–∏ –¥–∞–Ω–Ω—ã—Ö ‚Äî —ç—Ç–æ $ \tilde{X} V_k = U_k \Sigma_k $, –≥–¥–µ $ U_k $ –∏ $ \Sigma_k $ ‚Äî –ø–µ—Ä–≤—ã–µ $ k $ —Å—Ç–æ–ª–±—Ü–æ–≤ –∏ –¥–∏–∞–≥–æ–Ω–∞–ª—å–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ.



## 9. –ì–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∞—è –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è

- –ü–µ—Ä–≤–∞—è –≥–ª–∞–≤–Ω–∞—è –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–∞ ‚Äî —ç—Ç–æ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ, –≤ –∫–æ—Ç–æ—Ä–æ–º –¥–∞–Ω–Ω—ã–µ –∏–º–µ—é—Ç –Ω–∞–∏–±–æ–ª—å—à—É—é –¥–∏—Å–ø–µ—Ä—Å–∏—é.

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, PCA —Å—Ç—Ä–æ–∏—Ç –æ—Ä—Ç–æ–Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –±–∞–∑–∏—Å, –∞–¥–∞–ø—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –∫ —Å—Ç—Ä—É–∫—Ç—É—Ä–µ –¥–∞–Ω–Ω—ã—Ö.

- –ù–µ –≤—Å–µ–≥–¥–∞ —Ö–æ—Ä–æ—à–æ —Ä–∞–±–æ—Ç–∞–µ—Ç —Å –¥–∞–Ω–Ω—ã–º–∏, –∏–º–µ—é—â–∏–º–∏ –Ω–µ–ª–∏–Ω–µ–π–Ω—É—é —Å—Ç—Ä—É–∫—Ç—É—Ä—É.



## –ó–∞–∫–ª—é—á–µ–Ω–∏–µ

–ú–µ—Ç–æ–¥ –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç ‚Äî –º–æ—â–Ω—ã–π –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç –∞–Ω–∞–ª–∏–∑–∞ –¥–∞–Ω–Ω—ã—Ö, –æ—Å–Ω–æ–≤–∞–Ω–Ω—ã–π –Ω–∞ –ª–∏–Ω–µ–π–Ω–æ–π –∞–ª–≥–µ–±—Ä–µ, —Ç–µ–æ—Ä–∏–∏ –º–∞—Ç—Ä–∏—Ü –∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–µ. –ï–≥–æ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –æ–±—É—Å–ª–æ–≤–ª–µ–Ω–∞ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã –∏ —Å–ø–µ–∫—Ç—Ä–∞–ª—å–Ω–æ–≥–æ —Ä–∞–∑–ª–æ–∂–µ–Ω–∏—è. PCA —à–∏—Ä–æ–∫–æ –ø—Ä–∏–º–µ–Ω—è–µ—Ç—Å—è –≤ –º–∞—à–∏–Ω–Ω–æ–º –æ–±—É—á–µ–Ω–∏–∏, –∫–æ–º–ø—å—é—Ç–µ—Ä–Ω–æ–º –∑—Ä–µ–Ω–∏–∏, –±–∏–æ–∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–∫–µ, —Ñ–∏–Ω–∞–Ω—Å–∞—Ö –∏ –¥—Ä—É–≥–∏—Ö –æ–±–ª–∞—Å—Ç—è—Ö.

–í —Å–ª–µ–¥—É—é—â–∏—Ö –ª–µ–∫—Ü–∏—è—Ö –º—ã —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–æ–µ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ PCA, –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—é —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤, –≤—ã–±–æ—Ä —á–∏—Å–ª–∞ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç, —Å–≤—è–∑—å —Å –¥—Ä—É–≥–∏–º–∏ –º–µ—Ç–æ–¥–∞–º–∏ —Å–Ω–∏–∂–µ–Ω–∏—è —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ –∏ –ø—Ä–∏–º–µ—Ä—ã —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏.




# üî¢ –ß–∏—Å–ª–æ–≤–æ–π –ø—Ä–∏–º–µ—Ä PCA

## 1. –ò—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –º–∞—Ç—Ä–∏—Ü—É –æ–±—ä–µ–∫—Ç–æ–≤ $ X \in \mathbb{R}^{3 \times 2} $:

$$
X =
\begin{bmatrix}
2 & 3 \\
4 & 5 \\
6 & 7
\end{bmatrix}
$$

–ó–¥–µ—Å—å —É –Ω–∞—Å 3 –æ–±—ä–µ–∫—Ç–∞ –∏ 2 –ø—Ä–∏–∑–Ω–∞–∫–∞.



## 2. –¶–µ–Ω—Ç—Ä–∏—Ä–æ–≤–∞–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö

–°–Ω–∞—á–∞–ª–∞ –≤—ã—á–∏—Å–ª–∏–º —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ –∫–∞–∂–¥–æ–º—É –ø—Ä–∏–∑–Ω–∞–∫—É:

$$
\bar{x}_1 = \frac{2 + 4 + 6}{3} = 4,\quad \bar{x}_2 = \frac{3 + 5 + 7}{3} = 5
$$

–¢–µ–ø–µ—Ä—å –≤—ã—á—Ç–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∏–∑ –∫–∞–∂–¥–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞:

$$
\tilde{X} =
\begin{bmatrix}
2 - 4 & 3 - 5 \\
4 - 4 & 5 - 5 \\
6 - 4 & 7 - 5
\end{bmatrix}
=
\begin{bmatrix}
-2 & -2 \\
0 & 0 \\
2 & 2
\end{bmatrix}
$$


## 3. –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã

–§–æ—Ä–º—É–ª–∞:
$$
C = \frac{1}{n - 1} \tilde{X}^\top \tilde{X}, \quad n = 3
$$

–í—ã—á–∏—Å–ª–∏–º $ \tilde{X}^\top \tilde{X} $:

$$
\tilde{X}^\top =
\begin{bmatrix}
-2 & 0 & 2 \\
-2 & 0 & 2
\end{bmatrix},\quad
\tilde{X}^\top \tilde{X} =
\begin{bmatrix}
(-2)^2 + 0^2 + 2^2 & (-2)(-2) + 0 + 2 \cdot 2 \\
(-2)(-2) + 0 + 2 \cdot 2 & (-2)^2 + 0^2 + 2^2
\end{bmatrix}
=
\begin{bmatrix}
8 & 8 \\
8 & 8
\end{bmatrix}
$$

–¢–æ–≥–¥–∞ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞:

$$
C = \frac{1}{2} \cdot
\begin{bmatrix}
8 & 8 \\
8 & 8
\end{bmatrix}
=
\begin{bmatrix}
4 & 4 \\
4 & 4
\end{bmatrix}
$$



## 4. –ù–∞—Ö–æ–∂–¥–µ–Ω–∏–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–µ–ª –∏ –≤–µ–∫—Ç–æ—Ä–æ–≤

### a) –•–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏—á–µ—Å–∫–æ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ:

–î–ª—è –º–∞—Ç—Ä–∏—Ü—ã $ C $, –∏—â–µ–º —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∏–∑ —É—Ä–∞–≤–Ω–µ–Ω–∏—è:

$$
\det(C - \lambda I) = 0
$$

$$
\det
\begin{bmatrix}
4 - \lambda & 4 \\
4 & 4 - \lambda
\end{bmatrix}
= (4 - \lambda)^2 - 16 = \lambda^2 - 8\lambda = 0
$$

–†–µ—à–∞–µ–º:

$$
\lambda(\lambda - 8) = 0 \Rightarrow \lambda_1 = 8,\; \lambda_2 = 0
$$

### b) –°–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –≤–µ–∫—Ç–æ—Ä—ã:

#### –î–ª—è $ \lambda_1 = 8 $:

–†–µ—à–∞–µ–º —Å–∏—Å—Ç–µ–º—É:
$$
(C - 8I)v = 0
\Rightarrow
\begin{bmatrix}
-4 & 4 \\
4 & -4
\end{bmatrix}
\begin{bmatrix}
v_1 \\
v_2
\end{bmatrix}
= 0
\Rightarrow -4v_1 + 4v_2 = 0 \Rightarrow v_1 = v_2
$$

–ù–æ—Ä–º–∏—Ä—É–µ–º: $ v_1 = \frac{1}{\sqrt{2}}, v_2 = \frac{1}{\sqrt{2}} $

–ò—Ç–∞–∫, –ø–µ—Ä–≤—ã–π —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–π –≤–µ–∫—Ç–æ—Ä:
$$
v_1 = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$

#### –î–ª—è $ \lambda_2 = 0 $:

–ê–Ω–∞–ª–æ–≥–∏—á–Ω–æ –ø–æ–ª—É—á–∞–µ–º:
$$
(C - 0I)v = Cv = 0 \Rightarrow 4v_1 + 4v_2 = 0 \Rightarrow v_1 = -v_2
$$

–ù–æ—Ä–º–∏—Ä—É–µ–º: $ v_1 = \frac{1}{\sqrt{2}}, v_2 = -\frac{1}{\sqrt{2}} $

–í—Ç–æ—Ä–æ–π —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–π –≤–µ–∫—Ç–æ—Ä:
$$
v_2 = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ -1 \end{bmatrix}
$$

---

## 5. –í—ã–±–æ—Ä –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç

–ü–æ—Å–∫–æ–ª—å–∫—É $ \lambda_1 = 8 > \lambda_2 = 0 $, —Ç–æ –ø–µ—Ä–≤–∞—è –≥–ª–∞–≤–Ω–∞—è –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–∞ ‚Äî —ç—Ç–æ $ v_1 $. –ü—Ä–µ–¥–ø–æ–ª–æ–∂–∏–º, –º—ã —Ö–æ—Ç–∏–º –ø–æ–Ω–∏–∑–∏—Ç—å —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å –¥–æ $ k = 1 $.

–ú–∞—Ç—Ä–∏—Ü–∞ –ø—Ä–æ–µ–∫—Ü–∏–∏:
$$
V_k = v_1 = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$

---

## 6. –ü—Ä–æ–µ–∫—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –ø–µ—Ä–≤—É—é –≥–ª–∞–≤–Ω—É—é –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—É

$$
Z = \tilde{X} V_k =
\begin{bmatrix}
-2 & -2 \\
0 & 0 \\
2 & 2
\end{bmatrix}
\cdot
\frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ 1 \end{bmatrix}
=
\frac{1}{\sqrt{2}}
\begin{bmatrix}
-2 - 2 \\
0 + 0 \\
2 + 2
\end{bmatrix}
=
\frac{1}{\sqrt{2}}
\begin{bmatrix}
-4 \\
0 \\
4
\end{bmatrix}
$$

–¢–æ –µ—Å—Ç—å –Ω–æ–≤—ã–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –≤ –æ–¥–Ω–æ–º–µ—Ä–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ:

$$
Z = \begin{bmatrix}
-4/\sqrt{2} \\
0 \\
4/\sqrt{2}
\end{bmatrix}
$$



## 7. –û–±—ä—è—Å–Ω—ë–Ω–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è

–û–±—â–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è: $ \lambda_1 + \lambda_2 = 8 + 0 = 8 $

–û–±—ä—è—Å–Ω—ë–Ω–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è –¥–ª—è –ø–µ—Ä–≤–æ–π –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã: $ \frac{8}{8} = 1 $ –∏–ª–∏ 100%.

–¢–∞–∫ –∫–∞–∫ –≤—Ç–æ—Ä–æ–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ —Ä–∞–≤–Ω–æ –Ω—É–ª—é, —Ç–æ –≤—Å—ë —Å–æ–¥–µ—Ä–∂–∞–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –º–æ–∂–Ω–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç—å –æ–¥–Ω–æ–π –≥–ª–∞–≤–Ω–æ–π –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–æ–π.


# üìå –ò—Ç–æ–≥–∏

| –≠—Ç–∞–ø | –†–µ–∑—É–ª—å—Ç–∞—Ç |
|------|-----------|
| –¶–µ–Ω—Ç—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ $ \tilde{X} $ | $\begin{bmatrix}-2 & -2\\0 & 0\\2 & 2\end{bmatrix}$ |
| –ö–æ–≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ $ C $ | $\begin{bmatrix}4 & 4\\4 & 4\end{bmatrix}$ |
| –°–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è $ \lambda $ | $ \lambda_1 = 8, \; \lambda_2 = 0 $ |
| –°–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –≤–µ–∫—Ç–æ—Ä—ã $ v $ | $ v_1 = \frac{1}{\sqrt{2}}[1\;1]^\top,\; v_2 = \frac{1}{\sqrt{2}}[1\;-1]^\top $ |
| –ü—Ä–æ–µ–∫—Ü–∏–∏ –Ω–∞ –ø–µ—Ä–≤—É—é –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—É $ Z $ | $[-4/\sqrt{2},\; 0,\; 4/\sqrt{2}]^\top$ |


```python
import numpy as np
import matplotlib.pyplot as plt

class PCA:
    def __init__(self, n_components=2):
        """
        –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è PCA.

        –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
        - n_components: —á–∏—Å–ª–æ –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç –¥–ª—è –ø—Ä–æ–µ—Ü–∏—Ä–æ–≤–∞–Ω–∏—è
        """
        self.n_components = n_components
        self.components = None      # –ì–ª–∞–≤–Ω—ã–µ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã (—Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –≤–µ–∫—Ç–æ—Ä—ã)
        self.explained_variance = None  # –û–±—ä—è—Å–Ω—ë–Ω–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è
        self.explained_variance_ratio = None  # –î–æ–ª—è –æ–±—ä—è—Å–Ω—ë–Ω–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏
        self.mean = None            # –°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ –ø—Ä–∏–∑–Ω–∞–∫–∞–º

    def fit(self, X):
        """
        –û–±—É—á–µ–Ω–∏–µ PCA: –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç.

        –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
        - X: –º–∞—Ç—Ä–∏—Ü–∞ –¥–∞–Ω–Ω—ã—Ö —Ä–∞–∑–º–µ—Ä–æ–º (n_samples, n_features)
        """
        # –¶–µ–Ω—Ç—Ä–∏—Ä–æ–≤–∞–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö
        self.mean = np.mean(X, axis=0)
        X_centered = X - self.mean

        # –ö–æ–≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞
        n = X_centered.shape[0]
        C = (X_centered.T @ X_centered) / (n - 1)

        # –°–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∏ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –≤–µ–∫—Ç–æ—Ä—ã
        eigen_values, eigen_vectors = np.linalg.eig(C)

        # –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –ø–æ —É–±—ã–≤–∞–Ω–∏—é —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
        idx = np.argsort(eigen_values)[::-1]
        eigen_values = eigen_values[idx]
        eigen_vectors = eigen_vectors[:, idx]

        # –°–æ—Ö—Ä–∞–Ω—è–µ–º –ø–µ—Ä–≤—ã–µ n_components —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤
        self.components = eigen_vectors[:, :self.n_components]

        # –°–æ—Ö—Ä–∞–Ω—è–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –¥–∏—Å–ø–µ—Ä—Å–∏–∏
        self.explained_variance = eigen_values[:self.n_components]
        self.explained_variance_ratio = eigen_values[:self.n_components] / np.sum(eigen_values)

    def transform(self, X):
        """
        –ü—Ä–æ–µ–∫—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –≥–ª–∞–≤–Ω—ã–µ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã.

        –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
        - X: –º–∞—Ç—Ä–∏—Ü–∞ –¥–∞–Ω–Ω—ã—Ö —Ä–∞–∑–º–µ—Ä–æ–º (n_samples, n_features)

        –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
        - Z: –ø—Ä–æ–µ–∫—Ü–∏–∏ –≤ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ –º–µ–Ω—å—à–µ–π —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏
        """
        if self.mean is None:
            raise ValueError("PCA –Ω–µ –æ–±—É—á–µ–Ω. –í—ã–∑–æ–≤–∏—Ç–µ –º–µ—Ç–æ–¥ fit –ø–µ—Ä–µ–¥ transform.")

        X_centered = X - self.mean
        Z = X_centered @ self.components
        return Z

    def inverse_transform(self, Z):
        """
        –í–æ—Å—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –∏–∑ –ø—Ä–æ–µ–∫—Ü–∏–π.

        –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
        - Z: –ø—Ä–æ–µ–∫—Ü–∏–∏ –≤ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ –º–µ–Ω—å—à–µ–π —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏

        –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
        - X_reconstructed: –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –≤ –∏—Å—Ö–æ–¥–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ
        """
        if self.mean is None or self.components is None:
            raise ValueError("PCA –Ω–µ –æ–±—É—á–µ–Ω.")

        X_reconstructed = Z @ self.components.T + self.mean
        return X_reconstructed

    def plot_components_and_projections(self, X, title="PCA: –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—è –∫–æ–º–ø–æ–Ω–µ–Ω—Ç –∏ –ø—Ä–æ–µ–∫—Ü–∏–∏"):
        """
        –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö, –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç –∏ –∏—Ö –ø—Ä–æ–µ–∫—Ü–∏–π (—Ç–æ–ª—å–∫–æ –¥–ª—è 2D).
        """
        if X.shape[1] != 2 or self.components.shape[0] != 2:
            raise ValueError("–í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –≤–æ–∑–º–æ–∂–Ω–∞ —Ç–æ–ª—å–∫–æ –¥–ª—è 2D –¥–∞–Ω–Ω—ã—Ö.")

        X_centered = X - self.mean
        Z = self.transform(X)
        projections = self.inverse_transform(Z) - self.mean  # –ø—Ä–æ–µ–∫—Ü–∏–∏ –≤ —Ü–µ–Ω—Ç—Ä–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ

        plt.figure(figsize=(8, 6))
        plt.scatter(X_centered[:, 0], X_centered[:, 1], color='blue', label='–¶–µ–Ω—Ç—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ')

        # –†–∏—Å—É–µ–º –≥–ª–∞–≤–Ω—ã–µ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã
        for i in range(self.n_components):
            v = self.components[:, i]
            scale = np.sqrt(self.explained_variance[i]) * 3  # –º–∞—Å—à—Ç–∞–±–∏—Ä—É–µ–º –¥–ª—è –Ω–∞–≥–ª—è–¥–Ω–æ—Å—Ç–∏
            plt.quiver(0, 0, v[0] * scale, v[1] * scale, angles='xy', scale_units='xy', scale=1,
                       color='red', label=f'PC{i+1}')

        # –†–∏—Å—É–µ–º –ø—Ä–æ–µ–∫—Ü–∏–∏
        plt.scatter(projections[:, 0], projections[:, 1], color='green', label='–ü—Ä–æ–µ–∫—Ü–∏–∏')

        # –°–æ–µ–¥–∏–Ω—è–µ–º —Ç–æ—á–∫–∏ —Å –∏—Ö –ø—Ä–æ–µ–∫—Ü–∏—è–º–∏
        for x, proj in zip(X_centered, projections):
            plt.plot([x[0], proj[0]], [x[1], proj[1]], 'k--', alpha=0.5)

        plt.axhline(0, color='black', linewidth=0.5)
        plt.axvline(0, color='black', linewidth=0.5)
        plt.axis('equal')
        plt.xlabel("–ü—Ä–∏–∑–Ω–∞–∫ 1")
        plt.ylabel("–ü—Ä–∏–∑–Ω–∞–∫ 2")
        plt.title(title)
        plt.legend()
        plt.grid(True)
        plt.show()
```

```python
# –ò—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
X = np.array([
    [2, 3],
    [4, 5],
    [6, 7]
])

# –°–æ–∑–¥–∞–µ–º –∏ –æ–±—É—á–∞–µ–º PCA
pca = PCA(n_components=1)
pca.fit(X)

# –ü—Ä–æ–µ–∫—Ç–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ
Z = pca.transform(X)
print("–ü—Ä–æ–µ–∫—Ü–∏–∏ –Ω–∞ –ø–µ—Ä–≤—É—é –≥–ª–∞–≤–Ω—É—é –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—É:\n", Z)

# –í–æ—Å—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º –¥–∞–Ω–Ω—ã–µ –∏–∑ –ø—Ä–æ–µ–∫—Ü–∏–π
X_reconstructed = pca.inverse_transform(Z)
print("\n–í–æ—Å—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:\n", X_reconstructed)

# –í–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º
pca.plot_components_and_projections(X)
```
