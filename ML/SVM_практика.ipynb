#Как найти гиперплоскость в SVM для задачи классификации

#### Введение
Support Vector Machine (SVM) — это алгоритм машинного обучения, который используется для задач классификации и регрессии. Мы будем рассматривать задачу **бинарной классификации**, то есть разделения данных на два класса.

В этом примере мы научимся находить оптимальную гиперплоскость, которая разделяет два класса в пространстве признаков. Гиперплоскость — это линия (в двумерном пространстве), плоскость (в трёхмерном пространстве) или аналогичная структура в более высоких измерениях.



### Пример: Нахождение гиперплоскости

#### Данные:
У нас есть набор данных с двумя классами:

- Класс $+1$: точки $(1, 1)$, $(2, 2)$,
- Класс $-1$: точки $(-1, -1)$, $(-2, -2)$.

Наша цель — найти уравнение гиперплоскости, которая максимально разделяет эти два класса.



### Шаг 1: Общее уравнение гиперплоскости

В двумерном пространстве гиперплоскость — это прямая. Уравнение прямой можно записать как:

$$
w_1 x_1 + w_2 x_2 + b = 0
$$

где:
- $x_1$ и $x_2$ — координаты точки,
- $w_1$ и $w_2$ — веса (компоненты нормального вектора к гиперплоскости),
- $b$ — смещение (bias).

Это уравнение можно переписать в векторной форме:

$$
w^T x + b = 0
$$

где:
- $w = [w_1, w_2]$ — вектор весов,
- $x = [x_1, x_2]$ — вектор координат точки.



### Шаг 2: Определение маржи (margin)

Основная идея SVM заключается в том, чтобы найти такую гиперплоскость, которая **максимально разделяет** два класса. Для этого используется концепция **маржи**.

Маржа — это расстояние между гиперплоскостью и ближайшими точками из каждого класса. Эти ближайшие точки называются **опорными векторами**.

Расстояние от точки $x_i$ до гиперплоскости вычисляется по формуле:

$$
d = \frac{|w^T x_i + b|}{||w||}
$$

где $||w|| = \sqrt{w_1^2 + w_2^2}$ — евклидова норма вектора $w$.

Чтобы максимизировать маржу, нужно минимизировать $||w||$, при этом удовлетворяя условию корректной классификации всех точек.



### Шаг 3: Формулировка задачи оптимизации

Цель SVM — решить следующую задачу оптимизации:

$$
\min_{w, b} \frac{1}{2} ||w||^2
$$

при условиях:

$$
y_i (w^T x_i + b) \geq 1, \quad \forall i
$$

где:
- $y_i$ — метка класса ($+1$ или $-1$),
- $x_i$ — координаты $i$-й точки.

Эти условия означают, что каждая точка должна быть правильно классифицирована, и её расстояние до гиперплоскости должно быть не меньше единицы (с учётом масштабирования).



### Шаг 4: Применение условий к данным

Подставим наши данные:

1. Для точки $(1, 1)$ с меткой $+1$:
   $$
   w_1(1) + w_2(1) + b \geq 1
   $$
   $$
   w_1 + w_2 + b \geq 1
   $$

2. Для точки $(2, 2)$ с меткой $+1$:
   $$
   w_1(2) + w_2(2) + b \geq 1
   $$
   $$
   2w_1 + 2w_2 + b \geq 1
   $$

3. Для точки $(-1, -1)$ с меткой $-1$:
   $$
   w_1(-1) + w_2(-1) + b \leq -1
   $$
   $$
   -w_1 - w_2 + b \leq -1
   $$

4. Для точки $(-2, -2)$ с меткой $-1$:
   $$
   w_1(-2) + w_2(-2) + b \leq -1
   $$
   $$
   -2w_1 - 2w_2 + b \leq -1
   $$



### Шаг 5: Решение системы неравенств

Для упрощения предположим, что $w_1 = w_2 = w$ (это допустимо, так как данные симметричны). Тогда уравнение гиперплоскости примет вид:

$$
w(x_1 + x_2) + b = 0
$$

Подставим это предположение в неравенства:

1. $w(1 + 1) + b \geq 1$:
   $$
   2w + b \geq 1
   $$

2. $w(2 + 2) + b \geq 1$:
   $$
   4w + b \geq 1
   $$

3. $w(-1 - 1) + b \leq -1$:
   $$
   -2w + b \leq -1
   $$

4. $w(-2 - 2) + b \leq -1$:
   $$
   -4w + b \leq -1
   $$

Теперь решаем систему неравенств:

- Из $2w + b \geq 1$ и $-2w + b \leq -1$:
  $$
  b = 0, \quad w = \frac{1}{2}
  $$



### Шаг 6: Итоговое уравнение гиперплоскости

Подставим найденные значения $w = \frac{1}{2}$ и $b = 0$ в уравнение гиперплоскости:

$$
\frac{1}{2}(x_1 + x_2) = 0
$$

или эквивалентно:

$$
x_1 + x_2 = 0
$$



### Проверка решения

1. Для точки $(1, 1)$:
   $$
   1 + 1 = 2 > 0 \quad (\text{правильно классифицирована как } +1)
   $$

2. Для точки $(2, 2)$:
   $$
   2 + 2 = 4 > 0 \quad (\text{правильно классифицирована как } +1)
   $$

3. Для точки $(-1, -1)$:
   $$
   -1 - 1 = -2 < 0 \quad (\text{правильно классифицирована как } -1)
   $$

4. Для точки $(-2, -2)$:
   $$
   -2 - 2 = -4 < 0 \quad (\text{правильно классифицирована как } -1)
   $$



### Ответ

Итоговое уравнение гиперплоскости:

$$
\boxed{x_1 + x_2 = 0}
$$



```python
import numpy as np
import matplotlib.pyplot as plt

# Шаг 1: Определение данных
# Класс +1
X_plus = np.array([[1, 1], [2, 2]])
y_plus = np.array([1, 1])

# Класс -1
X_minus = np.array([[-1, -1], [-2, -2]])
y_minus = np.array([-1, -1])

# Объединяем данные
X = np.vstack((X_plus, X_minus))
y = np.hstack((y_plus, y_minus))

# Шаг 2: Находим параметры гиперплоскости
# Предположим, что w1 = w2 = w, b = 0 (симметричные данные)
w = np.array([0.5, 0.5])  # w1 = w2 = 0.5
b = 0  # Смещение

# Уравнение гиперплоскости: w1*x1 + w2*x2 + b = 0
# Преобразуем к виду x2 = -w1/w2 * x1 - b/w2
# Здесь w1 = w2, поэтому упрощается до x2 = -x1

# Генерируем координаты для отображения гиперплоскости
x1_range = np.linspace(-3, 3, 100)  # Диапазон значений x1
x2_hyperplane = -x1_range  # x2 = -x1

# Шаг 3: Визуализация
plt.figure(figsize=(8, 6))

# Отображаем точки класса +1
plt.scatter(X_plus[:, 0], X_plus[:, 1], color='blue', label='Класс +1')

# Отображаем точки класса -1
plt.scatter(X_minus[:, 0], X_minus[:, 1], color='red', label='Класс -1')

# Отображаем гиперплоскость
plt.plot(x1_range, x2_hyperplane, color='green', label='Гиперплоскость (x1 + x2 = 0)', linewidth=2)

# Добавляем легенду и метки осей
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # Горизонтальная линия x2 = 0
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # Вертикальная линия x1 = 0
plt.legend()
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Гиперплоскость SVM')
plt.grid(True)
plt.axis('equal')  # Сохраняем пропорции осей
plt.show()

# Шаг 4: Проверка решения
# Функция для проверки классификации
def check_classification(w, b, X, y):
    predictions = np.sign(np.dot(X, w) + b)
    correct = np.sum(predictions == y)
    total = len(y)
    return correct, total

correct, total = check_classification(w, b, X, y)
print(f"Правильно классифицировано {correct} из {total} точек.")
```

#### Введение
Support Vector Machine (SVM) может использоваться не только для задач классификации, но и для задач **регрессии**. В случае регрессии SVM пытается найти такую функцию (обычно линейную), которая минимизирует ошибку предсказания с учетом некоторого допустимого отклонения (epsilon).

Основная идея заключается в том, чтобы построить модель, которая "прогнозирует" значения целевой переменной $y$ как можно точнее, но при этом позволяет небольшие ошибки (в пределах epsilon) без штрафов.



### Пример: Нахождение функции регрессии

#### Данные:
У нас есть набор данных:

- Признак $x$: $[1, 2, 3, 4]$,
- Целевая переменная $y$: $[1.5, 2.0, 2.5, 3.0]$.

Наша цель — найти линейную функцию регрессии $f(x)$, которая минимизирует ошибку предсказания.



### Шаг 1: Общее уравнение функции регрессии

В простейшем случае функция регрессии является линейной:

$$
f(x) = w x + b
$$

где:
- $w$ — вес (наклон прямой),
- $b$ — смещение (bias).

Мы хотим найти оптимальные значения $w$ и $b$, которые минимизируют ошибку предсказания.



### Шаг 2: Epsilon-трубка (Epsilon-insensitive loss function)

В SVM для регрессии используется концепция **epsilon-трубки**. Это означает, что мы допускаем небольшие ошибки в пределах $\epsilon$ без штрафов. Если ошибка больше $\epsilon$, то она штрафуется пропорционально её величине.

Формально, функция потерь (loss function) выглядит так:

$$
L(y_i, f(x_i)) =
\begin{cases}
0, & \text{если } |y_i - f(x_i)| \leq \epsilon, \\
|y_i - f(x_i)| - \epsilon, & \text{если } |y_i - f(x_i)| > \epsilon.
\end{cases}
$$

Задача состоит в том, чтобы минимизировать сумму таких потерь для всех точек.



### Шаг 3: Формулировка задачи оптимизации

Цель SVM для регрессии — решить следующую задачу оптимизации:

$$
\min_{w, b, \xi_i, \xi_i^*} \frac{1}{2} ||w||^2 + C \sum_{i=1}^n (\xi_i + \xi_i^*)
$$

при условиях:

$$
y_i - (w x_i + b) \leq \epsilon + \xi_i,
$$
$$
(w x_i + b) - y_i \leq \epsilon + \xi_i^*,
$$
$$
\xi_i \geq 0, \quad \xi_i^* \geq 0, \quad \forall i,
$$

где:
- $w$ и $b$ — параметры модели,
- $\xi_i$ и $\xi_i^*$ — слабые переменные (slack variables), которые измеряют ошибки за пределами epsilon-трубки,
- $C$ — коэффициент регуляризации, который контролирует баланс между сложностью модели и величиной ошибок.



### Шаг 4: Применение условий к данным

Подставим наши данные:

- $x = [1, 2, 3, 4]$,
- $y = [1.5, 2.0, 2.5, 3.0]$.

Предположим, что $\epsilon = 0.1$. Тогда условия принимают вид:

1. Для точки $(x_1, y_1) = (1, 1.5)$:
   $$
   1.5 - (w \cdot 1 + b) \leq 0.1 + \xi_1,
   $$
   $$
   (w \cdot 1 + b) - 1.5 \leq 0.1 + \xi_1^*.
   $$

2. Для точки $(x_2, y_2) = (2, 2.0)$:
   $$
   2.0 - (w \cdot 2 + b) \leq 0.1 + \xi_2,
   $$
   $$
   (w \cdot 2 + b) - 2.0 \leq 0.1 + \xi_2^*.
   $$

3. Для точки $(x_3, y_3) = (3, 2.5)$:
   $$
   2.5 - (w \cdot 3 + b) \leq 0.1 + \xi_3,
   $$
   $$
   (w \cdot 3 + b) - 2.5 \leq 0.1 + \xi_3^*.
   $$

4. Для точки $(x_4, y_4) = (4, 3.0)$:
   $$
   3.0 - (w \cdot 4 + b) \leq 0.1 + \xi_4,
   $$
   $$
   (w \cdot 4 + b) - 3.0 \leq 0.1 + \xi_4^*.
   $$



### Шаг 5: Решение системы уравнений

Для упрощения предположим, что данные линейно зависимы, и попробуем найти $w$ и $b$ напрямую. Заметим, что данные образуют почти идеальную прямую:

$$
y = 0.5x + 1.0
$$

Проверим это уравнение на наших точках:

1. Для $x = 1$: $y = 0.5 \cdot 1 + 1.0 = 1.5$ (совпадает),
2. Для $x = 2$: $y = 0.5 \cdot 2 + 1.0 = 2.0$ (совпадает),
3. Для $x = 3$: $y = 0.5 \cdot 3 + 1.0 = 2.5$ (совпадает),
4. Для $x = 4$: $y = 0.5 \cdot 4 + 1.0 = 3.0$ (совпадает).

Таким образом, оптимальные параметры модели:

$$
w = 0.5, \quad b = 1.0
$$



### Шаг 6: Итоговая функция регрессии

Итоговое уравнение функции регрессии:

$$
\boxed{f(x) = 0.5x + 1.0}
$$

Эта функция точно описывает данные и удовлетворяет условиям задачи регрессии с $\epsilon = 0.1$.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR

# Шаг 1: Определение данных
X = np.array([[1], [2], [3], [4]])  # Признак x
y = np.array([1.5, 2.0, 2.5, 3.0])  # Целевая переменная y

# Шаг 2: Создание модели SVM для регрессии
# Используем линейный ядро (kernel='linear') и epsilon=0.1
model = SVR(kernel='linear', C=1.0, epsilon=0.1)

# Обучаем модель
model.fit(X, y)

# Получаем параметры модели
w = model.coef_[0][0]  # Вес (наклон прямой) — извлекаем единственное значение из массива
b = model.intercept_[0]  # Смещение (bias)
print(f"Параметры модели: w = {w}, b = {b}")

# Шаг 3: Генерация предсказаний
X_range = np.linspace(0, 5, 100).reshape(-1, 1)  # Диапазон значений x для построения графика
y_pred = model.predict(X_range)  # Предсказания модели

# Шаг 4: Визуализация
plt.figure(figsize=(8, 6))

# Отображаем исходные данные
plt.scatter(X, y, color='blue', label='Исходные данные')

# Отображаем функцию регрессии
plt.plot(X_range, y_pred, color='red', label=f'Функция регрессии (f(x) = {w:.2f}x + {b:.2f})', linewidth=2)

# Отображаем epsilon-трубку
epsilon = model.epsilon
plt.fill_between(
    X_range.flatten(),
    y_pred - epsilon,
    y_pred + epsilon,
    color='green',
    alpha=0.2,
    label=f'Epsilon-трубка (ε = {epsilon})'
)

# Добавляем легенду и метки осей
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('SVM для задачи регрессии')
plt.grid(True)
plt.show()

# Шаг 5: Проверка точности модели
# Предсказываем значения для исходных данных
y_pred_points = model.predict(X)
errors = np.abs(y - y_pred_points)

# Проверяем ошибки
within_epsilon = errors <= epsilon
print("Ошибки предсказания:")
for i, error in enumerate(errors):
    print(f"Точка {X[i][0]}: ошибка = {error:.2f}, внутри ε-трубки = {within_epsilon[i]}")
```
