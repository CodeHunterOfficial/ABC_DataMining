#Разработка программы тестирования на Python

#### Цель:
Разработать консольную программу для проведения тестирования, которая позволяет загружать вопросы из файла, перемешивать их и варианты ответов, проводить тестирование пользователя, сохранять результаты в файл и предоставлять статистику по выполнению теста. Программа должна быть удобной для пользователя, с четкими инструкциями и немедленной обратной связью после каждого ответа.



#### Требования к функционалу программы:

1. **Хранение вопросов**:
   - Вопросы должны храниться в текстовом файле.
   - Каждый вопрос должен содержать:
     - Текст вопроса.
     - 5 вариантов ответа.
     - Информацию о правильном варианте ответа (индекс или текст).
   - Формат хранения вопросов в файле должен быть четко определен (например, разделение элементов с помощью символа `|`).

2. **Перемешивание**:
   - Перед началом теста вопросы должны перемешиваться случайным образом.
   - Для каждого вопроса варианты ответов также должны перемешиваться, чтобы исключить возможность запоминания порядка правильных ответов.

3. **Тестирование**:
   - Вопросы задаются пользователю по одному.
   - Пользователь выбирает один из предложенных вариантов ответа (ввод номера ответа от 1 до 5).
   - Программа проверяет правильность выбранного ответа и сразу уведомляет пользователя о результате ("Правильно!" или "Неправильно."), после чего автоматически переходит к следующему вопросу.
   - Если пользователь вводит некорректный ответ (например, число вне диапазона или нечисловой символ), программа должна запросить повторный ввод до получения корректного ответа.

4. **Логирование результатов**:
   - Программа должна фиксировать время начала теста.
   - Программа должна фиксировать время окончания теста.
   - Результаты тестирования должны сохраняться в отдельный файл и включать следующую информацию:
     - Время начала теста.
     - Время окончания теста.
     - Общее количество вопросов.
     - Количество правильных ответов.
     - Процент правильных ответов.

5. **Обработка ошибок**:
   - Программа должна корректно обрабатывать некорректные действия пользователя (например, ввод символов вместо числа, выход за пределы допустимого диапазона ответов).
   - При возникновении ошибок чтения файла или некорректного формата данных программа должна выводить информативное сообщение и завершать работу.

6. **Интерфейс программы**:
   - Программа должна быть удобной для использования:
     - Четкие инструкции для пользователя.
     - Понятные сообщения о результатах теста.
     - Автоматический переход к следующему вопросу после проверки ответа.
     - Структурированный вывод информации.



#### Дополнительные требования:

1. **Формат файла с вопросами**:
   - Файл с вопросами должен быть текстовым.
   - Каждая строка файла содержит один вопрос, представленный в следующем формате:
     ```
     <текст вопроса>|<вариант ответа 1>|<вариант ответа 2>|<вариант ответа 3>|<вариант ответа 4>|<вариант ответа 5>|<правильный ответ>
     ```
   - Пример строки:
     ```
     Какой язык программирования используется для веб-разработки?|Python|JavaScript|Java|C++|Ruby|JavaScript
     ```

2. **Формат файла с результатами**:
   - Результаты тестирования сохраняются в текстовый файл.
   - Формат файла:
     ```
     Время начала теста: <дата и время>
     Время окончания теста: <дата и время>
     Общее количество вопросов: <число>
     Количество правильных ответов: <число>
     Процент правильных ответов: <число>%
     ```

3. **Методы реализации**:
   - Для хранения вопросов можно использовать матрицу (список списков) или класс.
   - Если используется класс, он должен содержать методы для перемешивания вариантов ответов и проверки правильности ответа.

4. **Ограничения**:
   - Вопросы должны быть строго с 5 вариантами ответа.
   - Правильный ответ должен быть указан явно в файле с вопросами.



#### Ожидаемый результат:
Программа должна обеспечивать проведение тестирования с учетом всех перечисленных требований. После завершения теста пользователь должен получить полную статистику по своей работе, а результаты должны быть сохранены в файл для дальнейшего анализа.



#### Пример работы программы:

1. **Ввод данных**:
   - Программа загружает вопросы из файла `questions.txt`.
   - Вопросы и варианты ответов перемешиваются.

2. **Процесс тестирования**:
   - Пользователь видит первый вопрос и варианты ответов:
     ```
     Вопрос 1/10: Какой язык программирования используется для веб-разработки?
     1. Ruby
     2. Python
     3. JavaScript
     4. Java
     5. C++
     ```
   - Пользователь вводит номер ответа: `3`.

3. **Обратная связь**:
   - Программа сразу проверяет ответ и уведомляет пользователя:
     ```
     Правильно!
     Переходим к следующему вопросу...
     ```

4. **Следующий вопрос**:
   - Программа автоматически показывает следующий вопрос:
     ```
     Вопрос 2/10: Какая столица Франции?
     1. Берлин
     2. Лондон
     3. Париж
     4. Мадрид
     5. Рим
     ```
   - Пользователь вводит номер ответа: `3`.

5. **Завершение теста**:
   - После ответа на все вопросы программа выводит статистику:
     ```
     Тестирование завершено!
     Общее количество вопросов: 10
     Количество правильных ответов: 8
     Процент правильных ответов: 80.00%
     ```

6. **Сохранение результатов**:
   - Результаты сохраняются в файл `results.txt`:
     ```
     Время начала теста: 2023-03-01 10:00:00
     Время окончания теста: 2023-03-01 10:15:00
     Общее количество вопросов: 10
     Количество правильных ответов: 8
     Процент правильных ответов: 80.00%
     ```



#### Заключение:
Программа должна быть гибкой, надежной и удобной для использования. Она должна соответствовать всем указанным требованиям и обеспечивать корректное проведение тестирования с сохранением результатов для последующего анализа.
