#Использование Docker в проектах Django в среде VS Code




## Введение

**Docker** — это платформа, которая позволяет создавать изолированные среды для выполнения приложений. Это особенно полезно для разработчиков Django, так как Docker гарантирует, что ваше приложение будет работать одинаково на любом устройстве или сервере, независимо от операционной системы или конфигурации.

В этой лекции мы подробно рассмотрим, как использовать Docker для работы с проектами Django в **Visual Studio Code (VS Code)**. Мы начнем с основ и пошагово настроим все необходимое. К концу лекции вы сможете самостоятельно создавать, настраивать и запускать Django-приложения внутри Docker-контейнеров, используя удобные инструменты VS Code.



## 1. Что такое Docker?

**Docker** — это инструмент, который позволяет:
- Создавать контейнеры — изолированные среды для выполнения приложений.
- Управлять зависимостями вашего проекта.
- Развертывать приложения одинаково на разных устройствах.

Проще говоря, Docker помогает вам "упаковать" ваше приложение вместе со всеми его зависимостями в одну единую "коробку", которую можно легко перемещать и запускать где угодно. Это особенно важно для командной работы, когда несколько разработчиков должны иметь одинаковые условия для разработки и тестирования.

Ключевые преимущества Docker:
- **Изолированность**: Каждый контейнер работает независимо от других процессов на хост-машине.
- **Повторяемость**: Один раз настроенная среда может быть точно воспроизведена на любом другом устройстве.
- **Эффективность**: Docker не требует создания целой виртуальной машины, а использует ресурсы операционной системы более эффективно.



## 2. Что такое Django?

**Django** — это высокопроизводительный фреймворк для создания веб-приложений на языке Python. Он предоставляет множество инструментов для быстрой разработки сложных сайтов, включая:
- ORM (Object-Relational Mapping) для работы с базами данных.
- Встроенные механизмы аутентификации пользователей.
- Систему администрирования (Django Admin).
- Поддержку маршрутизации URL.

Django следует принципу *"Don't Repeat Yourself"* (DRY), что делает код более чистым и поддерживаемым. Он идеально подходит для разработки как простых, так и масштабируемых веб-проектов.



## 3. Что такое VS Code?

**Visual Studio Code (VS Code)** — это легковесная интегрированная среда разработки (IDE), созданная компанией Microsoft. Она поддерживает множество языков программирования, включая Python, и имеет богатый набор расширений для различных задач.

Основные преимущества VS Code:
- **Лёгкость использования**: Интерфейс интуитивно понятен даже для начинающих разработчиков.
- **Мощные расширения**: Можно установить дополнительные плагины для работы с Git, Docker, форматированием кода и многим другим.
- **Поддержка множества языков**: Кроме Python, VS Code отлично работает с JavaScript, TypeScript, C++, Java и другими языками программирования.



## 4. Почему Docker + Django + VS Code?

Эта комбинация идеально подходит для разработки веб-приложений, потому что:
- **Docker** обеспечивает воспроизводимость среды разработки, что критически важно для командной работы.
- **Django** предоставляет мощные инструменты для создания веб-сайтов, позволяя сосредоточиться на бизнес-логике, а не на настройке окружения.
- **VS Code** делает процесс разработки удобным и эффективным благодаря своим расширениям и интеграции с различными инструментами.

С помощью Docker вы можете быть уверены, что ваше Django-приложение будет работать одинаково как на локальном компьютере, так и на сервере. А VS Code поможет вам писать и отлаживать код быстро и комфортно.

### 7.2 Содержимое сгенерированных файлов

#### 7.2.1 `Dockerfile`

`Dockerfile` — это инструкция для создания Docker-образа. Этот файл определяет все шаги, необходимые для подготовки среды выполнения вашего приложения внутри контейнера.

**Пример содержимого:**
```dockerfile
# Базовый образ Python 3.10 slim (легковесный)
FROM python:3-slim

# Объявляет, что приложение будет прослушивать порт 8000
EXPOSE 8000

# Отключает создание .pyc файлов внутри контейнера
ENV PYTHONDONTWRITEBYTECODE=1

# Отключает буферизацию вывода для удобства логирования
ENV PYTHONUNBUFFERED=1

# Копирует requirements.txt в контейнер для установки зависимостей
COPY requirements.txt .

# Устанавливает зависимости из requirements.txt
RUN python -m pip install --upgrade pip && \
    python -m pip install -r requirements.txt

# Устанавливает рабочую директорию внутри контейнера
WORKDIR /app

# Копирует все файлы проекта в рабочую директорию
COPY . /app

# Создает непривилегированного пользователя appuser с UID 5678
# Это повышает безопасность, так как приложение не запускается от root
RUN adduser -u 5678 --disabled-password --gecos "" appuser && \
    chown -R appuser /app

# Переключается на созданного пользователя appuser
USER appuser

# Запускает Gunicorn как основной процесс
CMD ["gunicorn", "--bind", "0.0.0.0:8000", "mysite.wsgi"]
```

**Объяснение каждой строки:**
1. **`FROM python:3-slim`**: Используется базовый образ Python 3.10 slim, который содержит только минимально необходимые компоненты для работы Python.
2. **`EXPOSE 8000`**: Объявляет, что приложение будет использовать порт 8000. Это помогает другим сервисам или хост-машине знать, какой порт используется.
3. **`ENV PYTHONDONTWRITEBYTECODE=1`**: Отключает создание `.pyc` файлов, которые обычно генерируются Python для ускорения загрузки модулей.
4. **`ENV PYTHONUNBUFFERED=1`**: Отключает буферизацию stdout и stderr, что важно для корректного логирования.
5. **`COPY requirements.txt .`**: Копирует файл `requirements.txt` в контейнер для установки зависимостей.
6. **`RUN python -m pip install -r requirements.txt`**: Устанавливает все зависимости, указанные в `requirements.txt`.
7. **`WORKDIR /app`**: Устанавливает рабочую директорию `/app`, где будут находиться файлы проекта.
8. **`COPY . /app`**: Копирует все файлы проекта из текущей директории в `/app`.
9. **`RUN adduser ...`**: Создает непривилегированного пользователя `appuser` с UID 5678 для запуска приложения.
10. **`USER appuser`**: Переключается на пользователя `appuser` для увеличения безопасности.
11. **`CMD ["gunicorn", "..."]`**: Запускает Gunicorn как основной процесс для обслуживания Django-приложения.

#### 7.2.2 `docker-compose.yml`

`docker-compose.yml` — это файл для определения и управления несколькими контейнерами одновременно. Он позволяет легко настраивать взаимодействие между сервисами.

**Пример содержимого:**
```yaml
version: '3.4'

services:
  mydjangoproject:
    image: mydjangoproject
    build:
      context: .
      dockerfile: ./Dockerfile
    ports:
      - "8000:8000"
    volumes:
      - .:/app
    environment:
      - DEBUG=1
```

**Объяснение каждого блока:**
1. **`version: '3.4'`**: Указывает версию формата файла `docker-compose`. Разные версии имеют разные возможности.
2. **`services`**: Определяет список сервисов, которые будут запущены в контейнерах.
3. **`build: context: . dockerfile: ./Dockerfile`**: Инструктирует Docker Compose использовать текущую директорию (`.`) как контекст сборки и `Dockerfile` для создания образа.
4. **`ports: - "8000:8000"`**: Перенаправляет порт 8000 контейнера на порт 8000 хост-машины, чтобы можно было получить доступ к приложению через браузер.
5. **`volumes: - .:/app`**: Монтирует текущую директорию локальной машины в `/app` внутри контейнера. Это позволяет автоматически обновлять код внутри контейнера при его изменении на локальной машине.
6. **`environment: - DEBUG=1`**: Устанавливает переменную окружения `DEBUG=1`, которая включает режим отладки Django.

#### 7.2.3 `.dockerignore`

Файл `.dockerignore` используется для исключения ненужных файлов и директорий из контейнера во время сборки образа. Это ускоряет процесс сборки и уменьшает размер конечного образа.

**Пример содержимого:**
```
**/__pycache__
**/.venv
**/.git
**/.gitignore
**/*.log
**/*.sqlite3
**/node_modules
**/Dockerfile*
**/docker-compose*
```

**Объяснение каждой строки:**
1. **`**/__pycache__`**: Исключает временные файлы Python, которые создаются при компиляции `.py` файлов.
2. **`**/.venv`**: Исключает виртуальное окружение Python, так как оно уже встроено в образ.
3. **`**/.git`**: Исключает файлы системы контроля версий Git.
4. **`**/*.log`**: Исключает файлы журналов, которые обычно создаются во время работы приложения.
5. **`**/*.sqlite3`**: Исключает файлы базы данных SQLite, так как они могут быть большими и неактуальными.
6. **`**/node_modules`**: Исключает папку `node_modules`, если проект использует JavaScript-зависимости.
7. **`**/Dockerfile*`**: Исключает сам `Dockerfile` и связанные с ним файлы, чтобы они не попали в контейнер.

---

#### 7.2.4 `docker-compose.debug.yml`

Этот файл используется для настройки отладочной среды с использованием Docker Compose.

**Пример содержимого:**
```yaml
version: '3.4'

services:
  mydjangoproject:
    image: mydjangoproject
    build:
      context: .
      dockerfile: ./Dockerfile
    command: ["sh", "-c", "pip install debugpy -t /tmp && python /tmp/debugpy --wait-for-client --listen 0.0.0.0:5678 manage.py runserver 0.0.0.0:8000 --nothreading --noreload"]
    ports:
      - "8000:8000"
      - "5678:5678"
    volumes:
      - .:/app
    environment:
      - DEBUG=1
```

**Объяснение каждого блока:**
1. **`command: ["sh", "-c", "..."]`**: Задает специальную команду для отладки:
   - Устанавливает `debugpy` для удаленной отладки.
   - Запускает `debugpy` в режиме ожидания подключения клиента на порту 5678.
   - Запускает Django-сервер без автоматической перезагрузки (`--noreload`) и многопоточности (`--nothreading`).
2. **`ports: - "8000:8000"`**: Перенаправляет порт 8000 для доступа к приложению.
3. **`ports: - "5678:5678"`**: Перенаправляет порт 5678 для подключения отладчика.
4. **`volumes: - .:/app`**: Монтирует текущую директорию локальной машины в `/app` для автоматического обновления кода.
5. **`environment: - DEBUG=1`**: Включает режим отладки Django.

## 13. Настройка Django для работы с PostgreSQL  

PostgreSQL — это мощная и надежная реляционная база данных, которая часто используется в производственных средах. Django отлично работает с PostgreSQL благодаря его поддержке в стандартной конфигурации.  

### Шаг 1: Установка PostgreSQL  
1. **Установите PostgreSQL**:  
   - Для Windows: Скачайте и установите [PostgreSQL](https://www.postgresql.org/download/).  
   - Для Linux: Используйте пакетный менеджер (например, `sudo apt install postgresql`).  
   - Для macOS: Используйте Homebrew (`brew install postgresql`).  

2. **Создайте пользователя и базу данных**:  
   - Откройте терминал и перейдите в PostgreSQL-среду:  
     ```bash  
     sudo -u postgres psql  
     ```  
   - Создайте нового пользователя:  
     ```sql  
     CREATE USER myuser WITH PASSWORD 'mypassword';  
     ```  
   - Создайте базу данных:  
     ```sql  
     CREATE DATABASE mydb OWNER myuser;  
     ```  
   - Выполните выход:  
     ```sql  
     \q  
     ```  



### Шаг 2: Установка зависимостей  
Чтобы Django мог работать с PostgreSQL, потребуется специальный драйвер `psycopg2`.  

1. Добавьте `psycopg2-binary` в файл `requirements.txt`:  
   ```
   psycopg2-binary
   ```  

2. Установите зависимости:  
   ```bash  
   pip install -r requirements.txt  
   ```  

> **Примечание:** Если возникнут проблемы с установкой, попробуйте использовать только `psycopg2`, но это может потребовать дополнительных системных библиотек.



### Шаг 3: Настройка `settings.py`  
Откройте файл `mysite/settings.py` и измените параметры базы данных:  

```python  
DATABASES = {  
    'default': {  
        'ENGINE': 'django.db.backends.postgresql',  # Тип базы данных  
        'NAME': 'mydb',                             # Название базы данных  
        'USER': 'myuser',                           # Пользователь  
        'PASSWORD': 'mypassword',                   # Пароль  
        'HOST': 'localhost',                        # Адрес сервера (localhost или IP)  
        'PORT': '5432',                             # Порт (по умолчанию 5432 для PostgreSQL)  
    }  
}  
```  



### Шаг 4: Применение миграций  
1. Создайте миграционные файлы:  
   ```bash  
   python manage.py makemigrations  
   ```  

2. Примените миграции:  
   ```bash  
   python manage.py migrate  
   ```  



### Шаг 5: Проверка работы  
1. Запустите сервер:  
   ```bash  
   python manage.py runserver  
   ```  

2. Если всё настроено правильно, Django должен успешно подключиться к PostgreSQL.  



## 14. Настройка Django для работы с MSSQL  

Microsoft SQL Server (MSSQL) — это еще одна популярная реляционная база данных, особенно в корпоративных средах. Чтобы использовать MSSQL с Django, потребуется специальный драйвер и дополнительные настройки.  



### Шаг 1: Установка MSSQL  
1. **Установите MSSQL**:  
   - Для Windows: Скачайте и установите [SQL Server](https://www.microsoft.com/en-us/sql-server/sql-server-downloads).  
   - Для Linux: Используйте официальную документацию Microsoft.  
   - Для macOS: Можно использовать Docker-образ MSSQL.  

2. **Создайте пользователя и базу данных**:  
   - Откройте SQL Server Management Studio (SSMS) или используйте командную строку.  
   - Создайте нового пользователя:  
     ```sql  
     CREATE LOGIN myuser WITH PASSWORD = 'mypassword';  
     ```  
   - Создайте базу данных:  
     ```sql  
     CREATE DATABASE mydb;  
     ```  
   - Назначьте права пользователю:  
     ```sql  
     USE mydb;  
     CREATE USER myuser FOR LOGIN myuser;  
     ALTER ROLE db_owner ADD MEMBER myuser;  
     ```  



### Шаг 2: Установка зависимостей  
Для работы с MSSQL потребуется специальный драйвер `django-mssql-backend` и `pyodbc`.  

1. Добавьте следующие зависимости в файл `requirements.txt`:  
   ```
   django-mssql-backend  
   pyodbc  
   ```  

2. Установите зависимости:  
   ```bash  
   pip install -r requirements.txt  
   ```  

> **Примечание:** Если вы используете macOS или Linux, вам также нужно установить ODBC-драйвер для MSSQL. Следуйте официальной документации Microsoft:  
> - [ODBC Driver for Linux/macOS](https://docs.microsoft.com/en-us/sql/connect/odbc/linux-mac/installing-the-microsoft-odbc-driver-for-sql-server)  



### Шаг 3: Настройка `settings.py`  
Откройте файл `mysite/settings.py` и измените параметры базы данных:  

```python  
DATABASES = {  
    'default': {  
        'ENGINE': 'sql_server.pyodbc',  # Тип базы данных  
        'NAME': 'mydb',                 # Название базы данных  
        'USER': 'myuser',               # Пользователь  
        'PASSWORD': 'mypassword',       # Пароль  
        'HOST': 'localhost',            # Адрес сервера (localhost или IP)  
        'PORT': '1433',                 # Порт (по умолчанию 1433 для MSSQL)  
        'OPTIONS': {  
            'driver': 'ODBC Driver 17 for SQL Server',  # Опция драйвера  
        },  
    }  
}  
```  

> **Важно:** Убедитесь, что версия драйвера совпадает с установленной на вашем компьютере.  



### Шаг 4: Применение миграций  
1. Создайте миграционные файлы:  
   ```bash  
   python manage.py makemigrations  
   ```  

2. Примените миграции:  
   ```bash  
   python manage.py migrate  
   ```  



### Шаг 5: Проверка работы  
1. Запустите сервер:  
   ```bash  
   python manage.py runserver  
   ```  

2. Если всё настроено правильно, Django должен успешно подключиться к MSSQL.

## 16. Добавление проекта под контроль Git



### 1. Инициализация Git в проекте

Откройте терминал в корневой директории вашего проекта (`D:\Projects\my_django_project`) и выполните следующую команду:

```bash
git init
```

**Что это делает?**
- Создает новое локальное репозиторий Git в вашем проекте.
- Появится скрытая папка `.git`, которая будет хранить всю историю изменений.



image.png



### 2. Создание `.gitignore` файла

`.gitignore` — это файл, который указывает Git, какие файлы и папки следует игнорировать при добавлении их в репозиторий. Для Django-проекта необходимо исключить временные файлы, базы данных SQLite, файлы виртуального окружения и другие ненужные элементы.

Создайте файл `.gitignore` в корневой директории проекта и добавьте следующее содержимое:

#### Пример содержимого `.gitignore` для Django-проекта:
```plaintext
# Системные файлы
*.log
*.pyc
__pycache__/
.vscode/
.idea/

# Файлы виртуального окружения
venv/
.env
ENV/
env.bak/
virtualenv/
VirtualEnv/

# Файлы Docker
Dockerfile*
docker-compose.*
.dockerignore

# Базы данных SQLite
*.sqlite3
*.sql

# Файлы Media и Static (если они не нужны в репозитории)
media/
static/

# Локальные конфигурационные файлы
local_settings.py
db.sqlite3

# Файлы IDE
.vs/
.settings/
.project

# Файлы сборки
build/
dist/
```


image.png


**Объяснение основных строк:**
1. **`*.pyc` и `__pycache__/`**: Исключает временные файлы Python.
2. **`venv/` и `.env`**: Исключает виртуальное окружение и переменные окружения.
3. **`*.sqlite3`**: Исключает файлы базы данных SQLite.
4. **`media/` и `static/`**: Исключает медиа- и статические файлы, если они не нужны в репозитории.
5. **`.vscode/` и `.idea/`**: Исключает настройки IDE (например, VS Code или PyCharm).



### 3. Добавление файлов в Git

Теперь добавьте все файлы проекта в Git, за исключением тех, которые указаны в `.gitignore`:

```bash
git add .
```

**Что это делает?**
- Добавляет все файлы в индекс Git, исключая те, что указаны в `.gitignore`.


### 4. Создание первого коммита

Выполните первый коммит с описанием изменений:

```bash
git commit -m "Initial commit"
```

**Что это делает?**
- Сохраняет текущее состояние файлов как первый коммит в истории Git.

### 5. (Опционально) Подключение удаленного репозитория

Если вы хотите разместить свой проект на GitHub, GitLab или другом удаленном сервисе, выполните следующие шаги:

1. **Создайте удаленный репозиторий** на GitHub или GitLab.
2. **Подключите его к локальному репозиторию:**

   ```bash
   git remote add origin <URL_удаленного_репозитория>
   ```

   Например:
   ```bash
   git remote add origin https://github.com/username/my-django-project.git
   ```

3. **Отправьте изменения в удаленный репозиторий:**

   ```bash
   git push -u origin main
   ```

   Если ветка называется `master`, используйте:
   ```bash
   git push -u origin master
   ```
