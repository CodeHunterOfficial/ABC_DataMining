#Основы DevOps


## 1. Введение в DevOps

**DevOps** — это культура и философия, которая объединяет разработку (Development) и эксплуатацию (Operations), создавая более гибкие, быстрые и надежные процессы через автоматизацию и межфункциональное сотрудничество. Это означает, что **Dev + Ops = DevOps**.

По сути, DevOps — это не просто технологическое решение или инструмент, а целая система, направленная на ускорение процесса доставки программного обеспечения конечному пользователю, снижение времени цикла разработки (time-to-market) и повышение качества продукта. Она основана на автоматизации ключевых этапов жизненного цикла ПО, таких как сборка, тестирование, развертывание и мониторинг.

Это тесное сотрудничество между командой разработчиков и командой эксплуатации для создания более качественного приложения. Более того, DevOps подразумевает изменение культуры работы: отдельные команды больше не действуют изолированно, а работают совместно, имея общую цель — выпускать продукты быстрее, качественнее и безопаснее.

При понимании DevOps важно учитывать следующие моменты:
- **DevOps — это не только объединение двух команд**, но и создание новой культуры, где все участники процесса взаимодействуют эффективно.
- **Это не отдельная команда или продукт**, а комплексный подход, который затрагивает все этапы разработки и эксплуатации.
- **Это не конкретный инструмент**, а набор методологий, практик и инструментов, которые помогают автоматизировать процессы.
- Люди DevOps — это члены внутренней команды, работающие либо на этапе разработки, либо на этапе эксплуатации, но с общей целью.

DevOps представляет собой систему, состоящую из людей, процессов и инструментов. Она объединяет различные команды (например, разработку и операции) с четко определенными процессами и эффективными инструментами для автоматической доставки программного обеспечения клиенту. Это набор методологий, используемых командами DevOps для ускорения качественной доставки ПО.

## 2. Почему DevOps необходим?

Чтобы понять важность DevOps, рассмотрим ситуацию без его применения:

В типичной ИТ-компании существуют две основные команды:
- Команда разработчиков, занимающаяся написанием кода и модульным тестированием.
- Группа эксплуатации, ответственная за обслуживание и мониторинг продукта.

Без DevOps:
- Команды работают полностью изолированно друг от друга.
- Большая часть времени тратится на ручное создание кода и развертывание в различных средах.
- Процесс характеризуется длительными периодами ожидания завершения работы других команд.
- Ручное выполнение задач приводит к ошибкам и увеличению времени их исправления.

Пример проблемы: Если команда разработки использует одну версию библиотеки, а команда эксплуатации другую, это может привести к несовместимости кода в производственной среде. Такие ситуации часто возникают в компаниях, где нет четкой координации между разработкой и эксплуатацией.

**Преимущества внедрения DevOps:**
- Увеличивает вероятность успешных релизов без ошибок.
- Упрощает весь процесс разработки и развертывания.
- Автоматизирует ручные процессы (сборка, выпуск и т.д.).
- Автоматизирует выполнение тестовых случаев.
- Настройка CI/CD (Continuous Integration / Continuous Delivery) в цикле выпуска.
- Реализация живого мониторинга.
- Содействие лучшему взаимодействию между командами.
- Снижение количества сбоев и необходимости откатов.
- Обеспечение постоянного улучшения процессов.

**Реальные примеры успешного внедрения DevOps:**
- **Netflix**: Компания перешла на микросервисную архитектуру и внедрила DevOps, что позволило ей быстро развивать новые функции и поддерживать высокую доступность сервиса.
- **Amazon**: Amazon Web Services (AWS) стала пионером в области DevOps, предлагая инструменты для автоматизации процессов разработки и эксплуатации.
- **Spotify**: Компания внедрила гибкую модель управления проектами, основанную на DevOps, что позволило ей выпускать обновления чаще и быстрее реагировать на изменения рынка.




## 3. Жизненный цикл DevOps

Жизненный цикл DevOps объединяет жизненные циклы разработки и эксплуатации. Он состоит из следующих ключевых этапов:

image.png




### 1. **ПЛАНИРОВАНИЕ (PLAN)**
На этом этапе происходит анализ требований заказчика и формирование стратегии разработки:
- Анализ требований заказчика.
- Определение типа продукции.
- Установка сроков для спринтов и финального продукта.
- Выбор технологий.
- Подбор инструментов.
- Формирование команды.
- Выбор методологии разработки (например, Agile).

### 2. **КОДИРОВАНИЕ (CODE)**
На этом этапе создается рабочий код:
- Создание кода с использованием методологии Agile.
- Разделение задач на спринты.
- Написание модульных тестов.
- Важно также писать документацию и следить за качеством кода.

### 3. **СБОРКА (BUILD)**
На этом этапе происходит компиляция кода:
- Первичная сборка кода на локальной машине разработчика.
- Автоматическая сборка после коммита в систему управления версиями (например, Git).
- Создание артефактов для дальнейшего использования.

### 4. **ТЕСТИРОВАНИЕ (TEST)**
На этом этапе проверяется качество кода:
- Автоматическое выполнение модульных тестов.
- Интеграционное тестирование.
- Тестирование безопасности.
- Нагрузочное тестирование.

### 5. **РЕЛИЗ (RELEASE)**
На этом этапе готовится артефакт для дальнейшего развертывания:
- Артефакт собирается и проверяется на соответствие требованиям.
- Подготавливается для развертывания в тестовой и производственной средах.

### 6. **РАЗВЕРТЫВАНИЕ (DEPLOY)**
На этом этапе происходит развертывание приложения:
- Автоматическое развертывание через CI/CD.
- Возможность ручного подтверждения перед продакшеном.
- Полная автоматизация для небольших изменений.
- Использование различных стратегий развертывания (например, blue-green deployment, canary release).

### 7. **ЭКСПЛУАТАЦИЯ И МОНИТОРИНГ (OPERATE & MONITOR)**
На этом этапе происходит управление и мониторинг системы:
- Управление системой.
- Мониторинг производительности и функциональности.
- Сбор метрик и аналитики для принятия решений о дальнейших улучшениях.


**Как это работает?**
Когда разработчик регистрирует код в системе управления версиями, запускается процесс CI/CD:
1. Получение кода из системы управления версиями.
2. Восстановление зависимостей (например, из NuGet, Maven).
3. Сборка кода.
4. Запуск модульных тестов.
5. Развертывание на промежуточные серверы (CI/CD).
6. Развертывание в production (CD или Continuous Deployment).




## 4. Конвейер CI/CD

В современных процессах разработки программного обеспечения скорость доставки становится ключевым фактором успеха. Чтобы ускорить этот процесс, используются различные механизмы, такие как Agile, автоматизация и специализированные инструменты. Это называется **DevOps**.

Рассмотрим классический процесс разработки и доставки ПО:

1. **Получение требований от бизнеса**.
2. **Составление плана преобразования требований в функциональность**.
3. **Написание кода разработчиками и помещение его в репозиторий** (например, TFS, GitHub).
4. **Выполнение процесса сборки** после регистрации кода в репозитории.
5. **Запуск тестовых случаев** для проверки кода.
6. **Развертывание кода на сервере DEV**, затем на QA.
7. **Подтверждение QA** и последующее развертывание в PROD.

**Проблемы классического процесса:**
1. Разработчики не знают, успешно ли прошла сборка после каждого коммита.
2. Необходимость ждать, пока другие разработчики завершат свои задачи.
3. Возможность возникновения ошибок при ручном развертывании.
4. Длительные задержки при выявлении проблем в коде.
5. Высокий риск ошибок при переходе от разработки к эксплуатации.

**Решение через CI/CD:**

- **Continuous Integration (CI)**: Автоматическая сборка и тестирование кода каждый раз, когда разработчик делает коммит. Это позволяет быстро выявлять ошибки и обеспечивать стабильность кодовой базы.
- **Continuous Delivery (CD)**: Автоматическое развертывание собранного кода на промежуточные серверы (DEV, QA) и подготовка к развертыванию в production после ручного утверждения.
- **Continuous Deployment**: Полностью автоматическое развертывание кода в production без ручного утверждения.


image.png

Как видно из приведенного выше изображения, мы видим, что как только разработчик регистрирует код в системе управления версиями, запускается Azure DevOps Pipeline. Получение кода из системы управления версиями, его восстановление, восстановление пакетов из NuGet, Maven и т. д., сборка кода и выполнение сценариев модульных тестов являются частью непрерывной интеграции (CI).


Включая непрерывную интеграцию, если артефакт сборки автоматически развертывается на любом промежуточном сервере, таком как DEV или QA, и развертывается в рабочей среде после утверждения вручную, это называется непрерывной доставкой (CD). Если ручное утверждение преобразуется в автоматическое развертывание, это называется непрерывным развертыванием. Итак, в основном Continuous Deployment — это то же самое, что Continuous Delivery, но без какого-либо утверждения вручную. Все идет в режиме авто одобрения.




### **Процесс создания образца приложения Django:**
1. Создание нового проекта Django с использованием Visual Studio Code.
2. Написание базового функционала и модульных тестов.
3. Настройка Git для хранения кода.
4. Создание Dockerfile для контейнеризации приложения.
5. Настройка Jenkins для автоматической сборки и тестирования кода.
6. Развертывание приложения на тестовом сервере с использованием Ansible.
7. Настройка мониторинга с помощью Prometheus и Grafana.
