Программа должна предоставлять возможность создания записей о недвижимости, владельцах, арендаторах, аренде, оплате и других данных. Также программа должна поддерживать выполнение различных операций с данными, таких как добавление, редактирование, удаление, поиск и анализ.



### Основные требования к реализации

#### 1. **Абстрактный класс `Property` (недвижимость)**
   - Создайте абстрактный класс `Property`, который будет являться базовым для всех типов недвижимости.
   - В классе определите:
     - Абстрактный метод `calculate_rental_cost`, который должен быть реализован в каждом подклассе.
     - Общие атрибуты для всех типов недвижимости: `property_id` (идентификатор), `address` (адрес), `area` (площадь), `monthly_rate` (ежемесячная ставка аренды), `is_available` (доступность).
     - Метод `__str__`, возвращающий строковое представление объекта (например, "Недвижимость: [address], Площадь: [area] кв.м").
     - Методы сравнения (`__lt__`, `__gt__`) для сравнения недвижимости по ежемесячной ставке или площади.
   - Используйте инкапсуляцию: сделайте атрибуты приватными и предоставьте доступ к ним через геттеры и сеттеры с аннотациями типов.

#### 2. **Наследование**
   - Создайте подклассы для различных типов недвижимости:
     - `Apartment` (квартира): добавьте атрибут `number_of_rooms` (количество комнат).
     - `House` (дом): добавьте атрибут `has_garden` (наличие сада).
     - `CommercialSpace` (коммерческое помещение): добавьте атрибут `business_type` (тип бизнеса).
   - В каждом подклассе переопределите метод `calculate_rental_cost` для расчета стоимости аренды с учетом дополнительных факторов (например, скидки на долгосрочную аренду).
   - Переопределите метод `__str__` для вывода дополнительной информации (например, "Квартира: [address], Комнат: [number_of_rooms]").

#### 3. **Композиция и агрегация**
   - Создайте класс `RentalAgreement`, который будет содержать информацию об аренде:
     - Атрибуты: `agreement_id` (идентификатор аренды), `tenant` (арендатор), `property` (недвижимость), `start_date` (дата начала аренды), `end_date` (дата окончания аренды), `total_cost` (общая стоимость).
     - Методы: `add_extra` (добавить дополнительную услугу), `remove_extra` (удалить дополнительную услугу), `calculate_total` (рассчитать общую стоимость).
   - Используйте **композицию**: добавьте атрибут `tenant_info` (информация об арендаторе) через отдельный класс `Tenant`.
   - Используйте **агрегацию**: аренда может существовать без дополнительных услуг, но услуги могут быть добавлены позже.

#### 4. **Интерфейсы для работы с сервисом**
   - Создайте интерфейс `Rentable`, который определяет метод `rent_property`. Этот метод должен быть реализован для аренды недвижимости.
   - Добавьте второй интерфейс `Reportable`, который определяет метод `generate_report`. Этот метод должен быть реализован для генерации отчетов об аренде (например, "Отчет по аренде: [данные]").

#### 5. **Миксины**
   - Создайте миксин `LoggingMixin`, который добавляет функциональность логирования действий с арендой и недвижимостью. Например, метод `log_action`, который записывает действия (например, "Аренда [agreement_id] создана").
   - Создайте второй миксин `NotificationMixin`, который добавляет функциональность отправки уведомлений. Например, метод `send_notification`, который отправляет уведомления (например, "Ваша аренда [address] подтверждена" или "Аренда [agreement_id] отменена").
   - Используйте оба миксина в подклассах для демонстрации множественного наследования.

#### 6. **Метаклассы**
   - Реализуйте метакласс `PropertyMeta`, который автоматически регистрирует все подклассы `Property` в реестре. Это позволит динамически создавать экземпляры недвижимости по имени типа.

#### 7. **Фабричные методы**
   - Создайте класс `PropertyFactory` с методом `create_property`, который принимает тип недвижимости (например, `"apartment"`, `"house"`, `"commercial"`) и создает соответствующий экземпляр класса.
   - Используйте этот метод для создания недвижимости в программе.

#### 8. **Цепочка обязанностей (Chain of Responsibility)**
   - Реализуйте паттерн "Цепочка обязанностей" для обработки запросов на изменение аренды. Например:
     - Менеджер по аренде может одобрить изменение, если оно незначительное.
     - Финансовый отдел может одобрить изменение, если оно требует пересмотра стоимости.
     - Директор может одобрить любое изменение.
   - Создайте цепочку обработчиков (`RentalManager`, `FinanceDepartment`, `Director`), которая последовательно передает запросы между звеньями.

#### 9. **Шаблонный метод (Template Method)**
   - Реализуйте шаблонный метод для стандартизации процесса аренды недвижимости. Создайте базовый класс `RentalProcess` с методом `rent_property`, который определяет общую структуру процесса:
     - Проверка доступности недвижимости.
     - Оформление аренды.
     - Подтверждение аренды.
   - Подклассы (`OnlineRentalProcess`, `OfflineRentalProcess`) должны реализовывать конкретные шаги процесса.

#### 10. **Декоратор для проверки прав доступа**
   - Создайте декоратор `check_permissions`, который проверяет права доступа пользователя перед выполнением определенных действий (например, аренда недвижимости или изменение аренды). Если у пользователя нет прав, выбрасывайте исключение `PermissionDeniedError`.

#### 11. **Исключения**
   - Создайте пользовательские исключения для обработки ошибок:
     - `InvalidPropertyError`: если недвижимость содержит некорректные данные.
     - `PermissionDeniedError`: если у пользователя нет прав доступа.
     - `RentalNotFoundError`: если аренда не найдена.

#### 12. **Сериализация и десериализация**
   - Реализуйте возможность сохранения и загрузки данных о недвижимости и аренде в файл (например, в формате JSON).
   - Добавьте методы `to_dict` и `from_dict` для преобразования объектов в словари и обратно.

#### 13. **Методы сравнения**
   - Реализуйте методы `__eq__`, `__lt__`, `__gt__` для сравнения недвижимости по ежемесячной ставке, площади или другим критериям.

#### 14. **Логирование и документация**
   - Настройте систему логирования с использованием библиотеки `logging`. Логи должны записываться в файл и в консоль.
   - Добавьте docstrings для всех методов и классов.



### Пример использования программы

1. Создание недвижимости:
   - Используйте `PropertyFactory` для создания недвижимости разных типов.
   - Добавьте недвижимость в сервис с помощью метода `add_property`.

2. Управление данными:
   - Редактируйте данные недвижимости через сеттеры.
   - Удаляйте недвижимость из сервиса с помощью метода `remove_property`.

3. Анализ данных:
   - Получите список всей недвижимости с помощью метода `get_all_properties`.
   - Найдите недвижимость по адресу с помощью метода `search_by_address`.

4. Логирование и уведомления:
   - Используйте миксины `LoggingMixin` и `NotificationMixin` для записи действий и отправки уведомлений.

5. Обработка исключений:
   - Обрабатывайте пользовательские исключения при попытке аренды недвижимости с некорректными данными или при отсутствии прав доступа.



### Заключение

Данная программа демонстрирует применение объектно-ориентированного программирования, включая наследование, композицию, агрегацию, интерфейсы, миксины, метаклассы, фабричные методы, паттерны проектирования (цепочка обязанностей, шаблонный метод) и другие концепции. ### Основные требования к реализации

#### 1. **Абстрактный класс `IndustrialEquipment` (промышленное оборудование)**

     - Методы сравнения (`__lt__`, `__gt__`) для сравнения оборудования по стоимости аренды или состоянию.
   - Используйте инкапсуляцию: сделайте атрибуты приватными и предоставьте доступ к ним через геттеры и сеттеры с аннотациями типов.

   - Используйте **композицию**: добавьте атрибут `customer_info` (информация о клиенте) через отдельный класс `Customer`.

#### 4. **Интерфейсы для работы с сервисом**

#### 5. **Миксины**

#### 6. **Метаклассы**

#### 7. **Фабричные методы**
   - Создайте класс `EquipmentFactory` с методом `create_equipment`, который принимает тип оборудования (например, `"forklift"`, `"crane"`, `"compressor"`) и создает соответствующий экземпляр класса.
   - Используйте этот метод для создания оборудования в программе.

#### 8. **Цепочка обязанностей (Chain of Responsibility)**
   - Реализуйте паттерн "Цепочка обязанностей" для обработки запросов на изменение аренды. Например:
     - Оператор может одобрить изменение, если оно незначительное.
     - Менеджер может одобрить изменение, если оно требует пересмотра стоимости.

   - Создайте цепочку обработчиков (`Operator`, `Manager`, `Admin`), которая последовательно передает запросы между звеньями.

     - Проверка доступности оборудования.
     - Оформление аренды.
     - Подтверждение аренды.
   - Подклассы (`OnlineRentalProcess`, `OfflineRentalProcess`) должны реализовывать конкретные шаги процесса.

#### 10. **Декоратор для проверки прав доступа**
   - Создайте декоратор `check_permissions`, который проверяет права доступа пользователя перед выполнением определенных действий (например, аренда оборудования или изменение аренды). Если у пользователя нет прав, выбрасывайте исключение `PermissionDeniedError`.

#### 11. **Исключения**
   - Создайте пользовательские исключения для обработки ошибок:
     - `InvalidEquipmentError`: если оборудование содержит некорректные данные.
     - `PermissionDeniedError`: если у пользователя нет прав доступа.
     - `RentalNotFoundError`: если аренда не найдена.

#### 12. **Сериализация и десериализация**
   - Реализуйте возможность сохранения и загрузки данных об оборудовании и аренде в файл (например, в формате JSON).
   - Добавьте методы `to_dict` и `from_dict` для преобразования объектов в словари и обратно.

#### 13. **Методы сравнения**
   - Реализуйте методы `__eq__`, `__lt__`, `__gt__` для сравнения оборудования по стоимости аренды, состоянию или другим критериям.

#### 14. **Логирование и документация**
   - Настройте систему логирования с использованием библиотеки `logging`. Логи должны записываться в файл и в консоль.
   - Добавьте docstrings для всех методов и классов.
