## 🗃️ **Модуль 9.5: Работа с базами данных — SQLite (без ООП)**

> **Добавлен как отдельный модуль — критически важный навык даже для начинающих**

- Что такое база данных? Зачем она нужна (vs файлы).
- Что такое SQLite? Встраиваемая, файловая, не требует сервера.
- **Модуль `sqlite3` — стандартная библиотека Python.**
- Подключение к БД: `sqlite3.connect('mydb.sqlite')`.
- Создание курсора: `cursor = conn.cursor()`.
- Выполнение SQL-запросов: `cursor.execute("CREATE TABLE ...")`.
- Основные команды SQL (в объёме курса):
  - `CREATE TABLE users (id INTEGER, name TEXT)`
  - `INSERT INTO users VALUES (1, 'Alice')`
  - `SELECT * FROM users`
  - `UPDATE users SET name='Bob' WHERE id=1`
  - `DELETE FROM users WHERE id=1`
- Получение данных: `cursor.fetchall()`, `cursor.fetchone()`.
- Фиксация изменений: `conn.commit()`.
- Закрытие: `conn.close()`.
- Параметризованные запросы: `cursor.execute("SELECT * FROM users WHERE id=?", (user_id,))` — защита от SQL-инъекций.
- Практические задачи:
  - Создание таблицы “Пользователи” или “Задачи”.
  - Добавление, чтение, обновление, удаление записей.
  - Поиск по условию.
  - Импорт данных из CSV в SQLite.
  - Экспорт данных из SQLite в JSON.
