#### Возможности моделей на основе LLM для генерации SQL-запросов




### 1. **Трансформация естественного языка в SQL**

Современные большие языковые модели (LLM) обладают способностью интерпретировать запросы, сформулированные на естественном языке, и преобразовывать их в корректные SQL-запросы. Это достигается благодаря предварительному обучению моделей на больших объемах параллельных данных, состоящих из текстовых запросов и соответствующих им SQL-запросов. Такой подход позволяет LLM понимать как семантику запросов пользователя, так и структуру базы данных, а также соблюдать правила синтаксиса SQL.

Например, запрос на естественном языке:  
*"Какие 10 альбомов по продажам являются лучшими?"*  

может быть автоматически преобразован в следующий SQL-запрос:  
```sql
SELECT a.AlbumId, a.Title, SUM(il.Quantity) AS TotalSales
FROM Album a
JOIN Track t ON a.AlbumId = t.AlbumId
JOIN InvoiceLine il ON t.TrackId = il.TrackId
GROUP BY a.AlbumId, a.Title
ORDER BY TotalSales DESC
LIMIT 10;
```

Результат выполнения данного запроса предоставляет не только список из 10 лучших альбомов по сумме продаж (`SUM(il.Quantity)`), но и может быть дополнительно визуализирован в виде диаграммы, отображающей распределение продаж среди этих альбомов (см. Рисунок 1).

````python
import sqlite3

# Создание подключения к базе данных
conn = sqlite3.connect('UniversityDB.db')
cursor = conn.cursor()

# Создание таблицы "Students"
cursor.execute('''
CREATE TABLE IF NOT EXISTS Students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    fname TEXT NOT NULL,       -- Имя
    lname TEXT NOT NULL,       -- Фамилия
    age INTEGER,               -- Возраст
    group_name TEXT            -- Группа
)
''')

# Создание таблицы "Teachers"
cursor.execute('''
CREATE TABLE IF NOT EXISTS Teachers (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    fname TEXT NOT NULL,       -- Имя
    lname TEXT NOT NULL,       -- Фамилия
    department TEXT            -- Кафедра
)
''')

# Создание таблицы "Courses"
cursor.execute('''
CREATE TABLE IF NOT EXISTS Courses (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,        -- Название курса
    teacher_id INTEGER,        -- ID преподавателя (внешний ключ)
    FOREIGN KEY (teacher_id) REFERENCES Teachers(id)
)
''')

# Создание таблицы "Enrollments" (связь между студентами и курсами)
cursor.execute('''
CREATE TABLE IF NOT EXISTS Enrollments (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    student_id INTEGER,        -- ID студента (внешний ключ)
    course_id INTEGER,         -- ID курса (внешний ключ)
    grade REAL,                -- Оценка
    FOREIGN KEY (student_id) REFERENCES Students(id),
    FOREIGN KEY (course_id) REFERENCES Courses(id)
)
''')

# Добавление данных в таблицу "Students"
students_data = [
    ('Иван', 'Иванов', 20, 'Группа_1'),
    ('Петр', 'Петров', 21, 'Группа_2'),
    ('Анна', 'Сидорова', 19, 'Группа_1'),
    ('Мария', 'Кузнецова', 22, 'Группа_3'),
    ('Алексей', 'Смирнов', 20, 'Группа_2')
]
cursor.executemany('INSERT INTO Students (fname, lname, age, group_name) VALUES (?, ?, ?, ?)', students_data)

# Добавление данных в таблицу "Teachers"
teachers_data = [
    ('Андрей', 'Сергеев', 'Математика'),
    ('Елена', 'Романова', 'Физика'),
    ('Олег', 'Петров', 'Программирование')
]
cursor.executemany('INSERT INTO Teachers (fname, lname, department) VALUES (?, ?, ?)', teachers_data)

# Добавление данных в таблицу "Courses"
courses_data = [
    ('Высшая математика', 1),
    ('Физика', 2),
    ('Python программирование', 3),
    ('Линейная алгебра', 1),
    ('Квантовая механика', 2)
]
cursor.executemany('INSERT INTO Courses (name, teacher_id) VALUES (?, ?)', courses_data)

# Добавление данных в таблицу "Enrollments"
enrollments_data = [
    (1, 1, None),  # Иван Иванов записан на "Высшая математика"
    (2, 2, None),  # Петр Петров записан на "Физика"
    (3, 3, None),  # Анна Сидорова записана на "Python программирование"
    (4, 1, None),  # Мария Кузнецова записана на "Высшая математика"
    (5, 4, None)   # Алексей Смирнов записан на "Линейная алгебра"
]
cursor.executemany('INSERT INTO Enrollments (student_id, course_id, grade) VALUES (?, ?, ?)', enrollments_data)

# Сохранение изменений
conn.commit()
````
