```python
# Установка SQLite (обычно уже доступна в Python)
import sqlite3

# Создание подключения к базе данных (если файла нет, он будет создан автоматически)
conn = sqlite3.connect('DBstudents.db')

# Создание объекта курсора для выполнения SQL-запросов
cursor = conn.cursor()

# Создание таблицы "Students"
cursor.execute('''
CREATE TABLE IF NOT EXISTS Students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    fname TEXT NOT NULL,       -- Имя
    lname TEXT NOT NULL,       -- Фамилия
    age INTEGER,               -- Возраст
    group_name TEXT            -- Группа
)
''')

# Добавление 10 записей о студентах
students_data = [
    ('Иван', 'Иванов', 20, 'Группа_1'),
    ('Петр', 'Петров', 21, 'Группа_2'),
    ('Анна', 'Сидорова', 19, 'Группа_1'),
    ('Мария', 'Кузнецова', 22, 'Группа_3'),
    ('Алексей', 'Смирнов', 20, 'Группа_2'),
    ('Екатерина', 'Васильева', 21, 'Группа_1'),
    ('Дмитрий', 'Попов', 23, 'Группа_3'),
    ('Ольга', 'Федорова', 19, 'Группа_2'),
    ('Сергей', 'Иванов', 22, 'Группа_1'),
    ('Татьяна', 'Морозова', 20, 'Группа_3')
]

# Вставка данных в таблицу (используем правильные имена полей: fname, lname, age, group_name)
cursor.executemany('INSERT INTO Students (fname, lname, age, group_name) VALUES (?, ?, ?, ?)', students_data)

# Сохранение изменений
conn.commit()

# Проверка: выборка всех студентов из таблицы
cursor.execute('SELECT * FROM Students')  # Исправлено название таблицы на "Students"
rows = cursor.fetchall()

# Вывод результатов
print("Таблица 'Студенты':")
for row in rows:
    print(row)

# Закрытие соединения
conn.close()
```
