```python
import transformers
import datasets
import torch
import evaluate
import sacrebleu
import rouge_score
import matplotlib.pyplot as plt
from datasets import load_dataset
from evaluate import load as load_metric  # Use evaluate.load instead of datasets.load_metric
from transformers import (
    BertTokenizer, BertForMaskedLM,
    RobertaTokenizer, RobertaForMaskedLM,
    T5Tokenizer, T5ForConditionalGeneration,
    AutoTokenizer, AutoModelForSeq2SeqLM,
    XLNetTokenizer, XLNetLMHeadModel,
    Trainer, TrainingArguments
)
import matplotlib.pyplot as plt
print("All libraries are successfully installed!")
```

```python
import os
from transformers import (
    BertTokenizer, BertForMaskedLM,
    RobertaTokenizer, RobertaForMaskedLM,
    AutoTokenizer, AutoModelForSeq2SeqLM,
    XLNetTokenizer, XLNetLMHeadModel,
    TrainingArguments, Trainer
)
from datasets import load_dataset
from evaluate import load as load_metric
import matplotlib.pyplot as plt

class GrammarCorrectionSystem:
    def __init__(self, model_name):
        self.model_name = model_name
        self.tokenizer = None
        self.model = None
        self.trainer = None
        self.history = {}

    def load_model(self):
        if "bert" in self.model_name.lower():
            self.tokenizer = BertTokenizer.from_pretrained(self.model_name)
            self.model = BertForMaskedLM.from_pretrained(self.model_name)
        elif "roberta" in self.model_name.lower():
            self.tokenizer = RobertaTokenizer.from_pretrained(self.model_name)
            self.model = RobertaForMaskedLM.from_pretrained(self.model_name)
        elif "t5" in self.model_name.lower() or "flan-t5" in self.model_name.lower():
            self.tokenizer = AutoTokenizer.from_pretrained(self.model_name)
            self.model = AutoModelForSeq2SeqLM.from_pretrained(self.model_name)
        elif "xlnet" in self.model_name.lower():
            self.tokenizer = XLNetTokenizer.from_pretrained(self.model_name)
            self.model = XLNetLMHeadModel.from_pretrained(self.model_name)
        else:
            raise ValueError(f"Unsupported model: {self.model_name}")

    def preprocess_data(self, dataset):
        def tokenize(examples):
            model_inputs = self.tokenizer(
                examples["input_text"], max_length=128, truncation=True, padding="max_length"
            )
            labels = self.tokenizer(
                examples["target_text"], max_length=128, truncation=True, padding="max_length"
            )
            model_inputs["labels"] = labels["input_ids"]
            return model_inputs

        return dataset.map(tokenize, batched=True)

    def train(self, train_dataset, test_dataset, output_dir="./results", epochs=3):
        # Отключаем wandb глобально
        os.environ["WANDB_DISABLED"] = "true"

        training_args = TrainingArguments(
            output_dir=output_dir,
            evaluation_strategy="epoch",
            learning_rate=5e-5,
            per_device_train_batch_size=8,
            per_device_eval_batch_size=8,
            num_train_epochs=epochs,
            weight_decay=0.01,
            save_total_limit=2,
            logging_dir=f"./logs/{self.model_name}",
            logging_steps=10,
            push_to_hub=False,
            report_to=None  # Отключает интеграцию с wandb
        )

        tokenized_train = self.preprocess_data(train_dataset)
        tokenized_test = self.preprocess_data(test_dataset)

        self.trainer = Trainer(
            model=self.model,
            args=training_args,
            train_dataset=tokenized_train,
            eval_dataset=tokenized_test,
            tokenizer=self.tokenizer
        )

        self.trainer.train()
        self.history = self.trainer.state.log_history  # Сохраняем историю обучения

    def evaluate(self, dataset):
        predictions = []
        references = []

        for example in dataset:
            input_ids = self.tokenizer(example["input_text"], return_tensors="pt").input_ids
            outputs = self.model.generate(input_ids, max_length=128, num_beams=5, early_stopping=True)
            corrected_text = self.tokenizer.decode(outputs[0], skip_special_tokens=True)
            predictions.append(corrected_text)
            references.append([example["target_text"]])

        bleu_metric = load_metric("sacrebleu")
        rouge_metric = load_metric("rouge")

        bleu_score = bleu_metric.compute(predictions=predictions, references=references)
        rouge_score = rouge_metric.compute(predictions=predictions, references=references)

        return bleu_score, rouge_score

    def plot_training_history(self):
        if not self.history:
            print("No training history available.")
            return

        loss = [entry["loss"] for entry in self.history if "loss" in entry]
        eval_loss = [entry["eval_loss"] for entry in self.history if "eval_loss" in entry]

        plt.figure(figsize=(12, 6))
        plt.plot(loss, label="Training Loss", color="blue")
        plt.plot(eval_loss, label="Validation Loss", color="orange")
        plt.title(f"Training and Validation Loss - {self.model_name}")
        plt.xlabel("Steps")
        plt.ylabel("Loss")
        plt.legend()
        plt.show()

    def correct_grammar(self, text):
        input_ids = self.tokenizer(f"Fix the grammar: {text}", return_tensors="pt").input_ids
        outputs = self.model.generate(input_ids, max_length=128, num_beams=5, early_stopping=True)
        corrected_text = self.tokenizer.decode(outputs[0], skip_special_tokens=True)
        return corrected_text


# Загрузка датасета JFLEG
dataset = load_dataset("jfleg")

# Предобработка данных
def preprocess_data(example):
    return {
        "input_text": example["sentence"],
        "target_text": example["corrections"][0]  # Используем первое исправление как целевой текст
    }

# Применяем предобработку ко всему датасету
dataset = dataset.map(preprocess_data)

# Check the available splits
print(dataset.keys())  # Output: dict_keys(['validation', 'test'])

# Since there is no 'train' split, we'll split 'validation' into train and validation subsets
if "train" not in dataset:
    print("No 'train' split found. Splitting 'validation' into train/validation subsets.")
    dataset = dataset["validation"].train_test_split(test_size=0.2, shuffle=True, seed=42)
    train_dataset = dataset["train"]
    val_dataset = dataset["test"]  # This will act as the validation set
else:
    train_dataset = dataset["train"]
    val_dataset = dataset["validation"]

test_dataset = dataset["test"]  # Use the 'test' split for final evaluation

# Сравнение моделей
models = [
    ("bert-base-uncased", "BERT"),
    ("roberta-base", "RoBERTa"),
    ("t5-small", "T5"),
    ("google/flan-t5-small", "Flan-T5"),
    ("xlnet-base-cased", "XLNet")
]

results = []

for model_name, model_label in models:
    print(f"Training {model_label}...")
    system = GrammarCorrectionSystem(model_name)
    system.load_model()
    system.train(train_dataset, test_dataset, output_dir=f"./results/{model_label}", epochs=3)
    bleu, rouge = system.evaluate(test_dataset)
    results.append((model_label, bleu["score"], rouge["rougeL"].mid.fmeasure))
    system.plot_training_history()

# Вывод результатов
print("\nComparison of Models:")
print("{:<10} {:<10} {:<10}".format("Model", "BLEU", "ROUGE-L"))
for model, bleu_score, rouge_score in results:
    print("{:<10} {:.4f}   {:.4f}".format(model, bleu_score, rouge_score))

# График сравнения BLEU
model_names = [result[0] for result in results]
bleu_scores = [result[1] for result in results]

plt.figure(figsize=(10, 6))
plt.bar(model_names, bleu_scores, color="green")
plt.title("Comparison of Models by BLEU Score")
plt.xlabel("Models")
plt.ylabel("BLEU Score")
plt.show()

# График сравнения ROUGE-L
rouge_scores = [result[2] for result in results]

plt.figure(figsize=(10, 6))
plt.bar(model_names, rouge_scores, color="blue")
plt.title("Comparison of Models by ROUGE-L Score")
plt.xlabel("Models")
plt.ylabel("ROUGE-L Score")
plt.show()
```
