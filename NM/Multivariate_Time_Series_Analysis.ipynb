```python
# Импорт библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf

# Загрузка данных о ценах на нефть (Brent и WTI)
brent = yf.download('BZ=F', start='2010-01-01', end='2023-10-01')
wti = yf.download('CL=F', start='2010-01-01', end='2023-10-01')

# Загрузка данных о ценах на природный газ (Henry Hub)
henry_hub = yf.download('NG=F', start='2010-01-01', end='2023-10-01')

# Загрузка данных о фондовых индексах (S&P 500, DAX, FTSE)
sp500 = yf.download('^GSPC', start='2010-01-01', end='2023-10-01')
dax = yf.download('^GDAXI', start='2010-01-01', end='2023-10-01')
ftse = yf.download('^FTSE', start='2010-01-01', end='2023-10-01')

# Создание DataFrame для каждого набора данных
brent_df = brent[['Close']].rename(columns={'Close': 'Brent'})
wti_df = wti[['Close']].rename(columns={'Close': 'WTI'})
henry_hub_df = henry_hub[['Close']].rename(columns={'Close': 'Henry_Hub'})
sp500_df = sp500[['Close']].rename(columns={'Close': 'S&P_500'})
dax_df = dax[['Close']].rename(columns={'Close': 'DAX'})
ftse_df = ftse[['Close']].rename(columns={'Close': 'FTSE'})

# Объединение всех DataFrame в один
data = pd.concat([brent_df, wti_df, henry_hub_df, sp500_df, dax_df, ftse_df], axis=1)

# Удаление строк с пропусками
data.dropna(inplace=True)

# Просмотр объединенных данных
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.vector_ar.var_model import VAR
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Визуализация временных рядов
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Brent'], label='Brent', color='blue')
plt.plot(data.index, data['WTI'], label='WTI', color='orange')
plt.plot(data.index, data['S&P_500'], label='S&P 500', color='green')
plt.plot(data.index, data['DAX'], label='DAX', color='red')
plt.plot(data.index, data['FTSE'], label='FTSE', color='purple')
plt.title('Цены на нефть и фондовые индексы')
plt.xlabel('Дата')
plt.ylabel('Цена')
plt.legend()
plt.grid()
plt.show()

# Корреляционная матрица
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица')
plt.show()

# Тест на стационарность (ADF)
def adf_test(series):
    result = adfuller(series)
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    if result[1] <= 0.05:
        print("Стационарный")
    else:
        print("Нестационарный")

for column in data.columns:
    print(f'Тест на стационарность для {column}:')
    adf_test(data[column])
    print('\n')

# Нормализация данных
scaler = StandardScaler()
df_scaled = scaler.fit_transform(data)

# PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_scaled)
plt.scatter(pca_result[:, 0], pca_result[:, 1])
plt.title('PCA')
plt.show()

# Векторная авторегрессионная модель (VAR)
model = VAR(df_scaled)
var_result = model.fit(maxlags=15, ic='aic')
print(var_result.summary())
```

Еще можно провести совместный анализ всех временных рядов, чтобы учесть их взаимосвязи и взаимодействия. Это особенно важно, если вы хотите понять, как изменения в одном активе влияют на другие или как они вместе формируют общую рыночную динамику. Вот несколько подходов для проведения такого анализа:


### **1. Векторная авторегрессия (VAR)**
#### **Цель**:
Изучить взаимосвязи между всеми временными рядами одновременно.

#### **Методология**:
- VAR — это модель, которая позволяет описать каждую переменную как линейную функцию прошлых значений всех переменных.
- Модель строится для всех временных рядов (Brent, WTI, Henry Hub, S&P 500, DAX, FTSE).
- Проводится анализ импульсно-откликной функции (Impulse Response Function) для оценки того, как изменение одного актива влияет на другие.
- Анализ разложения вариаций (Variance Decomposition) для определения, какой процент изменчивости каждого актива объясняется другими активами.

#### **Ожидаемые результаты**:
- Выявление причинно-следственных связей между активами.
- Определение того, какие активы наиболее сильно влияют на другие.
- Понимание того, как шоки в одном секторе (например, нефть) распространяются на другие секторы (фондовые индексы).

---

### **2. Кросскорреляционный анализ**
#### **Цель**:
Оценить лаговые зависимости между временными рядами.

#### **Методология**:
- Расчет кросскорреляций между всеми парами активов для разных лагов.
- Поиск значимых отклонений от нуля, которые указывают на наличие временной зависимости.
- Использование графиков кросскорреляций для визуализации результатов.

#### **Ожидаемые результаты**:
- Выявление задержек между изменениями в разных активах.
- Определение того, какие активы реагируют быстрее/медленнее на изменения других активов.

- Интерпретация главных компонент для понимания, какие активы больше всего влияют на каждую компоненту.

#### **Ожидаемые результаты**:

### **4. Совместное моделирование с использованием Transformer**
#### **Цель**:
Учесть долгосрочные и короткосрочные зависимости между всеми временными рядами.

#### **Методология**:
- Использование архитектуры Transformer для обработки многомерных временных рядов.
- Обучение модели на всех активных рядах одновременно.
- Анализ внимания (attention weights) для выявления важных взаимосвязей между активами.

#### **Ожидаемые результаты**:
- Понимание того, какие активы чаще "взаимодействуют" друг с другом.
- Выявление долгосрочных трендов и короткосрочных флуктуаций.
- Возможность точного прогнозирования будущих значений всех активов одновременно.

---

### **5. Коинтеграционный анализ**
#### **Цель**:
Изучить долгосрочные отношения между временными рядами.

#### **Методология**:
- Проверка на коинтеграцию с использованием теста Энгела-Грэнджера или Johansen.
- Если коинтеграция существует, построение ошибки коррекции (Error Correction Model, ECM) для анализа краткосрочных отклонений от долгосрочного равновесия.

#### **Ожидаемые результаты**:
- Выявление пар или групп активов, которые имеют долгосрочное равновесие.
- Определение того, как активы восстанавливаются после краткосрочных отклонений.
- Возможность использования этих отношений для арбитражных стратегий.

---

### **6. Графовый анализ взаимосвязей**
#### **Цель**:
Построить граф взаимосвязей между активами и проанализировать его структуру.

#### **Методология**:
- Создание матрицы корреляций или кросскорреляций между всеми активами.
- Преобразование матрицы в граф, где узлы — это активы, а ребра — связи между ними.
- Анализ центральности узлов (например, PageRank, Betweenness Centrality) для определения ключевых активов.
- Кластеризация графа для выделения групп активов с схожим поведением.

#### **Ожидаемые результаты**:
- Выявление "ключевых игроков" среди активов.
- Определение групп активов с сильной взаимосвязью.
- Понимание структуры взаимодействий между активами.

---

### **7. Мультивариатный GARCH**
#### **Цель**:
Моделировать условную волатильность и корреляции между всеми активами одновременно.

#### **Методология**:
- Использование мультивариатных GARCH-моделей (например, DCC-GARCH, BEKK-GARCH).
- Анализ динамики корреляций между активами во времени.
- Прогнозирование условной волатильности и корреляций.

#### **Ожидаемые результаты**:
- Выявление периодов повышенной корреляции между активами.
- Понимание того, как волатильность одного актива влияет на другие.
- Инструмент для управления рисками в многомерном портфеле.

---

### **8. Нейронные сети для многомерного прогнозирования**
#### **Цель**:
Построить единую модель для прогнозирования всех активов одновременно.

#### **Методология**:
- Использование рекуррентных нейронных сетей (RNN, LSTM) или Transformer для учета временных зависимостей.
- Обучение модели на всех временных рядах одновременно.
- Добавление дополнительных признаков (например, макроэкономические данные, технические индикаторы).

#### **Ожидаемые результаты**:
- Точный прогноз всех активов одновременно.
- Понимание того, какие активы являются наиболее предсказуемыми.
- Возможность использования модели для принятия инвестиционных решений.

---

### **9. Сравнение многомерных временных рядов через динамическую байесовскую сеть (DBN)**
#### **Цель**:
Моделировать сложные зависимости между активами с учетом скрытых факторов.

#### **Методология**:
- Использование DBN для представления временных рядов как вероятностной графовой модели.
- Определение скрытых состояний, которые влияют на все активы.
- Анализ переходных вероятностей для понимания динамики рынка.

#### **Ожидаемые результаты**:
- Выявление скрытых факторов, влияющих на все активы.
- Понимание того, как рынок переходит между различными состояниями.
- Возможность использования модели для прогнозирования будущих состояний рынка.



Элементы **нечеткой логики** (fuzzy logic) можно успешно использовать для анализа временных рядов, особенно когда данные имеют неопределенности, шум или сложные взаимосвязи, которые трудно описать точными числами. Нечеткая логика позволяет работать с приблизительными и нечеткими понятиями, что может быть полезно в финансовых данных, где часто присутствуют неопределенные факторы.

Вот как можно применить нечеткую логику для анализа ваших временных рядов:

---

### **1. Нечеткая классификация трендов**
#### **Цель**:
Классифицировать движения цен на активы (например, "медленный рост", "быстрый рост", "стабильность", "падение") с использованием нечетких множеств.

#### **Методология**:
- Определение нечетких множеств для различных типов трендов (например, "медленный рост" — это когда изменение цены находится в диапазоне 0–2%, "быстрый рост" — от 2% до 5%, и т.д.).
- Использование функций принадлежности (membership functions), таких как треугольные или трапециевидные, для определения степени принадлежности каждого значения к каждому множеству.
- Применение правил нечеткой логики для классификации текущего состояния рынка.

#### **Ожидаемые результаты**:
- Более интуитивное описание динамики цен с помощью человеческих терминов ("медленный рост", "падение").
- Возможность учитывать неопределенности в данных, например, когда движение цены находится между двумя категориями.

---

### **2. Нечеткий прогнозирование**
#### **Цель**:
Построить модель прогнозирования будущих значений на основе нечеткой логики.

#### **Методология**:
- Разделение исторических данных на входные и выходные переменные (например, прошлые значения цены как вход, будущие значения как выход).
- Создание нечеткой модели с использованием правил типа "ЕСЛИ... ТО..." (например, "ЕСЛИ цена Brent медленно растет И S&P 500 стабилен, ТО WTI будет расти").
- Обучение модели на исторических данных с использованием адаптивных методов, таких как ANFIS (Adaptive Neuro-Fuzzy Inference System).

#### **Ожидаемые результаты**:
- Модель, которая учитывает неопределенные зависимости между активами.
- Возможность получения предсказаний в виде интервалов или вероятностей вместо точных чисел.

---

### **3. Нечеткий анализ рисков**
#### **Цель**:
Оценить уровень риска для каждого актива с учетом неопределенностей.

#### **Методология**:
- Определение нечетких множеств для уровня риска (например, "низкий риск", "средний риск", "высокий риск").
- Расчет степени принадлежности каждого актива к каждому уровню риска на основе его волатильности, корреляций и других показателей.
- Построение системы правил для комбинирования различных факторов риска.

#### **Ожидаемые результаты**:
- Более гибкая оценка риска, учитывающая неопределенности.
- Возможность сравнения активов по их уровню риска с использованием нечетких терминов.

---

### **4. Нечеткий анализ событий**
#### **Цель**:
Исследовать влияние экономических или политических событий на активы с учетом неопределенностей.

#### **Методология**:
- Определение нечетких множеств для характера события (например, "слабое влияние", "среднее влияние", "сильное влияние").
- Использование функций принадлежности для оценки степени воздействия события на каждый актив.
- Построение системы правил для анализа последствий событий.

#### **Ожидаемые результаты**:
- Учет того, что одно и то же событие может иметь разное влияние на разные активы.
- Возможность оценки вероятности различных сценариев развития событий.

---

### **5. Нечеткая портфельная оптимизация**
#### **Цель**:
Найти оптимальное распределение активов в портфеле с учетом неопределенностей в доходности и риске.

#### **Методология**:
- Определение нечетких множеств для целевых показателей (например, "высокая доходность", "низкий риск").
- Использование нечеткого программирования для максимизации общей "удовлетворенности" портфеля.
- Включение дополнительных ограничений, таких как минимальная/максимальная доля активов.

#### **Ожидаемые результаты**:
- Портфель, который учитывает неопределенные цели инвестора.
- Возможность получения компромиссных решений, которые удовлетворяют нескольким критериям одновременно.

---

### **6. Нечеткий анализ взаимосвязей**
#### **Цель**:
Изучить взаимосвязи между активами с учетом неопределенностей.

#### **Методология**:
- Определение нечетких множеств для уровня корреляции (например, "слабая связь", "средняя связь", "сильная связь").
- Расчет нечетких коэффициентов корреляции с использованием функций принадлежности.
- Анализ динамики нечетких корреляций во времени.

#### **Ожидаемые результаты**:
- Более детальное понимание взаимосвязей между активами.
- Возможность выявления скрытых зависимостей, которые трудно увидеть с помощью классических методов.

---

### **7. Нечеткая система управления рисками**
#### **Цель**:
Разработать систему для автоматического управления рисками в портфеле.

#### **Методология**:
- Создание набора правил нечеткой логики для принятия решений о покупке/продаже активов (например, "ЕСЛИ волатильность высока И корреляция с рынком сильная, ТО продать актив").
- Интеграция системы с реальными данными для автоматического мониторинга портфеля.

#### **Ожидаемые результаты**:
- Автоматизированная система управления рисками.
- Возможность быстрой реакции на изменения на рынке.

---

### **Пример использования нечеткой логики: анализ трендов**

#### Шаги:
1. **Определение нечетких множеств**:
   - "Медленный рост": если изменение цены за день находится в диапазоне [0%, 2%].
   - "Быстрый рост": если изменение цены находится в диапазоне [2%, 5%].
   - "Стабильность": если изменение цены находится в диапазоне [-1%, 1%].
   - "Падение": если изменение цены меньше -1%.

2. **Функции принадлежности**:
   - Использование треугольных или трапециевидных функций для определения степени принадлежности каждого значения к каждому множеству.

3. **Правила нечеткой логики**:
   - ЕСЛИ изменение цены принадлежит "медленному росту", ТО текущий тренд = "рост".
   - ЕСЛИ изменение цены принадлежит "падению", ТО текущий тренд = "падение".

4. **Агрегация и дефаззификация**:
   - Комбинирование выводов из всех правил.
   - Преобразование нечеткого вывода в четкое значение (например, среднее арифметическое).

---

### **Преимущества использования нечеткой логики**:
- Учет неопределенностей и шума в данных.
- Более интуитивное описание процессов с помощью человеческих терминов.
- Гибкость для адаптации к меняющимся условиям.

Таким образом, нечеткая логика может стать мощным инструментом для анализа временных рядов, особенно в ситуациях, где классические методы сталкиваются с ограничениями из-за неопределенностей в данных.
