# **Анализ и прогнозирование веб-трафика: сравнение классических, машинных и глубоких методов для оптимизации пользовательской активности**


#### **Введение**

В эпоху цифровой трансформации веб-сайты стали ключевым инструментом для достижения бизнес-целей, таких как увеличение конверсий, улучшение пользовательского опыта и повышение прибыльности. Однако анализ производительности веб-сайта сопряжен с рядом вызовов, включая сезонные колебания трафика, наличие шумов в данных и сложность прогнозирования в условиях нестабильности. Временные ряды, отражающие данные о трафике и пользовательской активности, позволяют не только выявить ключевые тренды и сезонные паттерны, но и обнаружить аномалии, что делает их ценным инструментом для прогнозирования и принятия решений.

Согласно исследованию Vlahogianni, Karlaftis и Golias (2014), точное прогнозирование временных рядов, таких как веб-трафик, требует учета как внутренних закономерностей данных, так и внешних факторов, которые могут влиять на динамику [1]. В условиях растущей конкуренции на цифровых платформах эта задача становится критически важной для эффективного управления ресурсами, планирования маркетинговых кампаний и минимизации издержек.

Целью данной статьи является проведение сравнительного анализа классических, машинных и глубоких методов для анализа и прогнозирования веб-трафика с акцентом на повышение уровня вовлеченности пользователей. В работе рассматриваются два основных подхода:
1. **Прогнозирование без учета внешних факторов**, где модель строится исключительно на основе исторических данных временного ряда.
2. **Прогнозирование с учетом внешних факторов**, где учитываются дополнительные переменные, такие как маркетинговые кампании, сезонность, погодные условия и другие влияющие факторы.

Для решения задачи прогнозирования без учета внешних факторов используется функциональная зависимость, отражающая связь между прошлыми и будущими значениями временного ряда:
$$
Z(t) = F(Z(t-1), Z(t-2), Z(t-3), \dots) + \varepsilon_t,
$$
где $Z(t)$ — значения временного ряда в момент времени $t$, а $\varepsilon_t$ — ошибка модели. Основная задача заключается в минимизации среднего абсолютного отклонения (MAE) между прогнозируемыми и фактическими значениями:
$$
E = \frac{1}{P} \sum_{t=T+1}^{T+P} |\varepsilon_t| \to \min.
$$

Эта формула является стандартной для авторегрессионных моделей (AR), которые широко используются в прогнозировании временных рядов. Чучуева И.А. (2012) также использует аналогичные подходы в своей работе, где рассматривается прогнозирование на основе выборки максимального подобия [2].

Для прогнозирования с учетом внешних факторов модель расширяется за счет включения дополнительных переменных $X_1(t), X_2(t), \dots, X_S(t)$, которые могут влиять на исходный временной ряд:
$$
Z(t) = F(Z(t-1), Z(t-2), \dots, X_1(t), X_1(t-1), \dots, X_S(t), X_S(t-1), \dots) + \varepsilon_t.
$$
В этом случае задача также сводится к минимизации ошибки прогноза $E$, но с учетом дополнительных факторов. Такие модели, как SARIMAX, успешно применяются для учета внешних факторов в прогнозировании временных рядов [2].

Новизна исследования заключается в комплексном подходе, который объединяет классические методы анализа временных рядов, современные методы машинного обучения и глубокого обучения, что позволяет учитывать как внутренние закономерности данных, так и внешние факторы, влияющие на веб-трафик. На основе полученных данных предложены практические рекомендации по оптимизации маркетинговых стратегий и улучшению пользовательского опыта.

### **Методы исследования**

В рамках данного исследования применяются следующие методы анализа и прогнозирования временных рядов:

1. **Классические методы**:
   - ARIMA (Autoregressive Integrated Moving Average)
   - SARIMAX (Seasonal ARIMA with Exogenous Variables)
   - VAR (Vector Autoregression)
   - SVAR (Structural Vector Autoregression)

2. **Методы машинного обучения**:
   - XGBoost (Extreme Gradient Boosting)
   - LightGBM
   - Random Forest

3. **Методы глубокого обучения**:
   - LSTM (Long Short-Term Memory)
   - GRU (Gated Recurrent Unit)
   - CNN (Convolutional Neural Networks)

Каждый из этих методов имеет свои преимущества и недостатки при анализе временных рядов, что делает их применимыми в различных сценариях. В таблице ниже приведены ключевые характеристики каждого метода.

---

#### **Сравнение подходов**

| **Метод**         | **Преимущества**                                                                 | **Недостатки**                                                                 |
|--------------------|----------------------------------------------------------------------------------|--------------------------------------------------------------------------------|
| **ARIMA**          | Простота, интерпретируемость, подходит для линейных зависимостей.                | Не учитывает сезонность, требует стационарности данных.                        |
| **SARIMAX**        | Учет сезонности и внешних факторов.                                              | Сложность настройки, медленная работа на больших данных.                       |
| **VAR**            | Учет взаимосвязей между несколькими временными рядами.                          | Требует большого объема данных для оценки параметров.                          |
| **SVAR**           | Возможность анализа причинно-следственных связей.                               | Сложность настройки и интерпретации.                                           |
| **XGBoost**        | Высокая точность, устойчивость к шумам.                                          | Требует тщательной настройки гиперпараметров.                                  |
| **LightGBM**       | Быстрая обработка данных, эффективен для категориальных признаков.               | Склонность к переобучению на небольших данных.                                 |
| **Random Forest**  | Устойчивость к переобучению, простота интерпретации.                             | Менее точный по сравнению с бустинговыми методами.                             |
| **LSTM**           | Учет долгосрочных зависимостей, высокая точность для сложных данных.             | Требует больших объемов данных, высокая вычислительная сложность.              |
| **GRU**            | Быстрее обучается по сравнению с LSTM.                                           | Менее точный для очень сложных зависимостей.                                   |
| **CNN**            | Эффективен для выявления локальных паттернов.                                    | Требует специфической предобработки данных.                                    |


Таким образом, классические методы, такие как ARIMA и SARIMAX, подходят для задач с линейными зависимостями и стационарными данными. Методы машинного обучения, такие как XGBoost и LightGBM, демонстрируют высокую точность и устойчивость к шумам, но требуют тщательной настройки. Методы глубокого обучения, такие как LSTM и GRU, эффективны для сложных временных рядов, но требуют больших объемов данных и вычислительных ресурсов. Выбор метода зависит от специфики данных и задач исследования.

### **1.4. Удаление выбросов**
Выбросы могут искажать результаты моделирования. Их можно удалить или заменить.

#### Примеры:
- **Удаление выбросов на основе межквартильного размаха (IQR)**:



```python
# Функция для обработки выбросов и визуализации
def handle_outliers_and_visualize(data, column):
    # Вычисляем квартили и границы
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Визуализация до обработки выбросов
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    sns.boxplot(data[column])
    plt.title(f'До обработки выбросов ({column})')
    plt.xlabel(column)

    # Обрезаем выбросы
    data[column] = np.where(data[column] > upper_bound, upper_bound, data[column])
    data[column] = np.where(data[column] < lower_bound, lower_bound, data[column])

    # Визуализация после обработки выбросов
    plt.subplot(1, 2, 2)
    sns.boxplot(data[column])
    plt.title(f'После обработки выбросов ({column})')
    plt.xlabel(column)

    plt.tight_layout()
    plt.show()

# Обработка и визуализация для 'Event count'
handle_outliers_and_visualize(data, 'Event count')

# Обработка и визуализация для 'Average engagement time per session'
handle_outliers_and_visualize(data, 'Average engagement time per session')

# Вывод обработанных данных
print("Обработанные данные:")
data.head(5)
```


#### **2. Обработка категориальных переменных**

Категориальные переменные, такие как `Session primary channel group`, требуют преобразования в числовой формат для использования в моделях машинного обучения. Одним из наиболее распространенных методов является One-Hot Encoding.

##### **2.1. One-Hot Encoding**
One-Hot Encoding преобразует категориальные переменные в бинарные векторы, где каждая категория представляется отдельным столбцом. Это позволяет избежать введения искусственной порядковой зависимости между категориями.



```python
# Target Encoding (для SARIMAX)
from category_encoders import TargetEncoder

# One-Hot Encoding
data = pd.get_dummies(data, columns=['Session primary channel group'], drop_first=True)

# Label Encoding (для LightGBM)
label_encoder = LabelEncoder()
if 'Session primary channel group' in data.columns:
  data['Session primary channel group_encoded'] = label_encoder.fit_transform(data['Session primary channel group'])

target_encoder = TargetEncoder()
if 'Session primary channel group' in data.columns:
  data['Session primary channel group_encoded_target'] = target_encoder.fit_transform(data['Session primary channel group'], data['Engagement rate'])
```


**Преимущества One-Hot Encoding:**
1. **Отсутствие искусственной порядковой зависимости:** Категории не получают числовых значений, которые могут быть интерпретированы как имеющие порядок.
2. **Совместимость с моделями:** Большинство алгоритмов машинного обучения лучше работают с бинарными признаками.
3. **Упрощение интерпретации:** Каждая категория представлена отдельным признаком, что делает данные более прозрачными.

```python
# Предположим, что data_cleaned — это ваш DataFrame
# Разделение данных на признаки и целевую переменную
X = data_cleaned.drop(columns=['DateTime', 'Users'])  # Признаки (исключаем целевую переменную и временную метку)
y = data_cleaned['Users']  # Целевая переменная

# Разделение данных на обучающую и тестовую выборки (80:20)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Нормализация данных для моделей, чувствительных к масштабу (LSTM, GRU, CNN)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Разделение данных для ARIMA
y_train_arima = y_train
y_test_arima = y_test

# Разделение данных для SARIMAX
exog_train_sarimax = X_train[['Engaged sessions per user', 'Engagement rate']]
exog_test_sarimax = X_test[['Engaged sessions per user', 'Engagement rate']]

# Разделение данных для XGBoost и LightGBM
X_train_ml = X_train[['Engaged sessions per user', 'Engagement rate', 'Event count']]  # Без целевой переменной
X_test_ml = X_test[['Engaged sessions per user', 'Engagement rate', 'Event count']]

# Разделение данных для Random Forest
X_train_rf = X_train
X_test_rf = X_test

# Разделение данных для LSTM и GRU
sequence_length = 24  # Длина временного окна (например, 24 часа)
batch_size = 32  # Размер батча

# Преобразуем данные в последовательности для LSTM/GRU
train_generator = TimeseriesGenerator(
    data=X_train_scaled,
    targets=y_train.values,
    length=sequence_length,
    batch_size=batch_size
)

test_generator = TimeseriesGenerator(
    data=X_test_scaled,
    targets=y_test.values,
    length=sequence_length,
    batch_size=batch_size
)

# Удаление избыточных и малозначимых признаков
columns_to_drop = [
    'Session primary channel group_Organic Search',  # Низкая значимость
    'Session primary channel group_Organic Social',  # Низкая значимость
    'Session primary channel group_Referral',  # Низкая значимость
    'Month',  # Низкая значимость
]

# Удаляем признаки
X_train_reduced = X_train.drop(columns=columns_to_drop)
X_test_reduced = X_test.drop(columns=columns_to_drop)

# Нормализация данных для CNN
X_train_reduced_scaled = scaler.fit_transform(X_train_reduced)
X_test_reduced_scaled = scaler.transform(X_test_reduced)

# Разделение данных для CNN
def create_tf_dataset(data, targets, sequence_length, batch_size):
    dataset = tf.data.Dataset.from_tensor_slices((data, targets))
    dataset = dataset.window(sequence_length, shift=1, drop_remainder=True)
    dataset = dataset.flat_map(lambda x, y: tf.data.Dataset.zip((x.batch(sequence_length), y.batch(sequence_length))))
    dataset = dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return dataset

# Создание датасетов для CNN
X_train_cnn = create_tf_dataset(X_train_reduced_scaled, y_train.values, sequence_length, batch_size)
X_test_cnn = create_tf_dataset(X_test_reduced_scaled, y_test.values, sequence_length, batch_size)

# Вывод размеров выборок
print(f"Обучающая выборка для ARIMA: {len(y_train_arima)}, Тестовая выборка для ARIMA: {len(y_test_arima)}")
print(f"Обучающая выборка для SARIMAX: {exog_train_sarimax.shape}, Тестовая выборка для SARIMAX: {exog_test_sarimax.shape}")
print(f"Обучающая выборка для XGBoost/LightGBM: {X_train_ml.shape}, Тестовая выборка: {X_test_ml.shape}")
print(f"Обучающая выборка для Random Forest: {X_train_rf.shape}, Тестовая выборка: {X_test_rf.shape}")
print(f"Обучающая выборка для LSTM/GRU: {len(train_generator)} батчей, Тестовая выборка: {len(test_generator)} батчей")
print(f"Обучающая выборка для CNN: {len(list(X_train_cnn))} батчей, Тестовая выборка: {len(list(X_test_cnn))} батчей")
```

```python
# Нормализация целевой переменной (если необходимо)
scaler_y = StandardScaler()
y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1, 1)).flatten()
y_test_scaled = scaler_y.transform(y_test.values.reshape(-1, 1)).flatten()

# Проверка на стационарность с помощью теста Дики-Фуллера
adf_test = adfuller(y_train_scaled)
print(f"ADF Statistic: {adf_test[0]}")
print(f"p-value: {adf_test[1]}")
if adf_test[1] > 0.05:
    print("Ряд не стационарен. Рекомендуется дифференцирование.")
else:
    print("Ряд стационарен.")

# Графики автокорреляции (ACF) и частичной автокорреляции (PACF)
plot_acf(y_train_scaled, lags=40)
plt.title('График автокорреляции (ACF)')
plt.show()

plot_pacf(y_train_scaled, lags=40)
plt.title('График частичной автокорреляции (PACF)')
plt.show()

# Автоматический подбор параметров SARIMAX с помощью auto_arima
auto_model = auto_arima(
    y_train_scaled,
    exogenous=exog_train_sarimax,
    seasonal=True,
    m=7,  # Сезонность (например, недельная)
    trace=True,
    error_action='ignore',
    suppress_warnings=True
)
print(auto_model.summary())

# Обучение модели SARIMAX с подобранными параметрами
model = SARIMAX(
    endog=y_train_scaled,
    exog=exog_train_sarimax,
    order=auto_model.order,  # Используем параметры из auto_arima
    seasonal_order=auto_model.seasonal_order
)
results = model.fit(disp=False)

# Сохранение модели
joblib.dump(results, 'sarimax_model.pkl')

# Предсказание на тестовой выборке
forecast = results.get_forecast(steps=len(exog_test_sarimax), exog=exog_test_sarimax)
forecast_mean = forecast.predicted_mean

# Денормализация предсказаний
forecast_denorm = scaler_y.inverse_transform(forecast_mean.values.reshape(-1, 1)).flatten()

# Денормализация реальных значений
y_test_denorm = scaler_y.inverse_transform(y_test_scaled.reshape(-1, 1)).flatten()

# Оценка качества модели
mae = mean_absolute_error(y_test_denorm, forecast_denorm)
rmse = np.sqrt(mean_squared_error(y_test_denorm, forecast_denorm))
mape = mean_absolute_percentage_error(y_test_denorm, forecast_denorm)
print(f"MAE на тестовой выборке: {mae}")
print(f"RMSE на тестовой выборке: {rmse}")
print(f"MAPE на тестовой выборке: {mape}")

# Построение графика реальных данных и предсказаний
plt.figure(figsize=(10, 6))
plt.plot(y_test_denorm, label='Реальные данные', color='blue')
plt.plot(forecast_denorm, label='Предсказания', color='red', linestyle='--')
plt.title('Реальные данные vs Предсказания')
plt.xlabel('Время')
plt.ylabel('Users')
plt.legend()
plt.show()

# Анализ остатков
residuals = results.resid  # Остатки модели

# Тест Шапиро-Уилка на нормальность остатков
shapiro_test = shapiro(residuals)
print(f"Shapiro-Wilk test: p-value = {shapiro_test.pvalue}")

# Гистограмма остатков
plt.figure(figsize=(8, 5))
plt.hist(residuals, bins=30, edgecolor='black', color='lightblue')
plt.title('Гистограмма остатков')
plt.xlabel('Остатки')
plt.ylabel('Частота')
plt.show()

# График остатков во времени
plt.figure(figsize=(10, 6))
plt.plot(residuals)
plt.title('Остатки модели во времени')
plt.xlabel('Время')
plt.ylabel('Остатки')
plt.show()

# Кросс-валидация на временных рядах
tscv = TimeSeriesSplit(n_splits=5)
for train_index, test_index in tscv.split(y_train_scaled):
    cv_train, cv_test = y_train_scaled[train_index], y_train_scaled[test_index]
    cv_train_X, cv_test_X = exog_train_sarimax.iloc[train_index], exog_train_sarimax.iloc[test_index]

    # Обучение модели на cv_train
    cv_model = SARIMAX(
        endog=cv_train,
        exog=cv_train_X,
        order=auto_model.order,
        seasonal_order=auto_model.seasonal_order
    )
    cv_results = cv_model.fit(disp=False)

    # Прогнозирование на cv_test
    cv_forecast = cv_results.get_forecast(steps=len(cv_test_X), exog=cv_test_X)
    cv_forecast_mean = cv_forecast.predicted_mean

    # Оценка качества на cv_test
    cv_mae = mean_absolute_error(cv_test, cv_forecast_mean)
    print(f"MAE на кросс-валидации: {cv_mae}")

# Прогнозирование на будущее (пример)
future_steps = 30  # Количество шагов для прогноза
future_exog = exog_test_sarimax.iloc[:future_steps]  # Используем последние данные для прогноза
future_forecast = results.get_forecast(steps=future_steps, exog=future_exog)
future_forecast_mean = future_forecast.predicted_mean

# Денормализация прогноза
future_forecast_denorm = scaler_y.inverse_transform(future_forecast_mean.values.reshape(-1, 1)).flatten()

# График прогноза на будущее
plt.figure(figsize=(10, 6))
plt.plot(future_forecast_denorm, label='Прогноз на будущее', color='green', linestyle='--')
plt.title('Прогноз на будущее')
plt.xlabel('Время')
plt.ylabel('Users')
plt.legend()
plt.show()
```
