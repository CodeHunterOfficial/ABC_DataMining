## **1. Базовые команды (для новичков)**

Эти команды необходимы для повседневной работы с Git:

### **Настройка**
- `git config`: Настройка параметров Git.
  - Пример: `git config --global user.name "Your Name"`
  - Пример: `git config --global user.email "your.email@example.com"`

### **Инициализация репозитория**
- `git init`: Создание нового локального репозитория.
- `git clone <url>`: Клонирование удаленного репозитория.

### **Работа с файлами**
- `git add <file>`: Добавление изменений в индекс (staging area).
- `git commit -m "message"`: Фиксация изменений в репозитории.
- `git status`: Просмотр состояния рабочей директории и индекса.
- `git diff`: Просмотр изменений в файлах.

### **Работа с историей**
- `git log`: Просмотр истории коммитов.
- `git show <commit-hash>`: Просмотр деталей конкретного коммита.

### **Работа с удаленным репозиторием**
- `git push`: Отправка изменений в удаленный репозиторий.
- `git pull`: Получение изменений из удаленного репозитория.
- `git fetch`: Загрузка изменений из удаленного репозитория без слияния.

---

## **2. Команды для работы с ветками**

Ветки — это основа параллельной разработки. Эти команды помогут вам управлять ветками:

- `git branch`: Управление ветками.
  - `git branch`: Просмотр списка веток.
  - `git branch <branch-name>`: Создание новой ветки.
  - `git branch -d <branch-name>`: Удаление ветки.
- `git checkout <branch-name>`: Переключение между ветками.
- `git switch <branch-name>`: Альтернатива `checkout` для переключения веток (начиная с Git 2.23).
- `git merge <branch-name>`: Слияние одной ветки с другой.
- `git rebase <branch-name>`: Перенос коммитов одной ветки на другую.

---

## **3. Команды для управления изменениями**

Эти команды помогают исправлять ошибки или отменять изменения:

- `git reset`: Откат изменений (см. подробное описание выше).
  - `git reset --soft HEAD~1`: Откат последнего коммита, сохраняя изменения.
  - `git reset --hard HEAD~1`: Полный откат последнего коммита.
- `git revert <commit-hash>`: Безопасная отмена изменений через создание нового коммита.
- `git stash`: Временное сохранение изменений.
  - `git stash save "message"`: Сохранение изменений.
  - `git stash pop`: Применение последних сохраненных изменений.
  - `git stash list`: Просмотр списка сохраненных изменений.

---

## **4. Продвинутые команды**

Эти команды полезны для более сложных сценариев:

### **Управление удаленными репозиториями**
- `git remote`: Управление удаленными репозиториями.
  - `git remote -v`: Просмотр списка удаленных репозиториев.
  - `git remote add <name> <url>`: Добавление нового удаленного репозитория.
  - `git remote remove <name>`: Удаление удаленного репозитория.

### **Теги**
- `git tag`: Работа с тегами (метками версий).
  - `git tag <tag-name>`: Создание тега.
  - `git push --tags`: Отправка тегов в удаленный репозиторий.

### **Перебазирование и интерактивные команды**
- `git rebase -i`: Интерактивное перебазирование (редактирование истории коммитов).
- `git cherry-pick <commit-hash>`: Применение одного коммита из другой ветки.

### **Поиск и анализ**
- `git blame <file>`: Просмотр авторов строк в файле.
- `git bisect`: Поиск бага методом двоичного поиска.

---

## **5. Команды для работы с конфликтами**

Конфликты возникают при слиянии веток или применении изменений. Вот как с ними работать:
- `git merge --abort`: Отмена слияния в случае конфликта.
- `git mergetool`: Использование графического инструмента для разрешения конфликтов.

---

## **6. Дополнительные команды**

Эти команды не являются обязательными, но могут быть полезны:

- `git clean`: Удаление неотслеживаемых файлов.
  - `git clean -n`: Просмотр файлов, которые будут удалены.
  - `git clean -fd`: Удаление файлов и папок.
- `git reflog`: Просмотр истории всех действий в репозитории (полезно для восстановления потерянных коммитов).


# Лекция: Git Config — подробное руководство

`git config` — это одна из самых важных команд в Git, которая позволяет настроить параметры работы с репозиторием. Она используется для управления конфигурацией Git на разных уровнях: глобальном, локальном и системном. В этой лекции мы рассмотрим все аспекты работы с `git config`, включая основные настройки, уровни конфигурации, примеры использования и продвинутые возможности.

---

## **1. Основные концепции**

### **Что такое Git Config?**
Git хранит свои настройки в файлах конфигурации. Эти файлы содержат ключи и значения, которые определяют поведение Git. Например:
- Имя пользователя.
- Адрес электронной почты.
- Поведение пейджера (например, `less`).
- Настройки для работы с удаленным репозиторием.

Файлы конфигурации могут быть расположены на нескольких уровнях:
1. **Системный уровень** (`--system`) — применяется ко всем пользователям и всем репозиториям на компьютере.
2. **Глобальный уровень** (`--global`) — применяется ко всем репозиториям текущего пользователя.
3. **Локальный уровень** (`--local`) — применяется только к конкретному репозиторию.

---

## **2. Уровни конфигурации**

### **1. Системный уровень**
Файл конфигурации: `/etc/gitconfig` (на Linux/macOS) или `%ProgramFiles%\Git\etc\gitconfig` (на Windows).

- Применяется ко всем пользователям системы.
- Изменения требуют прав администратора.
- Пример настройки:
  ```bash
  git config --system user.name "System User"
  ```

### **2. Глобальный уровень**
Файл конфигурации: `~/.gitconfig` или `~/.config/git/config`.

- Применяется ко всем репозиториям текущего пользователя.
- Настройки глобального уровня переопределяют системные.
- Пример настройки:
  ```bash
  git config --global user.name "Your Name"
  git config --global user.email "your.email@example.com"
  ```

### **3. Локальный уровень**
Файл конфигурации: `.git/config` в корне репозитория.

- Применяется только к конкретному репозиторию.
- Настройки локального уровня переопределяют глобальные.
- Пример настройки:
  ```bash
  git config --local core.editor "nano"
  ```

---

## **3. Основные настройки**

### **1. Имя пользователя и email**
Эти настройки используются для подписи коммитов. Без них Git не позволит создать коммит.

```bash
git config --global user.name "Your Name"
git config --global user.email "your.email@example.com"
```

### **2. Редактор по умолчанию**
Git использует текстовый редактор для написания сообщений коммитов, слияний и других задач.

- Установка редактора:
  ```bash
  git config --global core.editor "nano"
  ```
- Возможные варианты:
  - `nano`
  - `vim`
  - `code` (Visual Studio Code)
  - `subl` (Sublime Text)

### **3. Пейджер (pager)**
Пейджер используется для вывода длинных данных, таких как `git log`.

- Отключение пейджера:
  ```bash
  git config --global core.pager ""
  ```
- Настройка пейджера:
  ```bash
  git config --global core.pager "less -R"
  ```

### **4. Цветовая схема**
Git может использовать цвета для улучшения читаемости вывода.

- Включение цветов:
  ```bash
  git config --global color.ui true
  ```
- Настройка цветов для конкретных команд:
  ```bash
  git config --global color.status auto
  git config --global color.branch auto
  ```

### **5. Алиасы (псевдонимы)**
Алиасы позволяют создавать сокращения для часто используемых команд.

- Создание алиаса:
  ```bash
  git config --global alias.co checkout
  git config --global alias.br branch
  git config --global alias.ci commit
  git config --global alias.st status
  ```
- Теперь можно использовать:
  ```bash
  git co main
  git st
  ```

---

## **4. Просмотр и редактирование конфигурации**

### **1. Просмотр текущих настроек**
- Просмотр всех настроек:
  ```bash
  git config --list
  ```
- Просмотр настроек определенного уровня:
  ```bash
  git config --global --list
  git config --local --list
  ```

### **2. Получение значения конкретного параметра**
```bash
git config user.name
git config user.email
```

### **3. Редактирование файла конфигурации**
- Открытие файла конфигурации в редакторе:
  ```bash
  git config --global --edit
  ```
- Ручное редактирование файла:
  - Глобальный: `~/.gitconfig`
  - Локальный: `.git/config`

---

## **5. Продвинутые настройки**

### **1. Настройка прокси**
Если вы работаете через прокси, настройте его:

```bash
git config --global http.proxy http://proxy.example.com:8080
git config --global https.proxy https://proxy.example.com:8080
```

Удаление прокси:
```bash
git config --global --unset http.proxy
git config --global --unset https.proxy
```

### **2. Кэширование учетных данных**
Для удобства работы с удаленными репозиториями можно настроить кэширование паролей.

- Включение кэширования на 1 час:
  ```bash
  git config --global credential.helper cache
  ```
- Включение кэширования на определенное время (например, 2 часа):
  ```bash
  git config --global credential.helper 'cache --timeout=7200'
  ```

### **3. Использование SSH вместо HTTPS**
Для работы с удаленными репозиториями через SSH:

- Убедитесь, что SSH-ключ создан:
  ```bash
  ssh-keygen -t rsa -b 4096 -C "your.email@example.com"
  ```
- Добавьте ключ в SSH-агент:
  ```bash
  eval "$(ssh-agent -s)"
  ssh-add ~/.ssh/id_rsa
  ```
- Настройте Git для использования SSH:
  ```bash
  git remote set-url origin git@github.com:username/repo.git
  ```

### **4. Настройка push.default**
Определяет, как Git обрабатывает команду `git push` при работе с несколькими ветками.

- Настройка режима:
  ```bash
  git config --global push.default simple
  ```
- Возможные значения:
  - `simple`: Push только текущей ветки.
  - `matching`: Push всех веток с совпадающими именами.
  - `current`: Push только текущей ветки, даже если нет соответствующей удаленной ветки.

---

## **6. Полезные примеры**

### **1. Удаление настройки**
```bash
git config --global --unset user.email
```

### **2. Установка нескольких значений**
Например, настройка нескольких пушей для одного репозитория:
```bash
git config --add remote.origin.push refs/heads/main
git config --add remote.origin.push refs/heads/feature
```

### **3. Настройка шаблона сообщений коммитов**
Создайте файл `.gitmessage.txt` с шаблоном сообщений:
```
# Type: feat, fix, docs, style, refactor, test, chore
# Subject: Short description of the change
#
# Body: Detailed explanation of the change
```
Настройте Git для использования шаблона:
```bash
git config --global commit.template ~/.gitmessage.txt
```

---

## **7. Заключение**

`git config` — это мощный инструмент для настройки Git под ваши нужды. Он позволяет управлять поведением Git на разных уровнях и оптимизировать рабочий процесс. Начните с базовых настроек (имя, email, редактор), а затем постепенно переходите к продвинутым параметрам (прокси, кэширование, алиасы).


# Лекция: Git Init — подробное руководство

Команда `git init` является одной из самых фундаментальных в Git. Она используется для создания нового репозитория, что является первым шагом при работе с системой контроля версий. В этой лекции мы рассмотрим все аспекты работы с `git init`, включая основные концепции, параметры команды, примеры использования и нюансы.

---

## **1. Основные концепции**

### **Что такое Git Init?**
Команда `git init` инициализирует новый Git-репозиторий. Это означает, что она создает скрытую директорию `.git`, которая содержит всю необходимую информацию для управления версиями проекта:
- Конфигурационные файлы.
- Историю коммитов.
- Информацию о ветках и тегах.
- Индекс (staging area).

После выполнения `git init` ваша рабочая директория становится Git-репозиторием, и вы можете начать отслеживать изменения в файлах.

---

## **2. Как работает `git init`**

### **1. Создание нового репозитория**
Когда вы выполняете команду `git init`, Git создает директорию `.git` в текущей папке. Эта директория содержит все метаданные репозитория.

**Структура директории `.git`:**
```
.git/
├── branches/
├── config
├── description
├── HEAD
├── hooks/
├── info/
├── objects/
└── refs/
```

- **`config`**: Файл конфигурации для репозитория.
- **`HEAD`**: Указывает на текущую ветку или коммит.
- **`objects/`**: Хранит объекты Git (blobs, trees, commits).
- **`refs/`**: Содержит ссылки на ветки и теги.

### **2. Типы инициализации**
Git позволяет инициализировать репозиторий двумя способами:
1. **В текущей директории**:
   ```bash
   git init
   ```
   - Создает репозиторий в текущей папке.
   - Полезно, если вы хотите начать отслеживать существующий проект.

2. **В новой директории**:
   ```bash
   git init <directory-name>
   ```
   - Создает новую папку и инициализирует репозиторий внутри неё.
   - Полезно, если вы начинаете проект с нуля.

---

## **3. Примеры использования**

### **1. Инициализация репозитория в текущей директории**
Если у вас уже есть проект, и вы хотите начать его отслеживание:
```bash
cd /path/to/project
git init
```
Это создаст директорию `.git` в текущей папке.

### **2. Инициализация репозитория в новой директории**
Если вы хотите создать новый проект:
```bash
git init my-new-project
```
Это создаст папку `my-new-project` и инициализирует репозиторий внутри неё.

### **3. Инициализация пустого репозитория**
Если вы хотите создать пустой репозиторий без рабочей директории (например, для bare-репозитория):
```bash
git init --bare
```
- Bare-репозиторий не имеет рабочей директории и используется как удаленный репозиторий для хранения данных.

---

## **4. Нюансы и особенности**

### **1. Повторная инициализация**
Если вы случайно выполните `git init` в уже инициализированной директории, ничего страшного не произойдет. Git просто сообщит, что репозиторий уже существует:
```
Reinitialized existing Git repository in /path/to/repo/.git/
```

### **2. Переопределение существующего репозитория**
Если вы хотите пересоздать репозиторий (например, чтобы очистить историю), удалите директорию `.git`:
```bash
rm -rf .git
git init
```
**Внимание:** Это приведет к потере всей истории репозитория!

### **3. Настройка после инициализации**
После выполнения `git init` рекомендуется настроить базовые параметры:
```bash
git config user.name "Your Name"
git config user.email "your.email@example.com"
```

---

## **5. Продвинутые параметры**

### **1. Инициализация bare-репозитория**
Bare-репозиторий используется для хранения данных и не имеет рабочей директории. Он часто применяется для удаленных репозиториев:
```bash
git init --bare my-repo.git
```

### **2. Указание шаблона для инициализации**
Git позволяет использовать шаблон для инициализации репозитория. Шаблон может содержать предопределенные файлы или настройки:
```bash
git init --template=/path/to/template
```

### **3. Инициализация с определенной веткой по умолчанию**
Начиная с Git 2.28, можно указать имя ветки по умолчанию:
```bash
git init -b main
```
Это создаст ветку `main` вместо устаревшей `master`.

---

## **6. Практические сценарии**

### **Сценарий 1: Начало работы с новым проектом**
1. Создайте новую директорию:
   ```bash
   mkdir my-project
   cd my-project
   ```
2. Инициализируйте репозиторий:
   ```bash
   git init
   ```
3. Добавьте файлы и сделайте первый коммит:
   ```bash
   echo "# My Project" > README.md
   git add README.md
   git commit -m "Initial commit"
   ```

### **Сценарий 2: Отслеживание существующего проекта**
1. Перейдите в директорию проекта:
   ```bash
   cd /path/to/existing-project
   ```
2. Инициализируйте репозиторий:
   ```bash
   git init
   ```
3. Добавьте файлы и сделайте первый коммит:
   ```bash
   git add .
   git commit -m "Initial commit"
   ```

### **Сценарий 3: Создание bare-репозитория**
1. Создайте bare-репозиторий:
   ```bash
   git init --bare my-repo.git
   ```
2. Добавьте его как удаленный репозиторий:
   ```bash
   git remote add origin /path/to/my-repo.git
   ```

---

## **7. Заключение**

`git init` — это простая, но важная команда, которая открывает возможности для работы с Git. Она позволяет начать отслеживать изменения в проекте и создавать репозитории с нуля. Понимание её работы и параметров поможет вам эффективно использовать Git в повседневной разработке.


# Лекция: Git Clone — подробное руководство

Команда `git clone` используется для создания локальной копии удаленного репозитория. Это один из самых важных инструментов в Git, так как он позволяет разработчикам загружать проекты с таких платформ, как GitHub, GitLab или Bitbucket, и начинать работать с ними локально. В этой лекции мы рассмотрим все аспекты работы с `git clone`, включая основные концепции, параметры команды, примеры использования и нюансы.

---

## **1. Основные концепции**

### **Что такое Git Clone?**
Команда `git clone` создает полную копию удаленного репозитория на вашем компьютере. Она выполняет следующие действия:
1. Создает новую директорию с именем репозитория.
2. Инициализирует Git-репозиторий (создает директорию `.git`).
3. Скачивает все файлы и историю коммитов.
4. Настройку связи с удаленным репозиторием (по умолчанию называется `origin`).

---

## **2. Как работает `git clone`**

### **1. Базовый синтаксис**
```bash
git clone <repository-url>
```
- `<repository-url>` — это адрес удаленного репозитория. Это может быть:
  - HTTPS URL: `https://github.com/username/repo.git`
  - SSH URL: `git@github.com:username/repo.git`

### **2. Что происходит при клонировании?**
1. Git создает новую директорию с именем репозитория.
2. Загружает все файлы и метаданные из удаленного репозитория.
3. Устанавливает ссылку на удаленный репозиторий под именем `origin`.
4. Переключается на ветку по умолчанию (обычно `main` или `master`).

---

## **3. Примеры использования**

### **1. Клонирование с использованием HTTPS**
HTTPS — это наиболее распространенный способ клонирования репозиториев:
```bash
git clone https://github.com/username/repo.git
```
- Не требует настройки SSH-ключей.
- Может запросить имя пользователя и пароль (или токен).

### **2. Клонирование с использованием SSH**
SSH обеспечивает безопасное подключение и не требует ввода пароля каждый раз:
```bash
git clone git@github.com:username/repo.git
```
- Требует настройки SSH-ключей на вашем компьютере.

### **3. Клонирование в другую директорию**
Если вы хотите клонировать репозиторий в конкретную папку:
```bash
git clone <repository-url> <directory-name>
```
Пример:
```bash
git clone https://github.com/username/repo.git my-project
```

### **4. Клонирование только определенной ветки**
Если вам нужна только одна ветка, используйте флаг `--branch` (`-b`):
```bash
git clone -b <branch-name> <repository-url>
```
Пример:
```bash
git clone -b feature-branch https://github.com/username/repo.git
```

---

## **4. Нюансы и особенности**

### **1. Клонирование без истории**
Если вам нужно клонировать только последний коммит (например, для экономии места), используйте флаг `--depth`:
```bash
git clone --depth 1 <repository-url>
```
Это создаст "мелкий" клон (shallow clone) с историей только одного коммита.

### **2. Клонирование bare-репозитория**
Bare-репозиторий — это репозиторий без рабочей директории. Он используется для хранения данных и часто применяется в серверах CI/CD:
```bash
git clone --bare <repository-url>
```

### **3. Переименование удаленного репозитория**
По умолчанию удаленный репозиторий называется `origin`. Вы можете изменить это имя с помощью флага `--origin`:
```bash
git clone --origin upstream <repository-url>
```

### **4. Клонирование с рекурсивным подмодулем**
Если репозиторий содержит подмодули, используйте флаг `--recurse-submodules`:
```bash
git clone --recurse-submodules <repository-url>
```
Это автоматически клонирует и инициализирует подмодули.

---

## **5. Проверка после клонирования**

После выполнения `git clone` рекомендуется проверить состояние репозитория:
1. Просмотреть список веток:
   ```bash
   git branch -a
   ```
2. Проверить текущую ветку:
   ```bash
   git status
   ```
3. Просмотреть историю коммитов:
   ```bash
   git log
   ```

---

## **6. Продвинутые сценарии**

### **1. Клонирование частичного репозитория**
Если репозиторий очень большой, можно клонировать только часть файлов:
```bash
git clone --filter=blob:none <repository-url>
```
Это скачает только структуру репозитория без содержимого файлов.

### **2. Клонирование через прокси**
Если вы работаете через прокси, настройте его перед клонированием:
```bash
git config --global http.proxy http://proxy.example.com:8080
git clone <repository-url>
```

### **3. Клонирование с переопределением авторизации**
Если вам нужно использовать другой аккаунт для клонирования, добавьте параметры авторизации в URL:
```bash
git clone https://username:token@github.com/username/repo.git
```

---

## **7. Заключение**

`git clone` — это простая, но мощная команда, которая позволяет разработчикам быстро начать работу с удаленными репозиториями. Она предоставляет множество параметров для настройки процесса клонирования и поддерживает различные сценарии использования.


# Лекция: Git Add `<file>` — подробное руководство

Команда `git add <file>` является ключевым инструментом в Git, который используется для добавления изменений в **staging area** (индекс). Это промежуточная область, где вы готовите изменения перед фиксацией (`git commit`). В этой лекции мы рассмотрим все аспекты работы с `git add <file>`, включая основные концепции, параметры команды, примеры использования и нюансы.

---

## **1. Основные концепции**

### **Что такое Git Add `<file>`?**
Команда `git add <file>` добавляет конкретный файл из рабочей директории в **staging area**. После этого изменения становятся доступными для коммита. Без выполнения `git add` Git не будет учитывать изменения при создании нового коммита.

### **Три состояния файлов в Git**
1. **Working Directory** — текущее состояние файлов в вашей рабочей директории.
2. **Staging Area (Index)** — промежуточная область, куда добавляются изменения перед коммитом.
3. **Repository** — финальное место, где хранятся зафиксированные изменения (`git commit`).

Команда `git add <file>` перемещает изменения из **Working Directory** в **Staging Area**.

---

## **2. Как работает `git add <file>`**

### **1. Добавление нового файла**
Если вы создали новый файл, Git изначально не отслеживает его. Чтобы начать отслеживание, используйте:
```bash
git add <file>
```

### **2. Добавление изменений в существующих файлах**
Если вы изменили уже отслеживаемый файл, Git отметит эти изменения как "modified". Чтобы добавить их в staging area:
```bash
git add <file>
```

### **3. Добавление удаленных файлов**
Если вы удалили файл, Git пометит это действие как "deleted". Чтобы добавить удаление в staging area:
```bash
git add <file>
```

---

## **3. Примеры использования**

### **1. Добавление нового файла**
1. Создайте новый файл:
   ```bash
   echo "Hello, Git!" > hello.txt
   ```
2. Добавьте файл в staging area:
   ```bash
   git add hello.txt
   ```

### **2. Добавление изменений в существующем файле**
1. Измените содержимое файла:
   ```bash
   echo "Updated content" >> README.md
   ```
2. Добавьте изменения:
   ```bash
   git add README.md
   ```

### **3. Добавление удаленного файла**
1. Удалите файл:
   ```bash
   rm old.txt
   ```
2. Добавьте удаление в staging area:
   ```bash
   git add old.txt
   ```

---

## **4. Проверка состояния**

После выполнения `git add <file>` рекомендуется проверять состояние репозитория:
```bash
git status
```
- Файлы в **staging area** будут отмечены как "Changes to be committed".
- Файлы с изменениями, которые еще не добавлены, будут отмечены как "Changes not staged for commit".

---

## **5. Нюансы и особенности**

### **1. Игнорирование файлов**
Git может игнорировать определенные файлы или папки с помощью файла `.gitignore`. Например:
```plaintext
# Ignore all .log files
*.log

# Ignore the build directory
/build/
```
Файлы, указанные в `.gitignore`, не будут добавлены в staging area, даже если вы выполните `git add <file>`.

### **2. Отмена добавления**
Если вы случайно добавили файл в staging area, его можно удалить:
```bash
git reset HEAD <file>
```

### **3. Добавление пустых каталогов**
Git не отслеживает пустые каталоги. Чтобы добавить пустой каталог, создайте внутри него файл `.gitkeep`:
```bash
mkdir empty-folder
touch empty-folder/.gitkeep
git add empty-folder/.gitkeep
```

### **4. Работа с большими файлами**
Для работы с большими файлами рекомендуется использовать [Git LFS](https://git-lfs.github.com/) (Large File Storage). Это позволяет хранить большие файлы вне основного репозитория.

---

## **6. Продвинутые сценарии**

### **1. Добавление изменений в конкретной строке**
Используйте `git add -p`, чтобы добавить только часть изменений из файла:
```bash
git add -p <file>
```
Эта команда запускает интерактивный режим, где вы можете выбрать конкретные изменения (hunks) для добавления:
- `y`: Добавить часть.
- `n`: Пропустить часть.
- `q`: Завершить процесс.

### **2. Автоматическое добавление изменений**
Вы можете настроить Git для автоматического добавления изменений:
```bash
git config --global alias.addall '!git add -A && git commit -m "Auto-commit"'
```

### **3. Добавление изменений с шаблонами**
Если вы хотите использовать шаблоны для сообщений коммитов, настройте их через `git config`:
```bash
git config commit.template ~/.gitmessage.txt
```

---

## **7. Заключение**

`git add <file>` — это простая, но мощная команда, которая позволяет контролировать, какие изменения будут включены в следующий коммит. Она предоставляет гибкость для управления изменениями в проекте и помогает поддерживать чистую историю коммитов.


# Лекция: Git Commit `-m "message"` — подробное руководство

Команда `git commit -m "message"` является одной из ключевых команд в Git, которая используется для фиксации изменений в репозитории. Она создает новый коммит, который представляет собой снимок состояния вашего проекта на определенный момент времени. В этой лекции мы рассмотрим все аспекты работы с `git commit`, включая основные концепции, параметры команды, примеры использования и нюансы.

---

## **1. Основные концепции**

### **Что такое Git Commit?**
Команда `git commit` создает новый коммит в репозитории. Коммит — это запись, которая содержит:
1. **Снимок изменений** (данные из staging area).
2. **Сообщение коммита** (описание изменений).
3. **Метаданные** (автор, дата, хеш коммита и ссылка на родительский коммит).

Флаг `-m "message"` позволяет добавить сообщение к коммиту прямо в командной строке.

### **Три состояния файлов в Git**
1. **Working Directory** — текущее состояние файлов.
2. **Staging Area (Index)** — промежуточная область, где готовятся изменения.
3. **Repository** — финальное место, где хранятся зафиксированные изменения.

Команда `git commit` перемещает изменения из **staging area** в **repository**.

---

## **2. Как работает `git commit -m "message"`**

### **1. Создание коммита**
Когда вы выполняете команду:
```bash
git commit -m "message"
```
Git создает новый коммит с указанным сообщением. Этот коммит становится частью истории репозитория.

### **2. Что происходит при коммите?**
1. Git создает уникальный хеш для нового коммита.
2. Сохраняет снимок изменений из staging area.
3. Добавляет метаданные (автор, дата, сообщение).
4. Обновляет указатель текущей ветки (например, `HEAD`) на новый коммит.

---

## **3. Примеры использования**

### **1. Базовый коммит**
Добавьте изменения в staging area и создайте коммит:
```bash
git add README.md
git commit -m "Update README file"
```

### **2. Многострочное сообщение**
Если сообщение длинное, используйте кавычки для переноса строки:
```bash
git commit -m "Add new feature:
- Implement user authentication
- Add login page
- Fix minor bugs"
```

### **3. Коммит без предварительного добавления в staging area**
Если вы хотите пропустить этап `git add`, используйте флаг `-a`:
```bash
git commit -am "Fix typo in documentation"
```
Флаг `-a` автоматически добавляет изменения в отслеживаемых файлах (новые файлы не будут включены).

---

## **4. Нюансы и особенности**

### **1. Важность сообщения коммита**
Сообщение коммита должно быть:
- **Кратким**: Первая строка должна быть заголовком (до 50 символов).
- **Описательным**: По возможности объясните, что и почему было сделано.
- **Структурированным**: Для длинных сообщений используйте пустую строку после заголовка.

Пример хорошего сообщения:
```plaintext
Add user authentication feature

- Implement login and registration pages
- Add password validation logic
- Update API endpoints for user management
```

### **2. Отмена последнего коммита**
Если вы случайно сделали коммит с ошибкой, его можно отменить:
```bash
git reset --soft HEAD~1
```
Это вернет изменения в staging area, но сохранит их.

### **3. Изменение последнего коммита**
Если вы забыли добавить изменения или допустили ошибку в сообщении, используйте:
```bash
git commit --amend -m "Updated message"
```
Это заменит последний коммит новым.

---

## **5. Продвинутые сценарии**

### **1. Интерактивное создание коммита**
Если вы хотите добавить только часть изменений из файла, используйте:
```bash
git add -p
git commit -m "Partially update code"
```

### **2. Подпись коммита GPG**
Для повышения безопасности можно подписывать коммиты с помощью GPG:
```bash
git commit -S -m "Secure commit with GPG signature"
```

### **3. Автоматическое форматирование сообщений**
Настройте шаблон для сообщений коммитов:
```bash
git config commit.template ~/.gitmessage.txt
```
Создайте файл `~/.gitmessage.txt` с шаблоном:
```plaintext
# Type: feat, fix, docs, style, refactor, test, chore
# Subject: Short description of the change
#
# Body: Detailed explanation of the change
```

---

## **6. Заключение**

`git commit -m "message"` — это простая, но мощная команда, которая позволяет фиксировать изменения в репозитории. Правильное использование сообщений коммитов помогает поддерживать чистую и понятную историю разработки.


# Лекция: Git Status — подробное руководство

Команда `git status` является одной из самых важных и часто используемых команд в Git. Она предоставляет информацию о текущем состоянии вашего репозитория, включая изменения в рабочей директории, staging area (индекс) и ветках. В этой лекции мы рассмотрим все аспекты работы с `git status`, включая основные концепции, примеры использования, нюансы и продвинутые сценарии.

---

## **1. Основные концепции**

### **Что такое Git Status?**
Команда `git status` показывает текущее состояние вашего репозитория. Она отвечает на следующие вопросы:
1. Какие файлы были изменены, но еще не добавлены в staging area?
2. Какие файлы находятся в staging area и готовы к коммиту?
3. На какой ветке вы находитесь?
4. Есть ли конфликты или другие особенности состояния?

### **Три состояния файлов в Git**
1. **Untracked** — файлы, которые Git еще не отслеживает.
2. **Modified** — файлы, которые были изменены, но не добавлены в staging area.
3. **Staged** — файлы, которые добавлены в staging area и готовы к коммиту.

---

## **2. Как работает `git status`**

### **1. Базовый вывод**
При выполнении команды:
```bash
git status
```
Git покажет:
- Текущую ветку.
- Список изменений в рабочей директории.
- Список изменений в staging area.
- Информацию о состоянии слияния или конфликтов (если они есть).

### **2. Пример вывода**
Пример вывода команды `git status`:
```plaintext
On branch main
Your branch is up to date with 'origin/main'.

Changes to be committed:
  (use "git restore --staged <file>..." to unstage)
        modified:   README.md

Changes not staged for commit:
  (use "git add <file>..." to update what will be committed)
  (use "git restore <file>..." to discard changes in working directory)
        modified:   index.html

Untracked files:
  (use "git add <file>..." to include in what will be committed)
        new-file.txt
```

---

## **3. Примеры использования**

### **1. Проверка текущего состояния**
Чтобы увидеть текущее состояние репозитория:
```bash
git status
```

### **2. Короткий формат вывода**
Для получения краткой информации используйте флаг `-s`:
```bash
git status -s
```
Пример вывода:
```plaintext
 M README.md
?? new-file.txt
```
- `M` — измененный файл.
- `??` — неотслеживаемый файл.

### **3. Проверка состояния после слияния**
Если слияние вызвало конфликты, `git status` покажет:
```plaintext
You have unmerged paths.
  (fix conflicts and run "git commit")
  (use "git merge --abort" to abort the merge)

Unmerged paths:
  (use "git add <file>..." to mark resolution)
        both modified:   file.txt
```

---

## **4. Нюансы и особенности**

### **1. Отображение неотслеживаемых файлов**
По умолчанию `git status` показывает неотслеживаемые файлы. Если вы хотите скрыть их, используйте флаг `--untracked-files=no`:
```bash
git status --untracked-files=no
```

### **2. Проверка состояния веток**
`git status` также показывает, насколько ваша локальная ветка отстает или опережает удаленную:
```plaintext
Your branch is ahead of 'origin/main' by 1 commit.
  (use "git push" to publish your local commits)
```

### **3. Игнорирование файлов**
Файлы, указанные в `.gitignore`, не будут отображаться в разделе "Untracked files". Например:
```plaintext
# .gitignore
*.log
/build/
```
Файлы `.log` и папка `build/` будут игнорироваться.

---

## **5. Продвинутые сценарии**

### **1. Проверка состояния конкретного файла**
Чтобы проверить состояние конкретного файла:
```bash
git status <file>
```

### **2. Проверка состояния с детализацией**
Для более подробной информации используйте флаг `-v`:
```bash
git status -v
```
Это покажет дифф изменений, которые находятся в staging area.

### **3. Автоматическое исправление состояния**
Git предлагает команды для исправления состояния прямо в выводе `git status`. Например:
- `git add <file>` — добавить файл в staging area.
- `git restore <file>` — отменить изменения в рабочей директории.
- `git restore --staged <file>` — убрать файл из staging area.

---

## **6. Заключение**

`git status` — это простая, но мощная команда, которая помогает вам понять текущее состояние вашего репозитория. Она предоставляет четкое представление о том, какие изменения требуют внимания, и помогает поддерживать чистую и организованную рабочую среду.


# Лекция: Git Diff — подробное руководство

Команда `git diff` является мощным инструментом в Git, который позволяет просматривать различия между файлами или состояниями репозитория. Она помогает разработчикам анализировать изменения, находить ошибки и лучше понимать текущее состояние проекта. В этой лекции мы рассмотрим все аспекты работы с `git diff`, включая основные концепции, примеры использования, нюансы и продвинутые сценарии.

---

## **1. Основные концепции**

### **Что такое Git Diff?**
Команда `git diff` показывает различия между:
1. Рабочей директорией и staging area.
2. Staging area и последним коммитом.
3. Различными коммитами, ветками или тегами.

Она отображает изменения в виде "diff" (разности), где добавленные строки помечаются знаком `+`, а удаленные — знаком `-`.

---

## **2. Как работает `git diff`**

### **1. Базовый вывод**
При выполнении команды:
```bash
git diff
```
Git покажет различия между рабочей директорией и staging area. Это полезно для анализа изменений, которые еще не были добавлены в staging area.

### **2. Пример вывода**
Пример вывода команды `git diff`:
```plaintext
diff --git a/README.md b/README.md
index 1234567..89abcde 100644
--- a/README.md
+++ b/README.md
@@ -1,3 +1,4 @@
 # Project Title

-Initial content
+Updated content
```
- `a/README.md` и `b/README.md` — это старая и новая версии файла.
- `-Initial content` — удаленная строка.

---

## **3. Основные режимы работы**

### **1. Различия между рабочей директорией и staging area**

Показывает изменения в рабочей директории, которые еще не добавлены в staging area.

```bash
git diff <commit-hash-1> <commit-hash-2>
```
Показывает изменения между двумя конкретными коммитами.

### **4. Различия между ветками**
```bash
git diff <branch-1>..<branch-2>
```
Показывает изменения между двумя ветками.

# Лекция: Git Log — подробное руководство

Команда `git log` является одним из ключевых инструментов в Git, который позволяет просматривать историю коммитов в репозитории. Она предоставляет детальную информацию о каждом коммите, включая хеш, автора, дату, сообщение и изменения. В этой лекции мы рассмотрим все аспекты работы с `git log`, включая основные концепции, примеры использования, нюансы и продвинутые сценарии.

---

## **1. Основные концепции**

### **Что такое Git Log?**
Команда `git log` показывает историю коммитов в текущей ветке или указанной области репозитория. Она отображает:
1. **Хеш коммита** (уникальный идентификатор).
2. **Автора** (имя и email).
3. **Дата и время** создания коммита.
4. **Сообщение коммита** (описание изменений).

### **Типы вывода**
- **Полный вывод**: Показывает всю доступную информацию о каждом коммите.
- **Краткий вывод**: Отображает только хеши и сообщения.
- **Графический вывод**: Визуализирует историю в виде графа.

---

## **2. Как работает `git log`**

### **1. Базовый вывод**
При выполнении команды:
```bash
git log
```
Git покажет историю коммитов в обратном хронологическом порядке (от последнего к первому). Пример вывода:
```plaintext
commit abc123456789abcdef123456789abcdef12345678
Author: John Doe <john.doe@example.com>
Date:   Mon Oct 23 10:00:00 2023 +0300

    Add user authentication feature

commit def987654321fedcba987654321fedcba98765432
Author: Jane Smith <jane.smith@example.com>
Date:   Sun Oct 22 15:30:00 2023 +0300

    Fix typo in README
```

### **2. Выход из режима просмотра**
Если вывод занимает несколько страниц, Git использует пейджер (например, `less`). Для выхода нажмите клавишу **`q`**.

---

## **3. Основные параметры**

Пример вывода:
```plaintext
* abc1234 Add user authentication feature
* def9876 Fix typo in README
```

### **3. Ограничение количества коммитов**
Чтобы увидеть последние N коммитов, используйте флаг `-n`:

### **4. Отображение изменений**
Для просмотра изменений в каждом коммите используйте флаг `-p`:
```bash
git log -p
```

### **5. Статистика изменений**
Для получения сводной информации о количестве изменений используйте флаг `--stat`:
```bash
git log --stat
```
Пример вывода:
```plaintext
 README.md | 2 +-
 index.html | 3 ++-
 2 files changed, 3 insertions(+), 2 deletions(-)
```

---

## **4. Примеры использования**

### **1. Просмотр истории конкретного файла**

- **Коммит** (`commit`): Показывает изменения, внесенные этим коммитом.

# Лекция: Git Push — подробное руководство

Команда `git push` является ключевым инструментом в Git, который используется для отправки локальных изменений в удаленный репозиторий. Она позволяет делиться своей работой с другими участниками команды или публиковать изменения на платформах, таких как GitHub, GitLab или Bitbucket. В этой лекции мы рассмотрим все аспекты работы с `git push`, включая основные концепции, примеры использования, нюансы и продвинутые сценарии.

---

## **1. Основные концепции**

### **Что такое Git Push?**
Команда `git push` отправляет локальные коммиты из вашей рабочей копии в удаленный репозиторий. Она обновляет удаленные ветки, чтобы они соответствовали локальным изменениям.

### **Основные компоненты**
1. **Локальная ветка**: Ветка, из которой вы отправляете изменения.
2. **Удаленная ветка**: Ветка в удаленном репозитории, куда вы отправляете изменения.
3. **Связь между ветками**: Установленная связь между локальной и удаленной веткой (например, через `origin/main`).

---

## **2. Как работает `git push`**

### **1. Базовый синтаксис**
```bash
git push <remote> <branch>
```
- `<remote>` — имя удаленного репозитория (обычно `origin`).
- `<branch>` — имя ветки, которую вы хотите отправить.

Пример:
```bash
git push origin main
```

### **2. Процесс выполнения**
1. Git проверяет, есть ли связь между локальной и удаленной веткой.
2. Если связь существует, Git отправляет локальные коммиты в удаленный репозиторий.
3. Удаленная ветка обновляется до последнего состояния локальной ветки.

---

## **3. Примеры использования**

### **1. Отправка изменений в текущую ветку**
Если вы находитесь в ветке `main` и хотите отправить изменения:
```bash
git push
```
Эта команда работает только если:
- Связь между локальной и удаленной веткой уже установлена.
- Имя удаленной ветки совпадает с локальной.

### **2. Отправка изменений в конкретную ветку**
Чтобы отправить изменения в определенную ветку:
```bash
git push origin feature-branch
```

### **3. Установка связи между ветками**
Если связь между локальной и удаленной веткой еще не установлена, используйте флаг `-u` (или `--set-upstream`):
```bash
git push -u origin main
```
После этого Git запомнит связь, и в будущем можно будет использовать просто:
```bash
git push
```

---

## **4. Нюансы и особенности**

### **1. Отправка всех веток**
Чтобы отправить все локальные ветки в удаленный репозиторий:
```bash
git push --all
```

### **2. Отправка тегов**
По умолчанию `git push` не отправляет теги. Чтобы отправить их:
```bash
git push --tags
```
Чтобы отправить теги вместе с веткой:
```bash
git push origin main --tags
```

### **3. Принудительная отправка**
Если удаленная ветка содержит коммиты, которых нет в локальной ветке, Git откажется выполнять `push`. Чтобы принудительно обновить удаленную ветку:
```bash
git push --force
```
**Важно:** Используйте `--force` с осторожностью, так как это может перезаписать историю и затереть чужие изменения.

### **4. Безопасная принудительная отправка**
Для безопасной принудительной отправки используйте флаг `--force-with-lease`:
```bash
git push --force-with-lease
```
Эта команда проверяет, что удаленная ветка не была изменена другими участниками.

---

## **5. Продвинутые сценарии**

### **1. Отправка изменений через прокси**
Если вы работаете через прокси, настройте его перед отправкой:
```bash
git config --global http.proxy http://proxy.example.com:8080
git push
```

### **2. Отправка изменений по SSH**
Если вы используете SSH для подключения к удаленному репозиторию:
```bash
git push git@github.com:username/repo.git main
```

### **3. Отправка изменений в новый удаленный репозиторий**
Если вы создали новый удаленный репозиторий, добавьте его как remote:
```bash
git remote add origin https://github.com/username/repo.git
git push -u origin main
```

### **4. Отправка изменений с переименованием ветки**
Если вы хотите отправить локальную ветку под другим именем:
```bash
git push origin local-branch:remote-branch
```

---

## **6. Решение проблем**

### **1. Конфликты при отправке**
Если удаленная ветка содержит коммиты, которых нет в локальной ветке:
1. Получите последние изменения:
   ```bash
   git pull
   ```
2. Разрешите конфликты, если они есть.
3. Повторите отправку:
   ```bash
   git push
   ```

### **2. Ошибка аутентификации**
Если Git запрашивает пароль или токен:
1. Убедитесь, что вы авторизованы на платформе (например, GitHub).
2. Для HTTPS используйте токен вместо пароля.
3. Для SSH настройте SSH-ключи.

---

## **7. Заключение**

`git push` — это важный инструмент для совместной работы над проектами. Она позволяет делиться изменениями с другими участниками и поддерживать актуальность удаленного репозитория. Правильное использование этой команды помогает избежать конфликтов и поддерживать чистую историю разработки.


# Лекция: Git Pull — подробное руководство

Команда `git pull` является одной из ключевых команд в Git, которая используется для получения изменений из удаленного репозитория и их интеграции в вашу локальную рабочую копию. Она объединяет две операции: `git fetch` (получение изменений) и `git merge` (слияние изменений). В этой лекции мы рассмотрим все аспекты работы с `git pull`, включая основные концепции, примеры использования, нюансы и продвинутые сценарии.

---

## **1. Основные концепции**

### **Что такое Git Pull?**
Команда `git pull` выполняет две задачи:
1. **Получение изменений** (`git fetch`): Скачивает последние изменения из удаленного репозитория.
2. **Слияние изменений** (`git merge`): Интегрирует полученные изменения в вашу текущую ветку.

По сути, `git pull` — это комбинация двух команд:
```bash
git fetch <remote>
git merge <branch>
```

---

## **2. Как работает `git pull`**

### **1. Базовый синтаксис**
```bash
git pull <remote> <branch>
```
- `<remote>` — имя удаленного репозитория (обычно `origin`).
- `<branch>` — имя ветки, из которой вы хотите получить изменения.

Пример:
```bash
git pull origin main
```

### **2. Процесс выполнения**
1. Git проверяет удаленное хранилище на наличие новых коммитов.
2. Если новые коммиты найдены, они загружаются в локальный репозиторий.
3. Загруженные изменения сливаются с текущей веткой.

---

## **3. Примеры использования**

### **1. Получение изменений из текущей ветки**
Если вы находитесь в ветке `main` и хотите получить изменения:
```bash
git pull
```
Эта команда работает только если:
- Связь между локальной и удаленной веткой уже установлена.
- Имя удаленной ветки совпадает с локальной.

### **2. Получение изменений из конкретной ветки**
Чтобы получить изменения из определенной ветки:
```bash
git pull origin feature-branch
```

### **3. Использование rebase вместо merge**
По умолчанию `git pull` выполняет слияние через `merge`. Если вы предпочитаете перебазирование (`rebase`), используйте флаг `--rebase`:
```bash
git pull --rebase
```
Это создаст более линейную историю.

---

## **4. Нюансы и особенности**

### **1. Установка связи между ветками**
Если связь между локальной и удаленной веткой еще не установлена, добавьте ее:
```bash
git branch --set-upstream-to=origin/main main
```
После этого можно использовать просто:
```bash
git pull
```

### **2. Автоматическое использование rebase**
Чтобы всегда использовать `rebase` вместо `merge` при выполнении `git pull`, настройте конфигурацию:
```bash
git config pull.rebase true
```

### **3. Проверка изменений перед слиянием**
Чтобы увидеть, какие изменения будут получены, используйте:
```bash
git fetch
git log HEAD..origin/main --oneline
```

### **4. Обработка конфликтов**
Если при слиянии возникают конфликты:
1. Разрешите конфликты вручную.
2. Добавьте исправленные файлы в staging area:
   ```bash
   git add <file>
   ```
3. Завершите слияние:
   ```bash
   git merge --continue
   ```

---

## **5. Продвинутые сценарии**

### **1. Получение изменений без слияния**
Если вы хотите только скачать изменения, но не сливать их:
```bash
git fetch
```
Затем вы можете вручную выполнить слияние или перебазирование:
```bash
git merge origin/main
```
или
```bash
git rebase origin/main
```

### **2. Получение изменений через прокси**
Если вы работаете через прокси, настройте его перед выполнением `git pull`:
```bash
git config --global http.proxy http://proxy.example.com:8080
git pull
```

### **3. Получение изменений по SSH**
Если вы используете SSH для подключения к удаленному репозиторию:
```bash
git pull git@github.com:username/repo.git main
```

### **4. Получение изменений с переименованием ветки**
Если вы хотите получить изменения из удаленной ветки под другим именем:
```bash
git pull origin remote-branch:local-branch
```

---

## **6. Решение проблем**

### **1. Конфликты при слиянии**
Если возникают конфликты при выполнении `git pull`:
1. Разрешите конфликты вручную.
2. Добавьте исправленные файлы:
   ```bash
   git add <file>
   ```
3. Завершите слияние:
   ```bash
   git merge --continue
   ```

### **2. Ошибка аутентификации**
Если Git запрашивает пароль или токен:
1. Убедитесь, что вы авторизованы на платформе (например, GitHub).
2. Для HTTPS используйте токен вместо пароля.
3. Для SSH настройте SSH-ключи.

### **3. Откат изменений после pull**
Если вы случайно выполнили `git pull` и хотите отменить слияние:
```bash
git merge --abort
```

---

## **7. Заключение**

`git pull` — это важный инструмент для синхронизации локальной копии проекта с удаленным репозиторием. Она позволяет получать последние изменения и поддерживать актуальность вашего кода. Правильное использование этой команды помогает избежать конфликтов и поддерживать чистую историю разработки.


# Лекция: Git Fetch — подробное руководство

Команда `git fetch` является одной из ключевых команд в Git, которая используется для получения изменений из удаленного репозитория. В отличие от `git pull`, она **не сливает** изменения с вашей локальной веткой, а только загружает их. Это позволяет вам просмотреть и проанализировать изменения перед их интеграцией. В этой лекции мы рассмотрим все аспекты работы с `git fetch`, включая основные концепции, примеры использования, нюансы и продвинутые сценарии.

---

## **1. Основные концепции**

### **Что такое Git Fetch?**
Команда `git fetch` загружает последние изменения из удаленного репозитория, но не применяет их к вашей локальной рабочей копии. Она:
1. Обновляет ссылки на удаленные ветки (например, `origin/main`).
2. Скачивает новые коммиты, теги и другие объекты.

Важно понимать, что `git fetch` **не изменяет** вашу текущую рабочую директорию или локальные ветки. Это делает её безопасным инструментом для анализа изменений перед слиянием.

---

## **2. Как работает `git fetch`**

### **1. Базовый синтаксис**
```bash
git fetch <remote>
```
- `<remote>` — имя удаленного репозитория (обычно `origin`).

Пример:
```bash
git fetch origin
```

### **2. Процесс выполнения**
1. Git связывается с удаленным репозиторием.
2. Загружает все новые коммиты, которые отсутствуют в локальном репозитории.
3. Обновляет ссылки на удаленные ветки (например, `origin/main`).

После выполнения `git fetch` вы можете:
- Просмотреть изменения.
- Выполнить слияние или перебазирование вручную.

---

## **3. Примеры использования**

### **1. Получение всех изменений**
Чтобы получить все изменения из удаленного репозитория:
```bash
git fetch
```
Эта команда работает с удаленным репозиторием по умолчанию (`origin`).

### **2. Получение изменений из конкретной ветки**
Чтобы получить изменения только из определенной ветки:
```bash
git fetch origin main
```

### **3. Получение всех веток**
Чтобы загрузить изменения из всех веток удаленного репозитория:
```bash
git fetch --all
```

### **4. Получение тегов**
По умолчанию `git fetch` не загружает теги. Чтобы загрузить их:
```bash
git fetch --tags
```

---

## **4. Нюансы и особенности**

### **1. Разница между `git fetch` и `git pull`**
- **`git fetch`**: Только загружает изменения, не изменяя локальные ветки.
- **`git pull`**: Загружает изменения и сразу сливает их с текущей веткой.

### **2. Просмотр изменений после fetch**
После выполнения `git fetch` вы можете просмотреть изменения:
```bash
git log HEAD..origin/main --oneline
```
Это покажет коммиты, которые есть в удаленной ветке, но отсутствуют в локальной.

### **3. Обновление удаленных веток**
Если удаленная ветка была переименована или удалена, используйте:
```bash
git fetch --prune
```
Это удалит ссылки на удаленные ветки, которых больше нет в удаленном репозитории.

### **4. Настройка автоматического prune**
Чтобы автоматически удалять неактуальные ссылки при каждом `fetch`, настройте:
```bash
git config fetch.prune true
```

---

## **5. Продвинутые сценарии**

### **1. Фильтрация изменений**
Чтобы загрузить только определенные файлы или части истории, используйте:
```bash
git fetch --filter=blob:none
```
Это создаст "мелкий" клон (shallow clone) без содержимого файлов.

### **2. Получение изменений через прокси**
Если вы работаете через прокси, настройте его перед выполнением `git fetch`:
```bash
git config --global http.proxy http://proxy.example.com:8080
git fetch
```

### **3. Получение изменений по SSH**
Если вы используете SSH для подключения к удаленному репозиторию:
```bash
git fetch git@github.com:username/repo.git
```

### **4. Создание локальной копии удаленной ветки**
Чтобы создать локальную копию удаленной ветки:
```bash
git fetch origin feature-branch
git checkout -b feature-branch origin/feature-branch
```

---

## **6. Решение проблем**

### **1. Ошибка аутентификации**
Если Git запрашивает пароль или токен:
1. Убедитесь, что вы авторизованы на платформе (например, GitHub).
2. Для HTTPS используйте токен вместо пароля.
3. Для SSH настройте SSH-ключи.

### **2. Конфликты при слиянии**
Если вы выполняете слияние после `git fetch` и возникают конфликты:
1. Разрешите конфликты вручную.
2. Добавьте исправленные файлы:
   ```bash
   git add <file>
   ```
3. Завершите слияние:
   ```bash
   git merge --continue
   ```

### **3. Отмена fetch**
Если вы хотите отменить последний `fetch` (например, чтобы удалить загруженные изменения):
```bash
git fetch --prune
```

---

## **7. Заключение**

`git fetch` — это безопасный и мощный инструмент для получения изменений из удаленного репозитория. Она позволяет вам анализировать изменения перед их интеграцией, что особенно важно в командной разработке. Правильное использование этой команды помогает поддерживать чистую историю разработки и избегать конфликтов.


# Лекция: Git Branch — подробное руководство

Команда `git branch` является одной из ключевых команд в Git, которая используется для управления ветками. Ветки позволяют разработчикам работать над различными частями проекта параллельно, не затрагивая основной код. В этой лекции мы рассмотрим все аспекты работы с `git branch`, включая основные концепции, примеры использования, нюансы и продвинутые сценарии.

---

## **1. Основные концепции**

### **Что такое Git Branch?**
Ветка (branch) в Git — это указатель на определенный коммит. Она позволяет разработчикам:
1. Работать над новыми функциями или исправлениями без влияния на основной код.
2. Экспериментировать с изменениями, не затрагивая рабочую версию проекта.
3. Объединять изменения обратно в основную ветку после завершения работы.

По умолчанию Git создает ветку `main` (или `master` в старых репозиториях).

---

## **2. Как работает `git branch`**

### **1. Базовый синтаксис**
```bash
git branch
```
Эта команда показывает список всех локальных веток. Текущая ветка помечается символом `*`.

Пример вывода:
```plaintext
* main
  feature-branch
  bugfix-branch
```

### **2. Создание новой ветки**
Чтобы создать новую ветку:
```bash
git branch <branch-name>
```
Пример:
```bash
git branch feature-branch
```

### **3. Переключение между ветками**
Чтобы переключиться на другую ветку:
```bash
git checkout <branch-name>
```
Начиная с Git 2.23, можно использовать команду `switch`:
```bash
git switch <branch-name>
```

### **4. Удаление ветки**
Чтобы удалить ветку:
```bash
git branch -d <branch-name>
```
Если ветка содержит неслитые изменения, Git откажется её удалять. Для принудительного удаления используйте флаг `-D`:
```bash
git branch -D <branch-name>
```

---

## **3. Примеры использования**

### **1. Просмотр списка веток**
Чтобы увидеть список всех локальных веток:
```bash
git branch
```

Чтобы увидеть список всех веток, включая удаленные:
```bash
git branch -a
```

### **2. Создание и переключение на новую ветку**
Чтобы создать новую ветку и сразу переключиться на неё:
```bash
git checkout -b <branch-name>
```
или
```bash
git switch -c <branch-name>
```

Пример:
```bash
git checkout -b feature-branch
```

### **3. Переименование ветки**
Чтобы переименовать текущую ветку:
```bash
git branch -m <new-branch-name>
```

Пример:
```bash
git branch -m feature-branch new-feature-branch
```

### **4. Удаление ветки**
Чтобы удалить локальную ветку:
```bash
git branch -d feature-branch
```

Чтобы удалить удаленную ветку:
```bash
git push origin --delete <branch-name>
```

---

## **4. Нюансы и особенности**

### **1. Разница между локальными и удаленными ветками**
- **Локальные ветки**: Хранятся на вашем компьютере.
- **Удаленные ветки**: Хранятся в удаленном репозитории (например, на GitHub).

Чтобы увидеть все ветки, включая удаленные:
```bash
git branch -a
```

### **2. Отслеживание удаленных веток**
При клонировании репозитория Git автоматически создает ссылки на удаленные ветки (например, `origin/main`). Чтобы создать локальную копию удаленной ветки:
```bash
git checkout -b <branch-name> origin/<branch-name>
```

### **3. Настройка связи между ветками**
Чтобы связать локальную ветку с удаленной:
```bash
git branch --set-upstream-to=origin/<branch-name>
```

### **4. Просмотр информации о ветках**
Чтобы увидеть последний коммит в каждой ветке:
```bash
git branch -v
```

Чтобы увидеть ветки, которые уже слиты в текущую:
```bash
git branch --merged
```

Чтобы увидеть ветки, которые еще не слиты:
```bash
git branch --no-merged
```

---

## **5. Продвинутые сценарии**

### **1. Создание ветки из определенного коммита**
Чтобы создать новую ветку из конкретного коммита:
```bash
git branch <branch-name> <commit-hash>
```

Пример:
```bash
git branch hotfix-branch abc1234
```

### **2. Создание ветки из тега**
Чтобы создать новую ветку из тега:
```bash
git branch <branch-name> <tag-name>
```

Пример:
```bash
git branch release-branch v1.0.0
```

### **3. Перебазирование ветки**
Чтобы перенести коммиты одной ветки на другую:
```bash
git rebase <base-branch>
```

Пример:
```bash
git checkout feature-branch
git rebase main
```

### **4. Слияние веток**
Чтобы слить одну ветку в другую:
```bash
git checkout main
git merge feature-branch
```

---

## **6. Решение проблем**

### **1. Конфликты при слиянии**
Если возникают конфликты при выполнении слияния:
1. Разрешите конфликты вручную.
2. Добавьте исправленные файлы:
   ```bash
   git add <file>
   ```
3. Завершите слияние:
   ```bash
   git merge --continue
   ```

### **2. Ошибка при удалении ветки**
Если Git отказывается удалить ветку из-за неслитых изменений:
```bash
git branch -D <branch-name>
```

### **3. Восстановление удаленной ветки**
Если вы случайно удалили ветку, её можно восстановить по хешу последнего коммита:
```bash
git checkout -b <branch-name> <commit-hash>
```

---

## **7. Заключение**

`git branch` — это мощный инструмент для управления ветками в Git. Она позволяет разработчикам эффективно организовывать работу над проектом, экспериментировать с изменениями и поддерживать чистую историю разработки. Правильное использование этой команды помогает избежать конфликтов и поддерживать порядок в репозитории.


# Лекция: Git Checkout и Git Switch — подробное руководство

Команды `git checkout` и `git switch` используются для переключения между ветками или восстановления файлов в рабочей директории. Начиная с Git 2.23, команда `git switch` была добавлена как более безопасная альтернатива `git checkout` для управления ветками. В этой лекции мы рассмотрим все аспекты работы с этими командами, включая основные концепции, примеры использования, нюансы и продвинутые сценарии.

---

## **1. Основные концепции**

### **Что такое Git Checkout?**
Команда `git checkout` изначально была универсальной командой для:
1. Переключения между ветками.
2. Восстановления файлов из staging area или последнего коммита.
3. Создания новых веток (в сочетании с флагом `-b`).

Пример:
```bash
git checkout main
```

### **Что такое Git Switch?**
Команда `git switch` была добавлена в Git 2.23 для упрощения работы с ветками. Она выполняет только одну задачу: переключение между ветками. Это делает её более понятной и безопасной по сравнению с `git checkout`.

Пример:
```bash
git switch main
```

---

## **2. Как работают эти команды**

### **1. Переключение между ветками**
Обе команды позволяют переключаться между существующими ветками.

#### Использование `git checkout`:
```bash
git checkout <branch-name>
```

#### Использование `git switch`:
```bash
git switch <branch-name>
```

Пример:
```bash
git checkout feature-branch
git switch feature-branch
```

### **2. Создание новой ветки**
Обе команды могут создавать новую ветку и сразу переключаться на неё.

#### Использование `git checkout`:
```bash
git checkout -b <new-branch-name>
```

#### Использование `git switch`:
```bash
git switch -c <new-branch-name>
```

Пример:
```bash
git checkout -b new-feature
git switch -c new-feature
```

---

## **3. Примеры использования**

### **1. Переключение на существующую ветку**
Чтобы переключиться на ветку `main`:

#### С помощью `git checkout`:
```bash
git checkout main
```

#### С помощью `git switch`:
```bash
git switch main
```

### **2. Создание и переключение на новую ветку**
Чтобы создать новую ветку `feature-branch` и сразу переключиться на неё:

#### С помощью `git checkout`:
```bash
git checkout -b feature-branch
```

#### С помощью `git switch`:
```bash
git switch -c feature-branch
```

### **3. Восстановление файла**
Команда `git checkout` также используется для восстановления файлов из staging area или последнего коммита. Команда `git switch` не поддерживает эту функциональность.

#### Восстановление файла из последнего коммита:
```bash
git checkout -- <file>
```

Пример:
```bash
git checkout -- README.md
```

#### Восстановление файла из staging area:
```bash
git restore <file>
```
(Начиная с Git 2.23 рекомендуется использовать `git restore` вместо `git checkout`.)

---

## **4. Нюансы и особенности**

### **1. Разница между `git checkout` и `git switch`**
- **`git checkout`**: Универсальная команда, которая может использоваться для переключения веток, восстановления файлов и других задач.
- **`git switch`**: Специализированная команда, предназначенная только для работы с ветками.

### **2. Безопасность**
- `git switch` более безопасна, так как она не позволяет случайно выполнить операции, связанные с восстановлением файлов.
- `git checkout` может быть опасной, если вы случайно восстановите файлы, потеряв несохраненные изменения.

### **3. Создание ветки из определенного коммита**
Чтобы создать новую ветку из конкретного коммита:

#### С помощью `git checkout`:
```bash
git checkout -b <branch-name> <commit-hash>
```

#### С помощью `git switch`:
```bash
git switch -c <branch-name> <commit-hash>
```

Пример:
```bash
git checkout -b hotfix-branch abc1234
git switch -c hotfix-branch abc1234
```

### **4. Переключение на удаленную ветку**
Чтобы переключиться на удаленную ветку, сначала нужно создать локальную копию:

#### С помощью `git checkout`:
```bash
git checkout -b <branch-name> origin/<branch-name>
```

#### С помощью `git switch`:
```bash
git switch <branch-name>
```

Пример:
```bash
git checkout -b feature-branch origin/feature-branch
git switch feature-branch
```

## **5. Продвинутые сценарии**

### **1. Отмена переключения**
Если вы начали переключение, но хотите отменить его (например, из-за конфликтов):

#### С помощью `git checkout`:
```bash
git checkout -
```

#### С помощью `git switch`:
```bash
git switch -
```
Эта команда переключается обратно на предыдущую ветку.

### **2. Создание ветки из тега**
Чтобы создать новую ветку из тега:

#### С помощью `git checkout`:
```bash
git checkout -b <branch-name> <tag-name>
```

#### С помощью `git switch`:
```bash
git switch -c <branch-name> <tag-name>
```

Пример:
```bash
git checkout -b release-branch v1.0.0
git switch -c release-branch v1.0.0
```

### **3. Переключение с принудительным откатом изменений**
Если у вас есть несохраненные изменения, Git откажется выполнить переключение. Чтобы принудительно откатить изменения:

#### С помощью `git checkout`:
```bash
git checkout -f <branch-name>
```

#### С помощью `git switch`:
```bash
git switch -f <branch-name>
```

---

## **6. Решение проблем**

### **1. Конфликты при переключении**
Если у вас есть несохраненные изменения, которые конфликтуют с целевой веткой:
1. Сохраните изменения в stash:
   ```bash
   git stash
   ```
2. Переключитесь на другую ветку:
   ```bash
   git switch <branch-name>
   ```
3. Восстановите изменения:
   ```bash
   git stash pop
   ```

### **2. Ошибка "detached HEAD"**
Если вы переключились на коммит, а не на ветку, Git переходит в состояние "detached HEAD". Это означает, что вы работаете напрямую с коммитом, а не с веткой. Чтобы избежать потери изменений:
1. Создайте новую ветку:
   ```bash
   git switch -c <branch-name>
   ```

---

## **7. Заключение**

`git checkout` и `git switch` — это ключевые инструменты для управления ветками в Git. Хотя `git checkout` остается универсальной командой, `git switch` предоставляет более безопасный и понятный способ работы с ветками. Правильное использование этих команд помогает эффективно организовывать работу над проектом и избегать ошибок.


# Лекция: Git Merge `<branch-name>` — подробное руководство

Команда `git merge` является одной из ключевых команд в Git, которая используется для объединения изменений из одной ветки в другую. Она позволяет интегрировать работу, выполненную в разных ветках, в единую историю. В этой лекции мы рассмотрим все аспекты работы с `git merge`, включая основные концепции, примеры использования, нюансы и продвинутые сценарии.

---

## **1. Основные концепции**

### **Что такое Git Merge?**
Команда `git merge` объединяет изменения из одной ветки в текущую ветку. Она создает новый коммит (merge commit), который содержит изменения из обеих веток.

### **Основные типы слияния**
1. **Fast-forward merge**: Если ветка, в которую вы хотите слить изменения, является прямым продолжением другой ветки, Git просто перемещает указатель текущей ветки на последний коммит другой ветки.
2. **Three-way merge**: Если ветки имеют расхождения, Git создает новый коммит слияния (merge commit), который объединяет изменения из обеих веток.

---

## **2. Как работает `git merge`**

### **1. Базовый синтаксис**
```bash
git merge <branch-name>
```
- `<branch-name>` — имя ветки, которую вы хотите слить с текущей веткой.

Пример:
```bash
git merge feature-branch
```

### **2. Процесс выполнения**
1. Git анализирует историю двух веток.
2. Если возможно, выполняется fast-forward merge.
3. Если невозможно, создается merge commit.

---

## **3. Примеры использования**

### **1. Fast-forward merge**
Если ветка `main` не имеет новых коммитов после создания `feature-branch`, Git выполнит fast-forward merge.

#### Исходное состояние:
```
main: A <- B
feature-branch: A <- B <- C <- D
```

#### Выполнение:
```bash
git checkout main
git merge feature-branch
```

#### Результат:
```
main: A <- B <- C <- D
```

### **2. Three-way merge**
Если ветка `main` имеет новые коммиты после создания `feature-branch`, Git создаст merge commit.

#### Исходное состояние:
```
main: A <- B <- X <- Y
feature-branch: A <- B <- C <- D
```

#### Выполнение:
```bash
git checkout main
git merge feature-branch
```

#### Результат:
```
main: A <- B <- X <- Y <- M
                   \       /
                    C <- D
```
Здесь `M` — это merge commit.

---

## **4. Нюансы и особенности**

### **1. Создание merge commit**
По умолчанию Git создает merge commit при трехстороннем слиянии. Чтобы принудительно создать merge commit даже при fast-forward merge, используйте флаг `--no-ff`:
```bash
git merge --no-ff feature-branch
```
Это полезно для сохранения истории ветвления.

### **2. Отмена слияния**
Если слияние вызвало проблемы, его можно отменить:
```bash
git merge --abort
```

### **3. Конфликты при слиянии**
Если изменения в ветках затрагивают одни и те же строки кода, возникают конфликты. Git помечает их следующим образом:
```plaintext
<<<<<<< HEAD
console.log("Main branch");
=======
console.log("Feature branch");
>>>>>>> feature-branch
```

#### Решение:
1. Откройте файлы с конфликтами.
2. Вручную выберите нужные изменения.
3. Добавьте исправленные файлы в staging area:
   ```bash
   git add <file>
   ```
4. Завершите слияние:
   ```bash
   git merge --continue
   ```

---

## **5. Продвинутые сценарии**

### **1. Слияние с определенным коммитом**
Чтобы слить изменения из конкретного коммита:
```bash
git merge <commit-hash>
```

### **2. Слияние с использованием стратегий**
Git предоставляет несколько стратегий слияния:
- `ours`: Игнорирует изменения из другой ветки.
- `theirs`: Игнорирует изменения из текущей ветки.
- `recursive`: Решает конфликты автоматически.

Пример:
```bash
git merge -s ours feature-branch
```

### **3. Squash слияние**
Чтобы объединить все коммиты из ветки в один коммит:
```bash
git merge --squash feature-branch
git commit -m "Merge feature-branch into main"
```

---

## **6. Решение проблем**

### **1. Конфликты при слиянии**
Если возникают конфликты:
1. Разрешите конфликты вручную.
2. Добавьте исправленные файлы:
   ```bash
   git add <file>
   ```
3. Завершите слияние:
   ```bash
   git merge --continue
   ```

### **2. Отмена слияния**
Если слияние вызвало проблемы:
```bash
git merge --abort
```

### **3. Удаление ветки после слияния**
После успешного слияния ветку можно удалить:
```bash
git branch -d feature-branch
```

---

## **7. Заключение**

`git merge` — это мощный инструмент для объединения изменений из разных веток. Она позволяет эффективно интегрировать работу над новыми функциями или исправлениями в основной код. Правильное использование этой команды помогает поддерживать чистую историю разработки и избегать конфликтов.


# Лекция: Git Rebase `<branch-name>` — подробное руководство

Команда `git rebase` является мощным инструментом в Git, который позволяет перенести коммиты одной ветки на другую. В отличие от `git merge`, которая создает новый коммит слияния, `git rebase` создает более линейную историю, что делает её особенно полезной для поддержания чистой и понятной истории разработки. В этой лекции мы рассмотрим все аспекты работы с `git rebase`, включая основные концепции, примеры использования, нюансы и продвинутые сценарии.

---

## **1. Основные концепции**

### **Что такое Git Rebase?**
Команда `git rebase` переносит коммиты текущей ветки на конец другой ветки. Она:
1. **Перебазирует** (перемещает) коммиты текущей ветки на базу целевой ветки.
2. Создает новые коммиты, которые являются копиями оригинальных коммитов, но основаны на новой базе.

### **Основные преимущества**
- **Линейная история**: История становится более чистой и легкой для понимания.
- **Избегание лишних merge commits**: В отличие от `git merge`, `git rebase` не создает дополнительные коммиты слияния.

### **Важно помнить**
- Не используйте `git rebase` для публичных веток, так как это может перезаписать историю и затереть изменения других участников.

---

## **2. Как работает `git rebase`**

### **1. Базовый синтаксис**
```bash
git rebase <branch-name>
```
- `<branch-name>` — имя ветки, на которую вы хотите перенести коммиты текущей ветки.

Пример:
```bash
git rebase main
```

### **2. Процесс выполнения**
1. Git временно сохраняет изменения из текущей ветки.
2. Перемещает указатель текущей ветки на последний коммит целевой ветки.
3. Применяет сохраненные изменения поверх целевой ветки.

---

## **3. Примеры использования**

### **1. Простое перебазирование**
Если вы хотите перенести коммиты из `feature-branch` на `main`:

#### Исходное состояние:
```
main: A <- B <- C
feature-branch: A <- D <- E
```

#### Выполнение:
```bash
git checkout feature-branch
git rebase main
```

#### Результат:
```
main: A <- B <- C
feature-branch: A <- B <- C <- D' <- E'
```
Здесь `D'` и `E'` — это новые коммиты, основанные на `C`.

### **2. Перебазирование с конфликтами**
Если возникают конфликты при перебазировании:
1. Разрешите конфликты вручную.
2. Добавьте исправленные файлы:
   ```bash
   git add <file>
   ```
3. Продолжите перебазирование:
   ```bash
   git rebase --continue
   ```

### **3. Отмена перебазирования**
Если перебазирование вызвало проблемы, его можно отменить:
```bash
git rebase --abort
```

---

## **4. Нюансы и особенности**

### **1. Интерактивное перебазирование**
Интерактивное перебазирование позволяет редактировать, удалять или объединять коммиты перед их применением. Для этого используйте флаг `-i`:
```bash
git rebase -i <base-commit>
```

Пример:
```bash
git rebase -i HEAD~3
```
Это откроет текстовый редактор с перечнем последних 3 коммитов:
```plaintext
pick abc123 Commit message 1
pick def456 Commit message 2
pick ghi789 Commit message 3
```
Вы можете изменить порядок, удалить коммиты или объединить их (squash).

### **2. Squash коммитов**
Чтобы объединить несколько коммитов в один:
1. Запустите интерактивное перебазирование:
   ```bash
   git rebase -i HEAD~3
   ```
2. Измените команду для коммитов:
   ```plaintext
   pick abc123 Commit message 1
   squash def456 Commit message 2
   squash ghi789 Commit message 3
   ```
3. Сохраните изменения и укажите новое сообщение коммита.

### **3. Перебазирование с принудительной отправкой**
Если вы перебазировали публичную ветку, потребуется принудительная отправка:
```bash
git push --force-with-lease
```
Используйте `--force-with-lease` вместо `--force`, чтобы избежать затирания чужих изменений.

---

## **5. Продвинутые сценарии**

### **1. Перебазирование на определенный коммит**
Чтобы перенести коммиты на конкретный коммит:
```bash
git rebase <commit-hash>
```

### **2. Перебазирование с использованием стратегий**
Git предоставляет несколько стратегий для разрешения конфликтов:
- `ours`: Игнорирует изменения из другой ветки.
- `theirs`: Игнорирует изменения из текущей ветки.

Пример:
```bash
git rebase -X theirs main
```

### **3. Автоматическое разрешение конфликтов**
Для автоматического разрешения конфликтов используйте флаг `--strategy-option`:
```bash
git rebase --strategy-option=theirs main
```

---

## **6. Решение проблем**

### **1. Конфликты при перебазировании**
Если возникают конфликты:
1. Разрешите конфликты вручную.
2. Добавьте исправленные файлы:
   ```bash
   git add <file>
   ```
3. Продолжите перебазирование:
   ```bash
   git rebase --continue
   ```

### **2. Отмена перебазирования**
Если перебазирование вызвало проблемы:
```bash
git rebase --abort
```

### **3. Восстановление после принудительной отправки**
Если вы случайно выполнили `git push --force` и затерли чужие изменения:
1. Получите старую версию ветки:
   ```bash
   git reflog
   ```
2. Восстановите её:
   ```bash
   git reset --hard <commit-hash>
   ```

---

## **7. Заключение**

`git rebase` — это мощный инструмент для создания линейной истории и упрощения работы с ветками. Однако его использование требует осторожности, особенно при работе с публичными ветками. Правильное применение этой команды помогает поддерживать чистую и организованную историю разработки.


# Лекция: Git Reset — подробное руководство

Команда `git reset` является одной из самых мощных и сложных команд в Git. Она позволяет отменять изменения, удалять коммиты или перемещать указатель текущей ветки (HEAD). Однако её использование требует осторожности, так как она может привести к потере данных, если применять её неправильно. В этой лекции мы рассмотрим все аспекты работы с `git reset`, включая основные концепции, примеры использования, нюансы и продвинутые сценарии.

---

## **1. Основные концепции**

### **Что такое Git Reset?**
Команда `git reset` используется для изменения состояния репозитория. Она позволяет:
1. Перемещать указатель текущей ветки (HEAD) на определенный коммит.
2. Отменять изменения в рабочей директории или staging area.
3. Удалять коммиты из истории.

### **Три уровня сброса**
Git предоставляет три режима работы с `git reset`, которые определяются флагами:
1. **`--soft`**: Перемещает указатель HEAD, но оставляет изменения в staging area.
2. **`--mixed`** (по умолчанию): Перемещает указатель HEAD и убирает изменения из staging area, но оставляет их в рабочей директории.
3. **`--hard`**: Полностью удаляет изменения, возвращая репозиторий к указанному состоянию.

---

## **2. Как работает `git reset`**

### **1. Базовый синтаксис**
```bash
git reset [--soft | --mixed | --hard] <commit>
```
- `<commit>` — хеш коммита, на который вы хотите переместить указатель HEAD.

Пример:
```bash
git reset --soft HEAD~1
```

### **2. Процесс выполнения**
1. Git перемещает указатель HEAD на указанный коммит.
2. В зависимости от режима (`--soft`, `--mixed`, `--hard`) изменяет состояние рабочей директории и staging area.

---

## **3. Примеры использования**

### **1. Отмена последнего коммита**
Если вы хотите отменить последний коммит, но сохранить изменения в рабочей директории:

#### Использование `--soft`:
```bash
git reset --soft HEAD~1
```
- Коммит будет удален из истории, но изменения останутся в staging area.

#### Использование `--mixed`:
```bash
git reset --mixed HEAD~1
```
- Коммит будет удален из истории, а изменения перенесены в рабочую директорию.

#### Использование `--hard`:
```bash
git reset --hard HEAD~1
```
- Коммит будет полностью удален вместе с изменениями.

---

### **2. Отмена нескольких коммитов**
Чтобы отменить последние 3 коммита:

#### Использование `--soft`:
```bash
git reset --soft HEAD~3
```

#### Использование `--mixed`:
```bash
git reset --mixed HEAD~3
```

#### Использование `--hard`:
```bash
git reset --hard HEAD~3
```

---

### **3. Сброс до определенного коммита**
Чтобы сбросить состояние репозитория до конкретного коммита:

#### Использование `--soft`:
```bash
git reset --soft abc1234
```

#### Использование `--mixed`:
```bash
git reset --mixed abc1234
```

#### Использование `--hard`:
```bash
git reset --hard abc1234
```

---

## **4. Нюансы и особенности**

### **1. Разница между режимами**
- **`--soft`**: Сохраняет изменения в staging area. Полезно, если вы хотите создать новый коммит с теми же изменениями.
- **`--mixed`**: Убирает изменения из staging area, но оставляет их в рабочей директории. Это режим по умолчанию.
- **`--hard`**: Полностью удаляет изменения. Используйте его с осторожностью, так как это необратимая операция.

### **2. Отмена изменений в рабочей директории**
Если вы хотите отменить изменения в рабочей директории, используйте:
```bash
git restore <file>
```
(Начиная с Git 2.23 рекомендуется использовать `git restore` вместо `git reset`.)

### **3. Откат отправленных коммитов**
Если вы уже отправили коммиты в удаленный репозиторий, использование `git reset` может вызвать проблемы. Вместо этого лучше использовать:
```bash
git revert <commit-hash>
```
Эта команда создает новый коммит, который отменяет изменения предыдущего коммита.

---

## **5. Продвинутые сценарии**

### **1. Сброс до удаленной ветки**
Если вы хотите сбросить локальную ветку до состояния удаленной ветки:
```bash
git fetch origin
git reset --hard origin/main
```

### **2. Отмена merge commit**
Если вы хотите отменить merge commit:
```bash
git reset --hard HEAD~1
```

### **3. Сброс с использованием reflog**
Если вы случайно выполнили `git reset` и потеряли данные, восстановите их с помощью `git reflog`:
```bash
git reflog
git reset --hard <commit-hash>
```

---

## **6. Решение проблем**

### **1. Потеря данных при `--hard`**
Если вы случайно выполнили `git reset --hard` и потеряли изменения:
1. Проверьте журнал `reflog`:
   ```bash
   git reflog
   ```
2. Восстановите потерянные коммиты:
   ```bash
   git reset --hard <commit-hash>
   ```

### **2. Конфликты при принудительной отправке**
Если вы сбросили публичную ветку и попытались отправить изменения:
```bash
git push --force-with-lease
```
Используйте `--force-with-lease`, чтобы избежать затирания чужих изменений.

---

## **7. Заключение**

`git reset` — это мощный инструмент для управления историей репозитория. Однако его использование требует осторожности, особенно при работе с публичными ветками. Правильное применение этой команды помогает поддерживать чистую и организованную историю разработки.


# Лекция: Git Revert `<commit-hash>` — подробное руководство

Команда `git revert` является безопасным способом отмены изменений в Git. В отличие от `git reset`, которая перезаписывает историю, `git revert` создает новый коммит, который отменяет изменения указанного коммита. Это делает её особенно полезной для работы с публичными репозиториями, где нельзя изменять уже опубликованную историю. В этой лекции мы рассмотрим все аспекты работы с `git revert`, включая основные концепции, примеры использования, нюансы и продвинутые сценарии.

---

## **1. Основные концепции**

### **Что такое Git Revert?**
Команда `git revert` создает новый коммит, который отменяет изменения, внесенные указанным коммитом. Она:
1. **Не изменяет историю**: Вместо перезаписи истории добавляет новый коммит.
2. **Безопасна для публичных репозиториев**: Не затирает чужие изменения.
3. **Отменяет конкретный коммит**: Вы можете точно указать, какие изменения нужно отменить.

### **Основные преимущества**
- **Обратимость**: Отмена изменений выполняется через новый коммит, что позволяет легко откатить её при необходимости.
- **Совместимость с публичными ветками**: Подходит для работы в команде, где история не должна быть изменена.

---

## **2. Как работает `git revert`**

### **1. Базовый синтаксис**
```bash
git revert <commit-hash>
```
- `<commit-hash>` — хеш коммита, который вы хотите отменить.

Пример:
```bash
git revert abc1234
```

### **2. Процесс выполнения**
1. Git анализирует указанный коммит.
2. Создает новый коммит, который отменяет изменения этого коммита.
3. Добавляет новый коммит в историю.

---

## **3. Примеры использования**

### **1. Отмена последнего коммита**
Если вы хотите отменить последний коммит:
```bash
git revert HEAD
```

#### Исходное состояние:
```
A <- B <- C (HEAD)
```

#### Результат:
```
A <- B <- C <- D (HEAD)
```
Здесь `D` — это новый коммит, который отменяет изменения из `C`.

### **2. Отмена конкретного коммита**
Если вы хотите отменить коммит `abc1234`:
```bash
git revert abc1234
```

#### Исходное состояние:
```
A <- B <- C <- D <- E (HEAD)
```

#### Результат:
```
A <- B <- C <- D <- E <- F (HEAD)
```
Здесь `F` — это новый коммит, который отменяет изменения из `D`.

### **3. Отмена нескольких коммитов**
Если вы хотите отменить несколько коммитов, выполните `git revert` для каждого из них:
```bash
git revert HEAD~2..HEAD
```
Это отменит последние 2 коммита.

---

## **4. Нюансы и особенности**

### **1. Конфликты при отмене**
Если изменения в указанном коммите конфликтуют с текущим состоянием репозитория:
1. Git остановит процесс и предложит разрешить конфликты.
2. Разрешите конфликты вручную.
3. Добавьте исправленные файлы:
   ```bash
   git add <file>
   ```
4. Завершите отмену:
   ```bash
   git revert --continue
   ```

### **2. Отмена merge commit**
Для отмены merge commit укажите дополнительный флаг `-m`:
```bash
git revert -m 1 <merge-commit-hash>
```
- `-m 1`: Отменяет изменения из первой родительской ветки.
- `-m 2`: Отменяет изменения из второй родительской ветки.

### **3. Отмена отмены**
Если вы случайно выполнили `git revert`, вы можете отменить её:
```bash
git revert <revert-commit-hash>
```

---

## **5. Продвинутые сценарии**

### **1. Отмена диапазона коммитов**
Чтобы отменить несколько коммитов одновременно:
```bash
git revert <oldest-commit>..<latest-commit>
```

Пример:
```bash
git revert abc1234..def5678
```

### **2. Отмена без создания коммита**
Если вы хотите только применить изменения, но не создавать коммит:
```bash
git revert --no-commit <commit-hash>
```

### **3. Отмена с использованием стратегий**
Git предоставляет несколько стратегий для разрешения конфликтов:
- `ours`: Игнорирует изменения из указанного коммита.
- `theirs`: Игнорирует текущие изменения.

Пример:
```bash
git revert -X theirs <commit-hash>
```

---

## **6. Решение проблем**

### **1. Конфликты при отмене**
Если возникают конфликты:
1. Разрешите конфликты вручную.
2. Добавьте исправленные файлы:
   ```bash
   git add <file>
   ```
3. Завершите отмену:
   ```bash
   git revert --continue
   ```

### **2. Отмена отмены**
Если вы случайно выполнили `git revert`, вы можете отменить её:
```bash
git revert <revert-commit-hash>
```

### **3. Отмена отправленных коммитов**
Если вы уже отправили коммиты в удаленный репозиторий, использование `git revert` — это безопасный способ отменить их:
```bash
git revert <commit-hash>
git push
```

---

## **7. Заключение**

`git revert` — это безопасный и надежный инструмент для отмены изменений в Git. Она создает новый коммит, который отменяет изменения указанного коммита, что делает её особенно полезной для работы с публичными репозиториями. Правильное использование этой команды помогает поддерживать чистую и организованную историю разработки.


# Лекция: Git Stash — подробное руководство

Команда `git stash` является мощным инструментом в Git, который позволяет временно сохранять изменения в рабочей директории без создания коммита. Это особенно полезно, когда вы работаете над одной задачей, но вам нужно быстро переключиться на другую задачу или ветку, не завершая текущие изменения. В этой лекции мы рассмотрим все аспекты работы с `git stash`, включая основные концепции, примеры использования, нюансы и продвинутые сценарии.

---

## **1. Основные концепции**

### **Что такое Git Stash?**
Команда `git stash` временно сохраняет («прячет») изменения в рабочей директории и staging area, чтобы вы могли переключиться на другую задачу или ветку. Эти изменения можно восстановить позже.

### **Основные преимущества**
- **Быстрое переключение**: Позволяет быстро переключаться между задачами без необходимости создания временных коммитов.
- **Сохранение состояния**: Сохраняет как отслеживаемые, так и неотслеживаемые файлы (опционально).
- **Восстановление изменений**: Восстанавливает изменения, когда они снова нужны.

---

## **2. Как работает `git stash`**

### **1. Базовый синтаксис**
```bash
git stash
```
Эта команда сохраняет текущие изменения в специальной области (stash).

### **2. Процесс выполнения**
1. Git сохраняет изменения в рабочей директории и staging area.
2. Убирает эти изменения из рабочей директории, возвращая её к состоянию последнего коммита.
3. Создает запись в стеке stash.

---

## **3. Примеры использования**

### **1. Простое сохранение изменений**
Если вы хотите временно сохранить изменения:
```bash
git stash
```

#### Исходное состояние:
```
Working Directory: Modified files
Staging Area: Staged changes
```

#### Результат:
```
Working Directory: Clean (no changes)
Staging Area: Empty
Stash: Saved changes
```

### **2. Сохранение с сообщением**
Чтобы добавить описание к сохраненным изменениям:
```bash
git stash save "Message describing the changes"
```

Пример:
```bash
git stash save "WIP: Feature implementation"
```

### **3. Восстановление изменений**
Чтобы восстановить последние сохраненные изменения:
```bash
git stash apply
```

#### Варианты:
- `git stash pop`: Восстанавливает изменения и удаляет их из стека.
- `git stash apply`: Восстанавливает изменения, но оставляет их в стеке.

---

## **4. Нюансы и особенности**

### **1. Сохранение неотслеживаемых файлов**
По умолчанию `git stash` сохраняет только отслеживаемые файлы. Чтобы сохранить неотслеживаемые файлы, используйте флаг `-u`:
```bash
git stash -u
```

### **2. Сохранение всех изменений**
Чтобы сохранить все изменения, включая игнорируемые файлы (например, `.gitignore`), используйте флаг `-a`:
```bash
git stash -a
```

### **3. Просмотр сохраненных изменений**
Чтобы увидеть список сохраненных изменений:
```bash
git stash list
```

Пример вывода:
```plaintext
stash@{0}: WIP on main: abc1234 Initial commit
stash@{1}: WIP on feature-branch: def5678 Add new feature
```

### **4. Просмотр содержимого конкретного stash**
Чтобы увидеть содержимое конкретного stash:
```bash
git stash show -p stash@{0}
```

---

## **5. Продвинутые сценарии**

### **1. Применение определенного stash**
Чтобы применить конкретный stash:
```bash
git stash apply stash@{1}
```

### **2. Удаление stash**
Чтобы удалить конкретный stash:
```bash
git stash drop stash@{1}
```

Чтобы удалить все stash:
```bash
git stash clear
```

### **3. Создание ветки из stash**
Чтобы создать новую ветку из stash:
```bash
git stash branch <branch-name> stash@{0}
```

Пример:
```bash
git stash branch new-feature stash@{0}
```

### **4. Автоматическое применение stash при pull**
Если у вас есть несохраненные изменения при выполнении `git pull`, Git предложит сохранить их:
```bash
git pull --autostash
```

---

## **6. Решение проблем**

### **1. Конфликты при применении stash**
Если возникают конфликты при применении stash:
1. Разрешите конфликты вручную.
2. Добавьте исправленные файлы:
   ```bash
   git add <file>
   ```
3. Завершите процесс:
   ```bash
   git stash drop
   ```

### **2. Потеря данных при очистке stash**
Если вы случайно выполнили `git stash clear` и потеряли данные:
1. Проверьте журнал `reflog`:
   ```bash
   git reflog
   ```
2. Восстановите потерянные изменения:
   ```bash
   git reset --hard <commit-hash>
   ```

---

## **7. Заключение**

`git stash` — это удобный инструмент для временного сохранения изменений, позволяющий быстро переключаться между задачами. Она помогает поддерживать чистоту рабочей директории и избегать создания временных коммитов. Правильное использование этой команды делает работу с Git более эффективной и организованной.


# Лекция: Git Remote — подробное руководство

Команда `git remote` используется для управления удаленными репозиториями в Git. Она позволяет добавлять, удалять и просматривать ссылки на удаленные репозитории, с которыми вы работаете. В этой лекции мы рассмотрим все аспекты работы с `git remote`, включая основные концепции, примеры использования, нюансы и продвинутые сценарии.

---

## **1. Основные концепции**

### **Что такое Git Remote?**
Удаленный репозиторий (remote) — это копия вашего проекта, хранящаяся на внешнем сервере (например, GitHub, GitLab или Bitbucket). Команда `git remote` позволяет:
1. **Добавлять** ссылки на удаленные репозитории.
2. **Просматривать** список удаленных репозиториев.
3. **Изменять** URL существующих удаленных репозиториев.
4. **Удалять** ссылки на удаленные репозитории.

По умолчанию при клонировании репозитория Git создает ссылку на удаленный репозиторий с именем `origin`.

---

## **2. Как работает `git remote`**

### **1. Базовый синтаксис**
```bash
git remote <subcommand> [options]
```
Основные подкоманды:
- `add`: Добавляет новый удаленный репозиторий.
- `remove` (или `rm`): Удаляет ссылку на удаленный репозиторий.
- `rename`: Переименовывает существующую ссылку.
- `set-url`: Изменяет URL удаленного репозитория.
- `show`: Показывает информацию о удаленном репозитории.
- `get-url`: Получает URL удаленного репозитория.

---

## **3. Примеры использования**

### **1. Просмотр списка удаленных репозиториев**
Чтобы увидеть список всех удаленных репозиториев:
```bash
git remote
```

Пример вывода:
```plaintext
origin
upstream
```

Чтобы увидеть более подробную информацию:
```bash
git remote -v
```

Пример вывода:
```plaintext
origin  https://github.com/username/repo.git (fetch)
origin  https://github.com/username/repo.git (push)
upstream  https://github.com/organization/repo.git (fetch)
upstream  https://github.com/organization/repo.git (push)
```

### **2. Добавление нового удаленного репозитория**
Чтобы добавить новый удаленный репозиторий:
```bash
git remote add <name> <url>
```

Пример:
```bash
git remote add origin https://github.com/username/repo.git
```

### **3. Удаление удаленного репозитория**
Чтобы удалить ссылку на удаленный репозиторий:
```bash
git remote remove <name>
```

Пример:
```bash
git remote remove upstream
```

### **4. Переименование удаленного репозитория**
Чтобы переименовать существующую ссылку:
```bash
git remote rename <old-name> <new-name>
```

Пример:
```bash
git remote rename origin github-repo
```

### **5. Изменение URL удаленного репозитория**
Чтобы изменить URL существующего удаленного репозитория:
```bash
git remote set-url <name> <new-url>
```

Пример:
```bash
git remote set-url origin git@github.com:username/repo.git
```

---

## **4. Нюансы и особенности**

### **1. Разница между fetch и push URL**
Удаленный репозиторий может иметь разные URL для операций `fetch` и `push`. Чтобы задать их отдельно:
```bash
git remote set-url --push <name> <push-url>
```

Пример:
```bash
git remote set-url --push origin https://github.com/username/repo-push.git
```

### **2. Работа с несколькими удаленными репозиториями**
Если вы работаете с несколькими удаленными репозиториями (например, `origin` и `upstream`), вы можете выполнять команды для каждого из них:
```bash
git fetch origin
git fetch upstream
```

### **3. Проверка текущего состояния**
Чтобы увидеть детальную информацию о удаленном репозитории:
```bash
git remote show <name>
```

Пример:
```bash
git remote show origin
```

Пример вывода:
```plaintext
* remote origin
  Fetch URL: https://github.com/username/repo.git
  Push  URL: https://github.com/username/repo.git
  HEAD branch: main
  Remote branches:
    main tracked
    feature-branch tracked
  Local branches configured for 'git pull':
    main merges with remote main
  Local refs configured for 'git push':
    main pushes to main (up to date)
```

---

## **5. Продвинутые сценарии**

### **1. Добавление нескольких удаленных репозиториев**
Если вы хотите работать с несколькими удаленными репозиториями, добавьте их с разными именами:
```bash
git remote add origin https://github.com/username/repo.git
git remote add upstream https://github.com/organization/repo.git
```

### **2. Использование SSH вместо HTTPS**
Для безопасного подключения используйте SSH:
```bash
git remote set-url origin git@github.com:username/repo.git
```

### **3. Синхронизация с upstream**
Если вы клонировали fork репозитория, добавьте ссылку на оригинальный репозиторий (`upstream`) для синхронизации:
```bash
git remote add upstream https://github.com/organization/repo.git
```

Затем получите изменения:
```bash
git fetch upstream
git merge upstream/main
```

### **4. Автоматическое обновление URL**
Если URL удаленного репозитория изменился, вы можете быстро обновить его:
```bash
git remote set-url origin https://new-url.com/repo.git
```

---

## **6. Решение проблем**

### **1. Ошибка аутентификации**
Если Git запрашивает пароль или токен:
1. Убедитесь, что вы авторизованы на платформе (например, GitHub).
2. Для HTTPS используйте токен вместо пароля.
3. Для SSH настройте SSH-ключи.

### **2. Конфликты при push**
Если возникают конфликты при отправке изменений:
1. Получите последние изменения:
   ```bash
   git pull
   ```
2. Разрешите конфликты, если они есть.
3. Отправьте изменения:
   ```bash
   git push
   ```

---

## **7. Заключение**

`git remote` — это ключевой инструмент для управления удаленными репозиториями в Git. Она позволяет эффективно взаимодействовать с внешними серверами, делиться изменениями и поддерживать актуальность проекта. # Лекция: Git Tag — подробное руководство

Команда `git tag` используется для создания и управления тегами в Git. Теги позволяют помечать определенные точки в истории проекта (например, релизы или важные версии). Они особенно полезны для маркировки стабильных версий программного обеспечения. В этой лекции мы рассмотрим все аспекты работы с `git tag`, включая основные концепции, примеры использования, нюансы и продвинутые сценарии.

---

### **Что такое Git Tag?**
Тег (tag) в Git — это ссылка на конкретный коммит. Он используется для:
1. **Маркировки важных моментов** в истории проекта (например, релизов).
2. **Идентификации версий** программного обеспечения.
3. Упрощения доступа к ключевым состояниям кода.

### **Типы тегов**
1. **Легковесные теги**: Простая ссылка на коммит.
2. **Аннотированные теги**: Содержат дополнительную информацию (сообщение, автор, дата).

---

## **2. Как работает `git tag`**

### **1. Базовый синтаксис**
```bash
git tag <tag-name>
```
Эта команда создает легковесный тег на текущем коммите.

Пример:
```bash
git tag v1.0.0
```

### **2. Создание аннотированного тега**
Чтобы создать аннотированный тег:
```bash
git tag -a <tag-name> -m "Message describing the tag"
```

Пример:
```bash
git tag -a v1.0.0 -m "Release version 1.0.0"
```

### **3. Просмотр списка тегов**
Чтобы увидеть список всех тегов:
```bash
git tag
```

Пример вывода:
```plaintext
v1.0.0
v1.1.0
v2.0.0
```

---

## **3. Примеры использования**

### **1. Создание тега на определенном коммите**
Если вы хотите создать тег на конкретном коммите:
```bash
git tag <tag-name> <commit-hash>
```

Пример:
```bash
git tag v1.0.1 abc1234
```

### **2. Просмотр информации о теге**
Чтобы увидеть информацию об аннотированном теге:
```bash
git show <tag-name>
```

Пример:
```bash
git show v1.0.0
```

### **3. Отправка тегов в удаленный репозиторий**

Пример:

Чтобы отправить все теги:
```bash
git push --tags
```

---

## **4. Нюансы и особенности**

### **1. Разница между легковесными и аннотированными тегами**
- **Легковесные теги**: Простые указатели на коммит. Не содержат дополнительной информации.
- **Аннотированные теги**: Содержат метаданные (сообщение, автор, дата). Рекомендуются для релизов.

### **2. Удаление тегов**
Чтобы удалить тег:
```bash
git tag -d <tag-name>
```

Пример:

## **5. Продвинутые сценарии**

### **2. Создание тега из другой ветки**

2. Создайте новый тег:
   ```bash
   git tag <tag-name> <commit-hash>
   ```

---

## **7. Заключение**

`git tag` — это мощный инструмент для маркировки важных моментов в истории проекта. Она позволяет эффективно организовывать релизы и поддерживать чистую историю разработки. Правильное использование этой команды помогает упростить процесс управления версиями программного обеспечения.


# Лекция: Git Rebase -i (Interactive Rebase) — подробное руководство

Команда `git rebase -i` (интерактивный rebase) является одной из самых мощных и гибких команд в Git. Она позволяет редактировать историю коммитов, объединять, переупорядочивать, изменять или удалять коммиты перед их применением. В этой лекции мы рассмотрим все аспекты работы с `git rebase -i`, включая основные концепции, примеры использования, нюансы и продвинутые сценарии.

---

## **1. Основные концепции**

### **Что такое Interactive Rebase?**
Интерактивный rebase (`git rebase -i`) позволяет вам:
1. **Редактировать историю**: Изменять сообщения коммитов.
2. **Объединять коммиты**: Объединять несколько коммитов в один.
3. **Удалять коммиты**: Убирать ненужные коммиты из истории.
4. **Переупорядочивать коммиты**: Менять порядок коммитов.
5. **Разделять коммиты**: Разбивать один коммит на несколько.

Эта команда особенно полезна для создания чистой и понятной истории разработки.

---

## **2. Как работает `git rebase -i`**

### **1. Базовый синтаксис**
```bash
git rebase -i <base-commit>
```
- `<base-commit>` — хеш коммита, начиная с которого вы хотите отредактировать историю.

Пример:
```bash
git rebase -i HEAD~3
```

### **2. Процесс выполнения**
1. Git открывает текстовый редактор с перечнем коммитов.
2. Вы редактируете список коммитов, указывая действия для каждого коммита.
3. Сохраняете изменения и закрываете редактор.
4. Git выполняет указанные действия.

---

## **3. Примеры использования**

### **1. Редактирование сообщений коммитов**
Если вы хотите изменить сообщение последнего коммита:

#### Шаг 1: Запустите интерактивный rebase:
```bash
git rebase -i HEAD~1
```

#### Шаг 2: В текстовом редакторе замените `pick` на `reword`:
```plaintext
pick abc1234 Commit message 1
```

Замените на:
```plaintext
reword abc1234 Updated commit message
```

#### Шаг 3: Сохраните изменения и укажите новое сообщение.

---

### **2. Объединение коммитов (Squash)**
Если вы хотите объединить последние 3 коммита в один:

#### Шаг 1: Запустите интерактивный rebase:
```bash
git rebase -i HEAD~3
```

#### Шаг 2: В текстовом редакторе замените `pick` на `squash` для коммитов, которые нужно объединить:
```plaintext
pick abc1234 Commit message 1
squash def5678 Commit message 2
squash ghi9012 Commit message 3
```

#### Шаг 3: Укажите новое сообщение для объединенного коммита.

---

### **3. Удаление коммитов**
Если вы хотите удалить коммит из истории:

#### Шаг 1: Запустите интерактивный rebase:
```bash
git rebase -i HEAD~3
```

#### Шаг 2: В текстовом редакторе удалите строку с ненужным коммитом:
```plaintext
pick abc1234 Commit message 1
pick def5678 Commit message 2
```

Здесь коммит `ghi9012` будет удален.

---

### **4. Переупорядочивание коммитов**
Если вы хотите изменить порядок коммитов:

#### Шаг 1: Запустите интерактивный rebase:
```bash
git rebase -i HEAD~3
```

#### Шаг 2: Переместите строки в текстовом редакторе:
```plaintext
pick def5678 Commit message 2
pick abc1234 Commit message 1
pick ghi9012 Commit message 3
```

---

### **5. Разделение коммитов**
Если вы хотите разделить один коммит на несколько:

#### Шаг 1: Запустите интерактивный rebase:
```bash
git rebase -i HEAD~1
```

#### Шаг 2: В текстовом редакторе замените `pick` на `edit`:
```plaintext
edit abc1234 Commit message 1
```

#### Шаг 3: Отмените изменения в рабочей директории:
```bash
git reset HEAD~
```

#### Шаг 4: Создайте новые коммиты:
```bash
git add <file>
git commit -m "First part of the split"
git add <file>
git commit -m "Second part of the split"
```

#### Шаг 5: Продолжите rebase:
```bash
git rebase --continue
```

---

## **4. Нюансы и особенности**

### **1. Потеря данных при редактировании истории**
Использование `git rebase -i` может привести к потере данных, если вы случайно удалите или измените важные коммиты. Для восстановления используйте:
```bash
git reflog
```

### **2. Конфликты при rebase**
Если возникают конфликты:
1. Разрешите конфликты вручную.
2. Добавьте исправленные файлы:
   ```bash
   git add <file>
   ```
3. Продолжите rebase:
   ```bash
   git rebase --continue
   ```

### **3. Отмена rebase**
Если rebase вызвал проблемы, его можно отменить:
```bash
git rebase --abort
```

---

## **5. Продвинутые сценарии**

### **1. Rebase с использованием стратегий**
Git предоставляет несколько стратегий для разрешения конфликтов:
- `ours`: Игнорирует изменения из другой ветки.
- `theirs`: Игнорирует текущие изменения.

Пример:
```bash
git rebase -X theirs main
```

### **2. Автоматическое разрешение конфликтов**
Для автоматического разрешения конфликтов используйте флаг `--strategy-option`:
```bash
git rebase --strategy-option=theirs main
```

---

## **6. Решение проблем**

### **1. Потеря данных при редактировании истории**
Если вы случайно удалили важный коммит:
1. Проверьте журнал `reflog`:
   ```bash
   git reflog
   ```
2. Восстановите потерянные коммиты:
   ```bash
   git reset --hard <commit-hash>
   ```

### **2. Конфликты при rebase**
Если возникают конфликты:
1. Разрешите конфликты вручную.
2. Добавьте исправленные файлы:
   ```bash
   git add <file>
   ```
3. Продолжите rebase:
   ```bash
   git rebase --continue
   ```

---

## **7. Заключение**

`git rebase -i` — это мощный инструмент для управления историей коммитов. Она позволяет создавать чистую и понятную историю разработки, что особенно важно для командной работы. Однако её использование требует осторожности, так как она может привести к потере данных, если применять её неправильно.


# Лекция: Git Cherry-Pick `<commit-hash>` — подробное руководство

Команда `git cherry-pick` является мощным инструментом в Git, который позволяет применить изменения из одного или нескольких коммитов к текущей ветке. Она особенно полезна, когда нужно перенести конкретные изменения из одной ветки в другую, не сливая всю ветку целиком. В этой лекции мы рассмотрим все аспекты работы с `git cherry-pick`, включая основные концепции, примеры использования, нюансы и продвинутые сценарии.

---

## **1. Основные концепции**

### **Что такое Git Cherry-Pick?**
Команда `git cherry-pick` применяет изменения из указанного коммита (или коммитов) к текущей ветке. Она:
1. Создает новый коммит с теми же изменениями.
2. Не изменяет историю исходной ветки.

### **Основные преимущества**
- **Гибкость**: Позволяет переносить конкретные изменения, а не всю ветку.
- **Избирательность**: Вы можете выбрать только те коммиты, которые вам нужны.
- **Независимость**: Не требует слияния или перебазирования всей ветки.

---

## **2. Как работает `git cherry-pick`**

### **1. Базовый синтаксис**
```bash
git cherry-pick <commit-hash>
```
- `<commit-hash>` — хеш коммита, который вы хотите применить.

Пример:
```bash
git cherry-pick abc1234
```

### **2. Процесс выполнения**
1. Git анализирует указанный коммит.
2. Применяет изменения из этого коммита к текущей ветке.
3. Создает новый коммит с теми же изменениями.

---

## **3. Примеры использования**

### **1. Применение одного коммита**
Если вы хотите применить изменения из одного коммита:
```bash
git cherry-pick <commit-hash>
```

#### Исходное состояние:
```
main: A <- B <- C
feature-branch: A <- D <- E
```

#### Выполнение:
```bash
git checkout main
git cherry-pick abc1234
```

#### Результат:
```
main: A <- B <- C <- D'
```
Здесь `D'` — это новый коммит, который содержит изменения из `D`.

---

### **2. Применение нескольких коммитов**
Если вы хотите применить несколько коммитов:
```bash
git cherry-pick <commit-hash-1> <commit-hash-2>
```

Пример:
```bash
git cherry-pick abc1234 def5678
```

---

### **3. Применение диапазона коммитов**
Если вы хотите применить диапазон коммитов:
```bash
git cherry-pick <start-commit>..<end-commit>
```

Пример:
```bash
git cherry-pick abc1234..def5678
```

Это применит все коммиты между `abc1234` и `def5678`, включая последний.

---

## **4. Нюансы и особенности**

### **1. Конфликты при cherry-pick**
Если возникают конфликты:
1. Разрешите конфликты вручную.
2. Добавьте исправленные файлы:
   ```bash
   git add <file>
   ```
3. Продолжите cherry-pick:
   ```bash
   git cherry-pick --continue
   ```

### **2. Отмена cherry-pick**
Если cherry-pick вызвал проблемы:
```bash
git cherry-pick --abort
```

### **3. Пропуск коммитов**
Если вы хотите пропустить коммит, вызвавший конфликт:
```bash
git cherry-pick --skip
```

---

## **5. Продвинутые сценарии**

### **1. Применение коммитов с опциями**
Вы можете использовать различные опции для настройки поведения `git cherry-pick`:
- `-n`: Применяет изменения, но не создает новый коммит.
  ```bash
  git cherry-pick -n <commit-hash>
  ```
- `-x`: Добавляет информацию о оригинальном коммите в сообщение нового коммита.
  ```bash
  git cherry-pick -x <commit-hash>
  ```

### **2. Применение коммитов из другой ветки**
Если вы хотите применить коммиты из другой ветки:
```bash
git cherry-pick <branch-name>~<number>
```

Пример:
```bash
git cherry-pick feature-branch~2
```

---

### **3. Автоматическое разрешение конфликтов**
Для автоматического разрешения конфликтов используйте стратегии слияния:
```bash
git cherry-pick -X theirs <commit-hash>
```

---

## **6. Решение проблем**

### **1. Потеря данных при cherry-pick**
Если вы случайно применили неправильный коммит:
1. Удалите последний коммит:
   ```bash
   git reset --hard HEAD~1
   ```

### **2. Конфликты при cherry-pick**
Если возникают конфликты:
1. Разрешите конфликты вручную.
2. Добавьте исправленные файлы:
   ```bash
   git add <file>
   ```
3. Продолжите cherry-pick:
   ```bash
   git cherry-pick --continue
   ```

---

## **7. Заключение**

`git cherry-pick` — это мощный инструмент для применения конкретных изменений из одной ветки в другую. Она позволяет гибко управлять историей коммитов и избирательно переносить изменения. Однако её использование требует осторожности, так как она может привести к конфликтам или дублированию изменений.


# Лекция: Git Blame `<file>` — подробное руководство

Команда `git blame` является мощным инструментом в Git, который позволяет определить, кто и когда внёс изменения в конкретные строки файла. Она особенно полезна для анализа истории изменений, поиска авторов ошибок или понимания контекста кода. В этой лекции мы рассмотрим все аспекты работы с `git blame`, включая основные концепции, примеры использования, нюансы и продвинутые сценарии.

---

## **1. Основные концепции**

### **Что такое Git Blame?**
Команда `git blame` показывает, кто изменил каждую строку файла, а также предоставляет следующую информацию:
1. **Хеш коммита**: Уникальный идентификатор коммита, в котором была изменена строка.
2. **Автор**: Имя и email человека, который внёс изменения.
3. **Дата**: Дата и время последнего изменения строки.
4. **Номер строки**: Порядковый номер строки в файле.
5. **Сама строка**: Содержимое строки.

Эта команда не изменяет репозиторий, а только отображает историю изменений.

---

## **2. Как работает `git blame`**

### **1. Базовый синтаксис**
```bash
git blame <file>
```
- `<file>` — имя файла, для которого вы хотите просмотреть историю изменений.

Пример:
```bash
git blame README.md
```

### **2. Пример вывода**
Пример вывода команды:
```plaintext
^abc1234 (John Doe 2023-10-01 10:00:00 +0300 1) # Project Title
def5678  (Jane Smith 2023-10-02 12:30:00 +0300 2)
ghi9012  (John Doe 2023-10-03 15:45:00 +0300 3) Initial content
```

Здесь:
- `^abc1234`: Хеш первого коммита, где была добавлена строка.
- `def5678`: Хеш последнего коммита, где была изменена строка.
- `(John Doe ...)`: Автор, дата и время изменения.
- `1`, `2`, `3`: Номера строк в файле.

---

## **3. Примеры использования**

### **1. Просмотр истории изменений файла**
Чтобы увидеть, кто и когда изменил каждую строку файла:
```bash
git blame README.md
```

### **2. Отображение только хешей коммитов**
Если вам нужны только хеши коммитов:
```bash
git blame -s README.md
```

Пример вывода:
```plaintext
^abc1234 1) # Project Title
def5678  2)
ghi9012  3) Initial content
```

### **3. Отображение изменений за определенный период**
Чтобы увидеть изменения, сделанные за последние 7 дней:
```bash
git blame --since=7.days README.md
```

---

### **4. Отображение изменений конкретного автора**
Если вы хотите увидеть только изменения, сделанные конкретным автором:
```bash
git blame --author="John Doe" README.md
```

---

### **5. Отображение изменений в другом формате**
Чтобы настроить формат вывода, используйте флаг `-e` (email вместо имени):
```bash
git blame -e README.md
```

Пример вывода:

# Лекция: Git Bisect — подробное руководство

Команда `git bisect` является мощным инструментом в Git, который помогает находить конкретный коммит, в котором была внесена ошибка или изменение. Она использует метод **двоичного поиска** (binary search), чтобы эффективно сузить круг подозрительных коммитов. В этой лекции мы рассмотрим все аспекты работы с `git bisect`, включая основные концепции, примеры использования, нюансы и продвинутые сценарии.

---

## **1. Основные концепции**

### **Что такое Git Bisect?**
Команда `git bisect` автоматизирует процесс поиска "вредоносного" коммита (bad commit), который вызвал ошибку или нежелательное поведение. Она работает следующим образом:
1. Вы указываете "хороший" коммит (good commit), где проблема отсутствовала.
2. Указываете "плохой" коммит (bad commit), где проблема проявляется.
3. Git автоматически переходит к середине истории между этими двумя коммитами.
4. Вы проверяете состояние кода и сообщаете Git, является ли текущий коммит "хорошим" или "плохим".
5. Процесс повторяется до тех пор, пока не будет найден конкретный коммит, в котором появилась проблема.

---

## **2. Как работает `git bisect`**

### **1. Базовый синтаксис**
```bash
git bisect start
git bisect bad
git bisect good <commit-hash>
```

### **2. Процесс выполнения**
1. Запустите режим бинарного поиска:
   ```bash
   git bisect start
   ```
2. Укажите текущий коммит как "плохой":
   ```bash
   git bisect bad
   ```
3. Укажите "хороший" коммит:
   ```bash
   git bisect good <commit-hash>
   ```
4. Git автоматически переместится к середине истории. Проверьте состояние кода.
5. Сообщите Git, является ли текущий коммит "хорошим" или "плохим":
   ```bash
   git bisect good
   ```
   или
   ```bash
   git bisect bad
   ```
6. Повторяйте шаги 4–5, пока Git не найдет проблемный коммит.

7. После завершения поиска вернитесь в обычный режим:
   ```bash
   git bisect reset
   ```

---

## **3. Примеры использования**

### **1. Поиск "вредоносного" коммита**
Предположим, вы знаете, что в последнем коммите есть ошибка, но она отсутствовала в более раннем коммите `abc1234`.

#### Шаг 1: Запустите режим бинарного поиска:
```bash
git bisect start
```

#### Шаг 2: Укажите текущий коммит как "плохой":
```bash
git bisect bad
```

#### Шаг 3: Укажите "хороший" коммит:
```bash
git bisect good abc1234
```

#### Шаг 4: Проверьте состояние кода:
Git переместится к середине истории. Проверьте, проявляется ли ошибка.

#### Шаг 5: Сообщите Git результат:
Если ошибка присутствует:
```bash
git bisect bad
```
Если ошибки нет:
```bash
git bisect good
```

#### Шаг 6: Повторяйте шаги 4–5, пока Git не найдет проблемный коммит.

#### Шаг 7: Вернитесь в обычный режим:
```bash
git bisect reset
```

---

### **2. Автоматизация проверок**
Если у вас есть скрипт для автоматической проверки состояния кода, вы можете использовать его с `git bisect run`:

#### Шаг 1: Создайте скрипт проверки:
Создайте файл `test.sh`:
```bash
#!/bin/bash
# Скрипт проверяет, проявляется ли ошибка
if ./run-tests.sh; then
  exit 0  # Тесты успешны ("хороший" коммит)
else
  exit 1  # Тесты провалены ("плохой" коммит)
fi
```

#### Шаг 2: Запустите автоматический поиск:
```bash
git bisect start
git bisect bad
git bisect good abc1234
git bisect run ./test.sh
```

Git автоматически выполнит бинарный поиск, используя ваш скрипт.

---

## **4. Нюансы и особенности**

### **1. Отмена процесса**
Если вы хотите прервать процесс бинарного поиска:
```bash
git bisect reset
```

### **2. Использование терминов "good" и "bad"**
- "Good" означает, что проблема отсутствует.
- "Bad" означает, что проблема присутствует.

Вы можете использовать другие термины, если это удобнее:
```bash
git bisect start
git bisect old
git bisect new
```

### **3. Работа с большими репозиториями**
Если история содержит тысячи коммитов, `git bisect` значительно ускоряет поиск проблемы по сравнению с ручным просмотром.

---

## **5. Продвинутые сценарии**

### **1. Поиск изменений в определенном файле**
Если вы хотите сосредоточиться на изменениях в конкретном файле:
```bash
git bisect start
git bisect bad
git bisect good abc1234
git checkout <file>
```

### **2. Поиск изменений в диапазоне коммитов**
Если вы знаете диапазон коммитов, где может быть проблема:
```bash
git bisect start HEAD abc1234
```

---

### **3. Сохранение состояния поиска**
Если вы хотите временно покинуть режим `git bisect`, но сохранить текущее состояние:
```bash
git bisect log > bisect.log
```

Чтобы восстановить состояние:
```bash
git bisect replay bisect.log
```

---

## **6. Решение проблем**

### **1. Конфликты при проверке состояния**
Если возникают конфликты при проверке состояния:
1. Разрешите конфликты вручную.
2. Продолжите процесс:
   ```bash
   git bisect good
   ```
   или
   ```bash
   git bisect bad
   ```

### **2. Потеря данных при отмене**
Если вы случайно выполнили `git bisect reset` и потеряли состояние:
1. Восстановите лог бинарного поиска:
   ```bash
   git bisect replay bisect.log
   ```

---

## **7. Заключение**

`git bisect` — это мощный инструмент для поиска "вредоносных" коммитов, которые вызывают ошибки или нежелательное поведение. Она автоматизирует процесс бинарного поиска, что делает её особенно полезной для анализа больших историй изменений. Однако её использование требует внимания к деталям, особенно при работе с автоматическими тестами.


# Лекция: Git Merge --abort — подробное руководство

Команда `git merge --abort` используется для отмены процесса слияния (merge), если возникли конфликты или другие проблемы. Она позволяет вернуть репозиторий в состояние, которое было до начала операции слияния. В этой лекции мы рассмотрим все аспекты работы с `git merge --abort`, включая основные концепции, примеры использования, нюансы и продвинутые сценарии.

---

## **1. Основные концепции**

### **Что такое Git Merge --abort?**
Команда `git merge --abort` отменяет текущую операцию слияния и возвращает репозиторий в состояние, предшествующее началу слияния. Она:
1. Убирает все изменения, связанные с неудачным слиянием.
2. Восстанавливает рабочую директорию и staging area до состояния перед выполнением `git merge`.

### **Основные преимущества**
- **Безопасность**: Позволяет избежать ошибок при сложных слияниях.
- **Простота**: Отменяет слияние одной командой.
- **Гибкость**: Может использоваться как на начальных этапах слияния, так и после разрешения конфликтов.

---

## **2. Как работает `git merge --abort`**

### **1. Базовый синтаксис**
```bash
git merge --abort
```

### **2. Процесс выполнения**
1. Git проверяет, находится ли репозиторий в состоянии слияния.
2. Если слияние активно, Git отменяет его и возвращает репозиторий в исходное состояние.
3. Все изменения, связанные с неудачным слиянием, удаляются.

---

## **3. Примеры использования**

### **1. Отмена слияния с конфликтами**
Если вы выполнили `git merge`, но возникли конфликты:
1. Выполните отмену слияния:
   ```bash
   git merge --abort
   ```

#### Исходное состояние:
```
main: A <- B <- C
feature-branch: A <- D <- E
```

#### После выполнения:
```
main: A <- B <- C
```

Рабочая директория и staging area возвращаются к состоянию до слияния.

---

### **2. Отмена слияния без конфликтов**
Если слияние уже завершено, но вы хотите отменить его:
1. Вернитесь к состоянию перед слиянием:
   ```bash
   git reset --hard HEAD~1
   ```

**Важно:** `git merge --abort` работает только в процессе слияния. Если слияние завершено, используйте `git reset`.

---

### **3. Отмена слияния после разрешения конфликтов**
Если вы начали разрешать конфликты, но решили отменить слияние:
1. Выполните:
   ```bash
   git merge --abort
   ```

---

## **4. Нюансы и особенности**

### **1. Когда можно использовать `git merge --abort`**
- **Активное слияние**: Команда работает только в процессе слияния.
- **Состояние MERGE_HEAD**: Git создает файл `.git/MERGE_HEAD` во время слияния. Если этот файл существует, `git merge --abort` доступен.

### **2. Отмена слияния после push**
Если слияние уже отправлено в удаленный репозиторий:
1. Отмените слияние локально:
   ```bash
   git merge --abort
   ```
2. Отправьте изменения обратно:
   ```bash
   git push --force-with-lease
   ```

**Важно:** Используйте `--force-with-lease`, чтобы избежать затирания чужих изменений.

---

### **3. Разница между `git merge --abort` и `git reset`**
- **`git merge --abort`**: Отменяет только слияние, если оно активно.
- **`git reset`**: Отменяет любые изменения, включая слияния.

---

## **5. Продвинутые сценарии**

### **1. Отмена слияния с использованием reflog**
Если вы случайно выполнили `git merge --abort` и потеряли важные изменения:
1. Проверьте журнал `reflog`:
   ```bash
   git reflog
   ```
2. Восстановите потерянные коммиты:
   ```bash
   git reset --hard <commit-hash>
   ```

---

### **2. Отмена слияния с несколькими конфликтами**
Если слияние вызвало множество конфликтов:
1. Отмените слияние:
   ```bash
   git merge --abort
   ```
2. Перебазируйте ветку, чтобы избежать конфликтов:
   ```bash
   git rebase main
   ```

---

### **3. Отмена слияния с сохранением изменений**
Если вы хотите сохранить изменения, связанные с неудачным слиянием:
1. Сохраните изменения в stash:
   ```bash
   git stash
   ```
2. Отмените слияние:
   ```bash
   git merge --abort
   ```
3. Восстановите изменения:
   ```bash
   git stash pop
   ```

---

## **6. Решение проблем**

### **1. Ошибка "fatal: There is no merge to abort"**
Если вы получили сообщение:
```plaintext
fatal: There is no merge to abort
```
Это означает, что слияние уже завершено или не было начато. В этом случае используйте:
```bash
git reset --hard HEAD~1
```

---

### **2. Потеря данных при отмене слияния**
Если вы случайно выполнили `git merge --abort` и потеряли важные изменения:
1. Проверьте журнал `reflog`:
   ```bash
   git reflog
   ```
2. Восстановите потерянные коммиты:
   ```bash
   git reset --hard <commit-hash>
   ```

---

## **7. Заключение**

`git merge --abort` — это простой и безопасный способ отменить слияние, если возникли конфликты или другие проблемы. Она помогает избежать ошибок и поддерживать чистую историю разработки. Однако её использование требует осторожности, особенно при работе с публичными репозиториями.


# Лекция: Git Mergetool — подробное руководство

Команда `git mergetool` используется для разрешения конфликтов слияния (merge conflicts) с помощью графических инструментов или специализированных программ. Она особенно полезна, когда конфликты сложны и их трудно решить вручную через текстовый редактор. В этой лекции мы рассмотрим все аспекты работы с `git mergetool`, включая основные концепции, примеры использования, нюансы и продвинутые сценарии.

---

## **1. Основные концепции**

### **Что такое Git Mergetool?**
Команда `git mergetool` запускает внешний инструмент для разрешения конфликтов слияния. Она:
1. Открывает выбранный инструмент для сравнения и редактирования конфликтующих файлов.
2. Позволяет визуализировать изменения и выбрать, какие строки сохранить.
3. Автоматически добавляет исправленные файлы в staging area после разрешения конфликтов.

### **Основные преимущества**
- **Визуализация**: Графические инструменты упрощают понимание конфликтов.
- **Автоматизация**: Инструмент автоматически обрабатывает файлы после разрешения конфликтов.
- **Гибкость**: Поддерживает множество популярных инструментов (например, VS Code, KDiff3, P4Merge).

---

## **2. Как работает `git mergetool`**

### **1. Базовый синтаксис**
```bash
git mergetool
```

### **2. Процесс выполнения**
1. Git проверяет, есть ли конфликты слияния.
2. Запускает выбранный инструмент для каждого конфликтующего файла.
3. После разрешения конфликтов добавляет исправленные файлы в staging area.

---

## **3. Примеры использования**

### **1. Разрешение конфликтов с помощью mergetool**
Если вы выполнили `git merge` и возникли конфликты:
1. Запустите `git mergetool`:
   ```bash
   git mergetool
   ```

#### Исходное состояние:
```
<<<<<<< HEAD
console.log("Main branch");
=======
console.log("Feature branch");
>>>>>>> feature-branch
```

#### После выполнения:
Инструмент откроет графический интерфейс, где вы сможете выбрать нужные строки.

---

### **2. Настройка mergetool**
Если вы хотите использовать определенный инструмент (например, VS Code):
1. Настройте его:
   ```bash
   git config --global merge.tool vscode
   ```
2. Убедитесь, что VS Code установлен и доступен в системе.

---

### **3. Использование с конкретным файлом**
Если вы хотите разрешить конфликт только для одного файла:
```bash
git mergetool <file>
```

Пример:
```bash
git mergetool README.md
```

---

## **4. Нюансы и особенности**

### **1. Поддерживаемые инструменты**
Git поддерживает множество инструментов для разрешения конфликтов:
- **VS Code**: Легкий и удобный редактор.
- **KDiff3**: Мощный инструмент для сравнения файлов.
- **P4Merge**: Графический инструмент для слияния.
- **Meld**: Популярный инструмент с открытым исходным кодом.

Чтобы увидеть список доступных инструментов:
```bash
git mergetool --tool-help
```

---

### **2. Настройка инструмента по умолчанию**
Чтобы установить инструмент по умолчанию:
```bash
git config --global merge.tool <tool-name>
```

Пример:
```bash
git config --global merge.tool meld
```

---

### **3. Работа с backup-файлами**
По умолчанию Git создает backup-файлы с расширением `.orig` после разрешения конфликтов. Чтобы отключить создание backup-файлов:
```bash
git config --global mergetool.keepBackup false
```

---

## **5. Продвинутые сценарии**

### **1. Использование нескольких инструментов**
Если вы хотите использовать разные инструменты для разных задач:
1. Настройте несколько инструментов:
   ```bash
   git config --global merge.tool vscode
   git config --global diff.tool meld
   ```

---

### **2. Автоматическое разрешение конфликтов**
Если вы хотите автоматически разрешить простые конфликты:
1. Используйте флаг `-X`:
   ```bash
   git merge -X theirs <branch>
   ```
2. Для сложных конфликтов используйте `git mergetool`.

---

### **3. Интеграция с IDE**
Многие современные IDE (например, IntelliJ IDEA, Visual Studio) поддерживают встроенные инструменты для разрешения конфликтов. Вы можете настроить их как mergetool:
```bash
git config --global merge.tool vscode
```

---

## **6. Решение проблем**

### **1. Ошибка "No merge tool found"**
Если вы получили сообщение:
```plaintext
No merge tool found
```
Это означает, что выбранный инструмент не установлен или не настроен. Установите его и настройте:
```bash
git config --global merge.tool vscode
```

---

### **2. Конфликты с несколькими файлами**
Если возникают конфликты с несколькими файлами:
1. Запустите `git mergetool`:
   ```bash
   git mergetool
   ```
2. Разрешите конфликты для каждого файла.

---

### **3. Отмена изменений**
Если вы случайно сохранили неправильные изменения:
1. Отмените слияние:
   ```bash
   git merge --abort
   ```
2. Повторите процесс.

---

## **7. Заключение**

`git mergetool` — это мощный инструмент для разрешения конфликтов слияния с помощью графических средств. Она упрощает работу с сложными конфликтами и помогает поддерживать чистую историю разработки. Однако её использование требует настройки и выбора подходящего инструмента.


# Лекция: Git Clean — подробное руководство

Команда `git clean` используется для удаления неотслеживаемых файлов и директорий из рабочей директории. Она особенно полезна, когда нужно очистить проект от временных файлов, артефактов сборки или других ненужных данных. В этой лекции мы рассмотрим все аспекты работы с `git clean`, включая основные концепции, примеры использования, нюансы и продвинутые сценарии.

---

## **1. Основные концепции**

### **Что такое Git Clean?**
Команда `git clean` удаляет файлы и директории, которые:
1. Не находятся под версионным контролем (не отслеживаются Git).
2. Не игнорируются через `.gitignore`.

### **Основные преимущества**
- **Очистка рабочей директории**: Удаляет временные файлы, артефакты сборки и другие ненужные данные.
- **Безопасность**: Предоставляет флаги для предварительного просмотра и безопасного удаления.
- **Гибкость**: Позволяет настраивать поведение через различные параметры.

---

## **2. Как работает `git clean`**

### **1. Базовый синтаксис**
```bash
git clean [options]
```

### **2. Процесс выполнения**
1. Git проверяет, какие файлы и директории не отслеживаются.
2. Удаляет их в зависимости от указанных параметров.

---

## **3. Примеры использования**

### **1. Просмотр неотслеживаемых файлов**
Чтобы увидеть, какие файлы будут удалены:
```bash
git clean -n
```
Флаг `-n` (или `--dry-run`) показывает, что будет удалено, но не выполняет удаление.

Пример вывода:
```plaintext
Would remove untracked-file.txt
Would remove build/
```

---

### **2. Удаление неотслеживаемых файлов**
Чтобы удалить неотслеживаемые файлы:
```bash
git clean -f
```
Флаг `-f` (или `--force`) принудительно удаляет файлы.

---

### **3. Удаление неотслеживаемых директорий**
Чтобы удалить неотслеживаемые директории:
```bash
git clean -fd
```
Флаг `-d` добавляет удаление директорий.

---

### **4. Удаление игнорируемых файлов**
Чтобы удалить файлы, игнорируемые через `.gitignore`:
```bash
git clean -fx
```
Флаг `-x` добавляет удаление игнорируемых файлов.

---

## **4. Нюансы и особенности**

### **1. Безопасное использование**
- **Предварительный просмотр**: Всегда используйте `-n`, чтобы увидеть, что будет удалено.
- **Подтверждение**: Используйте `-i` для интерактивного режима, где Git запрашивает подтверждение для каждого файла:
  ```bash
  git clean -i
  ```

---

### **2. Разница между `-f` и `-fd`**
- **`-f`**: Удаляет только файлы.
- **`-fd`**: Удаляет файлы и директории.

---

### **3. Исключение определенных файлов**
Если вы хотите исключить определенные файлы или директории:
```bash
git clean -f --exclude=<pattern>
```

Пример:
```bash
git clean -f --exclude=*.log
```
Это удалит все неотслеживаемые файлы, кроме `.log`.

---

## **5. Продвинутые сценарии**

### **1. Удаление всех неотслеживаемых файлов и директорий**
Чтобы удалить все неотслеживаемые файлы и директории, включая игнорируемые:
```bash
git clean -fdx
```

---

### **2. Удаление файлов в конкретной директории**
Если вы хотите очистить только определенную директорию:
```bash
git clean -fd <directory>
```

Пример:
```bash
git clean -fd build/
```

---

### **3. Интерактивное удаление**
Чтобы удалить файлы в интерактивном режиме:
```bash
git clean -i
```

Git предложит выбрать, какие файлы удалить:
```plaintext
Would remove the following items:
  untracked-file.txt    build/
*** Commands ***
    1: clean                2: filter by pattern    3: select by numbers    4: ask each             5: quit
    6: help
What now>
```

---

## **6. Решение проблем**

### **1. Ошибка "fatal: clean.requireForce defaults to true"**
Если вы получили сообщение:
```plaintext
fatal: clean.requireForce defaults to true and neither -i, -n, nor -f given; refusing to clean
```
Это означает, что Git требует явного подтверждения. Добавьте флаг `-f`:
```bash
git clean -f
```

---

### **2. Восстановление удаленных файлов**
Если вы случайно удалили важные файлы:
1. Проверьте журнал `reflog`:
   ```bash
   git reflog
   ```
2. Восстановите потерянные файлы из резервной копии или другого источника.

---

### **3. Удаление игнорируемых файлов**
Если вы хотите удалить только игнорируемые файлы:
```bash
git clean -fX
```
Флаг `-X` удаляет только игнорируемые файлы.

---

## **7. Заключение**

`git clean` — это мощный инструмент для очистки рабочей директории от неотслеживаемых файлов и директорий. Она помогает поддерживать порядок в проекте и избегать загромождения ненужными данными. Однако её использование требует осторожности, особенно при работе с важными файлами.


# Лекция: Git Reflog — подробное руководство

Команда `git reflog` является мощным инструментом в Git, который позволяет просматривать историю всех изменений в вашем локальном репозитории. Она записывает каждое действие, выполненное в репозитории (например, коммиты, переключения веток, слияния, сбросы и т.д.), даже если эти изменения не отражены в основной истории коммитов. В этой лекции мы рассмотрим все аспекты работы с `git reflog`, включая основные концепции, примеры использования, нюансы и продвинутые сценарии.

---

## **1. Основные концепции**

### **Что такое Git Reflog?**
Команда `git reflog` показывает журнал ссылок (reference log), который содержит:
1. **Хеши коммитов**: Уникальные идентификаторы состояний репозитория.
2. **Действия**: Информацию о том, какие операции были выполнены (например, `commit`, `checkout`, `reset`).
3. **Временные метки**: Дату и время выполнения каждого действия.

Reflog хранится локально и существует только на вашем компьютере. Это делает его полезным для восстановления потерянных данных или отмены ошибочных действий.

---

## **2. Как работает `git reflog`**

### **1. Базовый синтаксис**
```bash
git reflog
```

### **2. Процесс выполнения**
1. Git проверяет локальный журнал ссылок.
2. Показывает список всех действий, выполненных в репозитории.
3. Каждое действие имеет уникальный хеш и описание.

Пример вывода:
```plaintext
abc1234 HEAD@{0}: commit: Add new feature
def5678 HEAD@{1}: checkout: moving from main to feature-branch
ghi9012 HEAD@{2}: reset: moving to HEAD~1
```

Здесь:
- `abc1234`: Хеш коммита.
- `HEAD@{0}`: Ссылка на состояние HEAD.
- `commit: Add new feature`: Описание действия.

---

## **3. Примеры использования**

### **1. Просмотр журнала ссылок**
Чтобы увидеть историю всех действий:
```bash
git reflog
```

Пример:
```plaintext
abc1234 HEAD@{0}: commit: Add new feature
def5678 HEAD@{1}: checkout: moving from main to feature-branch
ghi9012 HEAD@{2}: reset: moving to HEAD~1
jkl3456 HEAD@{3}: merge feature-branch: Merge branch 'feature-branch'
```

---

### **2. Восстановление потерянного коммита**
Если вы случайно выполнили `git reset --hard` и потеряли коммит:
1. Найдите хеш потерянного коммита в `git reflog`:
   ```bash
   git reflog
   ```
2. Восстановите его:
   ```bash
   git reset --hard <commit-hash>
   ```

Пример:
```bash
git reset --hard abc1234
```

---

### **3. Восстановление удаленной ветки**
Если вы случайно удалили ветку:
1. Найдите последний коммит ветки в `git reflog`:
   ```bash
   git reflog
   ```
2. Создайте новую ветку на основе этого коммита:
   ```bash
   git checkout -b <branch-name> <commit-hash>
   ```

Пример:
```bash
git checkout -b feature-branch def5678
```

---

### **4. Отмена ошибочного действия**
Если вы выполнили ошибочное действие (например, `git reset` или `git rebase`):
1. Найдите предыдущее состояние в `git reflog`.
2. Вернитесь к этому состоянию:
   ```bash
   git reset --hard HEAD@{1}
   ```

---

## **4. Нюансы и особенности**

### **1. Локальная природа reflog**
- **Локальный журнал**: Reflog существует только на вашем компьютере и не отправляется в удаленный репозиторий.
- **Срок хранения**: Записи в reflog автоматически удаляются через 90 дней (по умолчанию).

---

### **2. Разница между `git log` и `git reflog`**
- **`git log`**: Показывает историю коммитов текущей ветки.
- **`git reflog`**: Показывает все действия, выполненные в репозитории, включая те, которые не отражены в истории.

---

### **3. Настройка срока хранения**
Если вы хотите изменить срок хранения записей в reflog:
```bash
git config gc.reflogExpire <time>
```

Пример:
```bash
git config gc.reflogExpire 30.days
```

---

## **5. Продвинутые сценарии**

### **1. Поиск конкретного действия**
Если вы хотите найти конкретное действие:
```bash
git reflog | grep "commit"
```

Пример:
```bash
git reflog | grep "Add new feature"
```

---

### **2. Восстановление после принудительной отправки**
Если вы случайно выполнили `git push --force` и затерли чужие изменения:
1. Найдите предыдущее состояние в `git reflog`.
2. Восстановите его:
   ```bash
   git reset --hard <commit-hash>
   ```
3. Отправьте изменения обратно:
   ```bash
   git push --force-with-lease
   ```

---

### **3. Использование reflog для анализа**
Если вы хотите проанализировать историю изменений:
```bash
git reflog show <branch-name>
```

Пример:
```bash
git reflog show main
```

---

## **6. Решение проблем**

### **1. Потеря данных после очистки**
Если записи в reflog были автоматически очищены:
1. Используйте `git fsck` для поиска "висячих" объектов:
   ```bash
   git fsck --full --no-reflogs --unreachable --lost-found
   ```

---

### **2. Ошибка "fatal: bad object"**
Если вы получили сообщение:
```plaintext
fatal: bad object HEAD@{0}
```
Это означает, что объект был удален из reflog. Восстановите его с помощью `git fsck`.

---

### **3. Конфликты при восстановлении**
Если возникают конфликты при восстановлении:
1. Разрешите конфликты вручную.
2. Добавьте исправленные файлы:
   ```bash
   git add <file>
   ```
3. Завершите процесс:
   ```bash
   git reset --hard
   ```

---

## **7. Заключение**

`git reflog` — это мощный инструмент для восстановления потерянных данных и анализа истории изменений в репозитории. Она помогает избежать потери важной информации и исправить ошибочные действия. Однако её использование требует внимания к деталям, особенно при работе с командами, изменяющими историю.
