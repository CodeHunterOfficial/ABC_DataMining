### **Тема 1: Основы Git и организация командной работы**

* **Введение в Git:** Концепция системы контроля версий, её важность в современной разработке.
* **Базовые команды Git:** `init`, `add`, `commit`, `status`, `log`. Работа с историей изменений.
* **Удаленные репозитории:** Клонирование (`clone`), отправка изменений (`push`), получение изменений (`pull`).
* **Основы групповой разработки:** Ветвление (`branch`), переключение между ветками (`checkout`), слияние (`merge`). Стратегии ветвления (Git Flow, GitHub Flow).
* **Разрешение конфликтов:** Пошаговое руководство по устранению конфликтов при слиянии.
* **Практика:** Студенты создают общий репозиторий в **Azure Repos**, учатся работать с ветками, выполнять pull request'ы и разрешать конфликты в команде.

### **Тема 5: Непрерывная интеграция (CI) с Azure Pipelines**

* **Концепция CI:** Автоматизация сборки, тестирования и проверки качества кода.
* **Настройка CI-пайплайна:** Создание YAML-пайплайна в **Azure Pipelines**.
* **Этапы CI:** Получение кода из **Azure Repos**, восстановление зависимостей, сборка ASP.NET Core-приложения, запуск всех тестов.
* **Публикация артефактов:** Сохранение результатов сборки и тестов для дальнейшего использования.
* **Практика:** Студенты настраивают CI-пайплайн, который автоматически запускает сборку и все тесты после каждого push'а в главную ветку.



### **Тема 6: Контейнеризация приложения с Docker**

* **Введение в Docker:** Что такое контейнеры, образы, Dockerfile. Преимущества контейнеризации.
* **Написание Dockerfile для ASP.NET Core:** Оптимизация Dockerfile для многоступенчатой сборки (multi-stage build).
* **Сборка и запуск контейнера:** Команды `docker build` и `docker run`.
* **Управление образами:** `docker images`, `docker rmi`.
* **Репозитории образов:** Введение в **Docker Hub** и **Azure Container Registry (ACR)**. Публикация образа в ACR.
* **Практика:** Студенты "упаковывают" своё ASP.NET Core-приложение в Docker-контейнер, запускают его локально и публикуют образ в ACR.


### **Тема 7: Непрерывное развертывание (CD) с Azure Pipelines**

* **Концепция CD:** Автоматизация доставки и развертывания приложения.
* **Настройка CD-пайплайна:** Создание YAML-пайплайна для развертывания.
* **Развертывание в Azure App Service:** Настройка развертывания Docker-контейнера в **Azure App Service**.
* **Переменные и окружения:** Управление конфигурацией для разных сред (Dev, Staging, Production).
* **Практика:** Студенты настраивают полный CI/CD-пайплайн, который автоматически собирает, тестирует, создает образ и разворачивает их приложение в Azure при каждом успешном коммите.



### **Тема 8: Логирование, мониторинг и безопасность**

* **Логирование в ASP.NET Core:** Использование ILogger, настройка провайдеров логирования (Console, Debug, Azure Application Insights).
* **Мониторинг с Azure Monitor:** Сбор метрик, создание дашбордов, настройка оповещений.
* **Azure Application Insights:** Отслеживание производительности, ошибок и использования приложения.
* **Управление секретами:** Безопасное хранение чувствительных данных с помощью **Azure Key Vault** и интеграция с ASP.NET Core.
* **Практика:** Интеграция логирования и мониторинга в приложение, настройка Key Vault для хранения строк подключения к базе данных.



### **Тема 9: Введение в MLOps на Azure**

* **Что такое MLOps:** Концепция и отличие от традиционного DevOps.
* **Жизненный цикл ML-модели:** От данных до развертывания.
* **Azure Machine Learning:** Обзор сервиса, создание рабочей области.
* **Развертывание ML-модели:** Базовый пример развертывания обученной модели как REST API в **Azure Container Instance** или **Azure Kubernetes Service**.
* **Интеграция с приложением:** Как ASP.NET Core-приложение может взаимодействовать с развернутой ML-моделью.
* **Практика:** Развертывание простой предварительно обученной ML-модели как веб-сервиса и интеграция её вызовов в ASP.NET Core-приложение.



### **Тема 10: Итоговый проект: Построение готового продукта**

* **Групповая работа:** Студенты делятся на группы и разрабатывают полноценный продукт, интегрируя все изученные технологии.
* **Полный цикл DevOps:** Проект должен пройти весь путь: от планирования в **Azure Boards** и управления кодом в **Azure Repos** до автоматизированного CI/CD-развертывания в Azure.
* **Включение MLOps:** Если применимо, включение ML-компонента, развернутого через MLOps.
* **Презентация и защита:** Каждая группа демонстрирует свой готовый продукт, объясняет архитектуру, процесс разработки и развертывания.
