# Учебная программа: Полный курс по DevOps и разработке на ASP.NET Core с Azure

## Тема 1: Основы Git и организация командной работы
- **Введение в Git**: Что такое система контроля версий (VCS) и почему Git является стандартом индустрии.
- **Базовые операции Git**: `git init`, `git add`, `git commit`, `git status`, `git log`. Работа с историей изменений.
- **Удаленные репозитории**: Клонирование репозитория (`git clone`), отправка изменений (`git push`), получение изменений (`git pull`).
- **SSH-клонирование**: Настройка SSH-ключей и использование их для безопасного взаимодействия с удаленными репозиториями.
- **Стратегии ветвления**: Обзор популярных стратегий (Gitflow, GitHub Flow, GitLab Flow) и их применение.
- **Основы групповой разработки**: Ветвление (`git branch`), переключение между ветками (`git checkout`), слияние веток (`git merge`).
- **Разрешение конфликтов**: Пошаговое руководство по устранению конфликтов при слиянии кода.
- **Практика**: Создание общего репозитория в Azure Repos, настройка SSH-доступа, работа с ветками, выполнение pull request'ов и разрешение конфликтов в команде.

## Тема 2: Azure Boards и Agile-управление проектом
- **Введение в Agile-методологии**: Краткий обзор Scrum и Kanban, их принципы и преимущества.
- **Azure Boards**: Обзор интерфейса, создание организации и проекта.
- **Рабочие элементы**: Создание и управление Epic, Feature, User Story, Task, Bug. Понимание иерархии и жизненного цикла элементов.
- **Бэклог и спринты**: Приоритизация задач, планирование итераций, оценка трудозатрат.
- **Доски**: Использование Kanban-досок для визуализации рабочего процесса и отслеживания прогресса.
- **Кастомные отчеты и дашборды**: Создание и использование настраиваемых отчетов и дашбордов в Azure Boards для визуализации метрик проекта и команды.
- **Практика**: Планирование проекта в Azure Boards, создание бэклога, разбивка задач и распределение их по спринтам, отслеживание прогресса на досках и создание кастомных отчетов.

## Тема 3: Разработка REST API на ASP.NET Core
- **Архитектура REST**: Глубокое погружение в принципы RESTful API, HTTP-методы (GET, POST, PUT, DELETE, PATCH), коды состояния HTTP.
- **Создание ASP.NET Core Web API**: Структура проекта, контроллеры, маршрутизация, модели данных.
- **Обработка входных данных**: Валидация моделей, привязка данных.

- **Тестирование API с Postman**: Использование Postman для отправки запросов, сохранения коллекций и автоматизации базовых тестов API.
- **Документирование API с Swagger/OpenAPI**: Интеграция Swashbuckle для автоматической генерации интерактивной документации API.
- **Практика**: Разработка REST API для основной части продукта (например, CRUD-операции для нескольких сущностей), реализация аутентификации/авторизации, документирование его с помощью Swagger и тестирование в Postman.

## Тема 4: Паттерны проектирования и тестирование в ASP.NET Core (с CQRS)
- **Паттерны проектирования**: Введение в основные паттерны (Dependency Injection, Repository, Unit of Work) и их применение в ASP.NET Core.
- **Внедрение зависимостей (DI)**: Глубокое понимание DI-контейнера ASP.NET Core для создания слабосвязанного и тестируемого кода.
- **CQRS (Command Query Responsibility Segregation)**:
  - Введение в CQRS: Объяснение принципов разделения операций чтения (Queries) и записи (Commands). Преимущества CQRS (масштабируемость, производительность, гибкость).
  - Идемпотентность команд: Понимание концепции идемпотентности для команд в распределенных системах.
  - Реализация CQRS: Использование библиотек типа MediatR для создания команд, запросов и их обработчиков.
- **Unit-тестирование**: Написание модульных тестов для бизнес-логики и обработчиков CQRS с использованием xUnit/NUnit и Moq.
- **Интеграционное тестирование**: Тестирование взаимодействия между компонентами (например, API-контроллера с обработчиками CQRS, или обработчиков с базой данных).
- **Практика**: Рефакторинг части существующего API-проекта с применением CQRS, написание модульных и интеграционных тестов для новых компонентов.

## Тема 5: Непрерывная интеграция (CI) с Azure Pipelines
- **Концепция CI**: Автоматизация сборки, тестирования и проверки качества кода.
- **Azure Pipelines**: Обзор сервиса, агенты пайплайнов (Microsoft-hosted vs self-hosted), пулы агентов.
- **Создание CI-пайплайна (YAML)**: Структура файла azure-pipelines.yml, триггеры, переменные.
- **Этапы CI**: Получение кода из Azure Repos, восстановление зависимостей, сборка ASP.NET Core-приложения, запуск всех модульных и интеграционных тестов.
- **Публикация артефактов**: Сохранение результатов сборки и тестов для дальнейшего использования.
- **Политики ветвления**: Настройка обязательных проверок пайплайна и ревью для pull request'ов.
- **Практика**: Настройка CI-пайплайна, который автоматически запускает сборку и все тесты после каждого push'а в главную ветку, а также настройка политики ветвления.

## Тема 6: Контейнеризация приложения с Docker
- **Введение в Docker**: Что такое контейнеры, образы, Dockerfile. Преимущества контейнеризации в разработке и развертывании.
- **Написание Dockerfile для ASP.NET Core**: Оптимизация Dockerfile для многоступенчатой сборки (multi-stage build) для уменьшения размера образа.
- **Сборка и запуск контейнера**: Команды `docker build`, `docker run`, `docker ps`, `docker logs`.
- **Управление образами**: `docker images`, `docker rmi`, `docker prune`.
- **Репозитории образов**: Введение в Docker Hub и Azure Container Registry (ACR). Публикация образа в ACR.
- **Практика**: "Упаковка" ASP.NET Core-приложения в Docker-контейнер, запуск его локально и публикация образа в ACR.

## Тема 7: Инфраструктура как код (IaC) и Управление базами данных
- **Введение в IaC**: Что такое Infrastructure as Code, преимущества и инструменты.
- **Azure Resource Manager (ARM) шаблоны / Bicep**: Основы синтаксиса, развертывание базовых ресурсов Azure (App Service Plan, App Service, Key Vault, SQL Database).
- **Управление базами данных**:
  - Интеграция с Azure SQL Database / Azure Cosmos DB.
  - Миграции баз данных (например, Entity Framework Core Migrations).
  - Управление схемами и данными.
- **Практика**: Создание ARM/Bicep шаблонов для развертывания необходимых ресурсов Azure для своего приложения (App Service, SQL Database, Key Vault) и настройка миграций базы данных.

## Тема 8: Непрерывное развертывание (CD) с Azure Pipelines
- **Концепция CD**: Автоматизация доставки и развертывания приложения в различные среды.
- **Настройка CD-пайплайна**: Создание YAML-пайплайна для развертывания.
- **Развертывание в Azure App Service**: Настройка развертывания Docker-контейнера в Azure App Service (Linux).
- **Развертывание в Azure Container Instances (ACI)**: Быстрое развертывание контейнеров для тестирования или небольших сервисов.
- **Продвинутые стратегии развертывания**: Обзор Blue/Green развертываний и Canary развертываний.
- **Переменные и окружения**: Управление конфигурацией для разных сред (Dev, Staging, Production) с использованием переменных пайплайна и Azure Key Vault.
- **Практика**: Настройка полного CI/CD-пайплайна, который автоматически собирает, тестирует, создает образ и разворачивает приложение в Azure при каждом успешном коммите, возможно, с использованием Blue/Green стратегии.

## Тема 9: Логирование, мониторинг, безопасность и устранение неполадок
- **Продвинутое логирование в ASP.NET Core**: Использование `ILogger`, настройка различных провайдеров (Console, Debug, File, Serilog).
- **Централизованное логирование**: Обзор Azure Log Analytics и Azure Monitor Logs.
- **Мониторинг с Azure Monitor**: Сбор метрик, создание дашбордов, настройка оповещений по ключевым показателям.
- **Azure Application Insights**: Глубокий мониторинг производительности, ошибок, зависимостей и использования приложения.
- **Управление секретами**: Безопасное хранение чувствительных данных с помощью Azure Key Vault и интеграция его с ASP.NET Core-приложением и пайплайнами.
- **Безопасность API**: Краткий обзор OWASP Top 10 для API и рекомендации по их предотвращению.
- **Устранение неполадок (Troubleshooting)**: Практические подходы к отладке приложений в Azure, анализу логов и метрик для выявления и устранения проблем.
- **Стоимость и оптимизация ресурсов Azure**: Краткое обсуждение того, как отслеживать и оптимизировать затраты в Azure, используя Azure Cost Management.
- **Практика**: Интеграция Serilog и Application Insights в приложение, настройка Key Vault для хранения строк подключения к базе данных и других секретов. Практика по выявлению и устранению проблем на основе логов и метрик.

## Тема 10: Асинхронная обработка и распределенные сообщения (Azure Service Bus / Kafka)
- **Асинхронная обработка**: Понятие фоновых задач, их необходимость.
- **Планировщики задач в ASP.NET Core**: Использование библиотек типа Hangfire или Quartz.NET для выполнения отложенных и повторяющихся задач.
- **Введение в брокеры сообщений**:
  - **Azure Service Bus**: Основы очередей и топиков, преимущества для надежной доставки сообщений.
  - **Введение в Kafka**: Что такое распределенная потоковая платформа сообщений. Основные концепции: топики, продюсеры, консьюмеры, брокеры.
- **Использование брокеров сообщений в ASP.NET Core**: Отправка и получение сообщений. Примеры использования (например, для обработки событий, очередей).
- **Практика**: Интеграция Hangfire для выполнения фоновых задач (например, отправка email) и настройка базового взаимодействия с Azure Service Bus или Kafka для обмена сообщениями между сервисами.

## Тема 12: Итоговый проект: Построение готового продукта
- **Групповая работа**: Студенты делятся на группы и разрабатывают полноценный продукт, интегрируя все изученные технологии.
- **Полный цикл DevOps**: Проект должен пройти весь путь: от планирования в Azure Boards и управления кодом в Azure Repos до автоматизированного CI/CD-развертывания в Azure с использованием IaC.
- **Включение продвинутых аспектов**: Использование CQRS, логирования, мониторинга, управления секретами, фоновых задач, брокеров сообщений и, если применимо, ML-компонента.
