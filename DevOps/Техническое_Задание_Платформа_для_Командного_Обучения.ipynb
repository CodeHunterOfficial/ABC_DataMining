# Техническое Задание: Платформа для Командного Обучения
## Версия: 1.0  
## Дата: 16 июля 2025 г.

### 1. Введение
Настоящий документ определяет требования к разработке веб-платформы "Платформа для Командного Обучения". Целью проекта является создание современного, масштабируемого и надежного программного решения, которое значительно упростит процесс организации и проведения командного обучения, а также взаимодействия между преподавателями и студентами. Система будет разработана с применением передовых практик DevOps, облачных технологий Microsoft Azure и архитектурных паттернов.

### 2. Цели Проекта
- **Автоматизация управления обучением**: Упрощение создания курсов, формирования команд, публикации заданий и оценки работ.
- **Улучшение взаимодействия**: Предоставление удобных инструментов для коммуникации между участниками образовательного процесса.
- **Повышение эффективности**: Сокращение рутинных операций для преподавателей и студентов.
- **Демонстрация современных практик**: Создание проекта, который служит примером использования полного цикла DevOps, облачных технологий и архитектурных паттернов в реальных условиях.
- **Создание готового продукта**: Разработка системы, пригодной для дальнейшего развития и эксплуатации в образовательной среде.

### 3. Целевая Аудитория
- **Преподаватели**: Пользователи с расширенными правами, ответственные за создание и управление курсами, командами, заданиями и оценками.
- **Студенты**: Основные пользователи, участвующие в курсах, выполняющие задания, сдающие работы и получающие обратную связь.

### 4. Функциональные Требования
Система должна предоставлять следующий функционал:

#### 4.1. Управление Пользователями и Доступом
- **Регистрация**: Возможность для новых пользователей зарегистрироваться в системе. По умолчанию регистрируются как "Студенты". Изменение роли на "Преподаватель" должно производиться администратором системы.
- **Вход в систему**: Аутентификация пользователей по Email и паролю. Система должна выдавать JWT-токен для последующих авторизованных запросов.
- **Авторизация на основе ролей**: Разграничение доступа к функционалу системы в зависимости от роли пользователя (Преподаватель/Студент). Например, только преподаватель может создавать курсы или оценивать работы.
- **Управление профилем**: Пользователи должны иметь возможность просматривать и редактировать свои основные данные (например, Email, пароль).

#### 4.2. Управление Курсами
- **Создание курсов**: Преподаватели должны иметь возможность создавать новые курсы, указывая название, описание и генерируя уникальный код для присоединения.
- **Редактирование/Удаление курсов**: Преподаватели могут изменять информацию о курсах или удалять их.
- **Просмотр курсов**:
  - Преподаватели видят все созданные ими курсы.
  - Студенты видят только те курсы, к которым они присоединились.
- **Присоединение к курсу**: Студенты могут присоединиться к курсу, введя его уникальный код.

#### 4.3. Управление Командами
- **Формирование команд**: Преподаватели могут создавать команды в рамках курса, добавляя в них студентов, уже присоединившихся к этому курсу.
- **Управление составом команд**: Преподаватели могут добавлять/удалять студентов из команд.
- **Просмотр команд**: Студенты могут видеть состав своей команды в рамках курса.

#### 4.4. Управление Заданиями и Сдачей Работ
- **Создание заданий**: Преподаватели могут создавать задания для конкретного курса или команды, указывая название, подробное описание и срок сдачи.
- **Редактирование/Удаление заданий**: Преподаватели могут изменять или удалять опубликованные задания.
- **Просмотр заданий**: Студенты видят список заданий, относящихся к их курсам.
- **Сдача работ**: Студенты могут сдавать работы, прикрепляя ссылку на внешний ресурс (например, GitHub-репозиторий или облачное хранилище файлов).

#### 4.5. Оценка Работ
- **Просмотр сданных работ**: Преподаватели могут просматривать все сданные работы по конкретному заданию.
- **Выставление оценки**: Преподаватели могут выставлять числовую оценку (например, от 0 до 100) за каждую сданную работу.
- **Комментарии к работе**: Преподаватели могут оставлять текстовые комментарии к сданным работам.

#### 4.6. Система Асинхронных Уведомлений и Планирование Задач
- **Уведомления об оценке**: После выставления оценки за работу, система должна автоматически генерировать и отправлять уведомление студенту (например, по Email или через внутреннюю систему сообщений).
- **Фоновая обработка**: Отправка уведомлений должна происходить асинхронно, не блокируя основной процесс выставления оценки.
- **Планирование задач**: Система должна иметь возможность планировать и выполнять периодические фоновые задачи (например, отправка ежедневных напоминаний о сроках сдачи заданий, сбор статистики).

#### 4.7. (Опционально) Интеллектуальный Анализ Заданий (MLOps)
- **Предварительная оценка текста**: Возможность для преподавателя получить предварительную "интеллектуальную" оценку или классификацию сданной текстовой работы (например, по сложности, полноте ответа) с использованием ML-модели.
- **Развертывание ML-модели**: ML-модель должна быть развернута как отдельный веб-сервис, доступный через API.

### 5. Нефункциональные Требования

#### 5.1. Производительность
- **Время ответа API**: Среднее время ответа на запросы API не должно превышать 500 миллисекунд.
- **Загрузка страниц**: Время загрузки клиентских страниц должно быть минимальным для комфортного взаимодействия с пользователем.
- **Масштабируемость**: Система должна быть спроектирована для горизонтального масштабирования API-сервиса и фоновых воркеров для обработки возрастающей нагрузки.

#### 5.2. Безопасность
- **Управление секретами**: Все конфиденциальные данные (строки подключения к БД, ключи API, учетные данные) должны храниться и управляться безопасным способом с использованием облачных хранилищ секретов (например, Azure Key Vault).
- **Защита API**: API должен быть защищен от распространенных веб-уязвимостей (OWASP Top 10), включая SQL-инъекции, XSS, CSRF, а также от несанкционированного доступа.
- **Аутентификация/Авторизация**: Надежная реализация аутентификации на основе JWT-токенов.

#### 5.3. Надежность
- **Обработка ошибок**: Система должна корректно обрабатывать ошибки на всех уровнях и предоставлять информативные сообщения пользователям и в логи.
- **Отказоустойчивость**: Асинхронные очереди сообщений должны обеспечивать надежную доставку уведомлений даже при временных сбоях сервисов. Механизмы повторных попыток должны быть предусмотрены.

#### 5.4. Мониторинг и Логирование
- **Централизованное логирование**: Все события, ошибки и предупреждения в приложении должны логироваться и быть доступны для централизованного анализа.
- **Мониторинг производительности**: Система должна собирать метрики производительности (время ответа, загрузка CPU/RAM, количество запросов) и предоставлять дашборды для их визуализации.
- **Оповещения**: Возможность настройки автоматических оповещений о критических ошибках или превышении пороговых значений метрик.

#### 5.5. Удобство использования (UX/UI)
- **Интуитивный интерфейс**: Пользовательский интерфейс должен быть простым, понятным и интуитивно понятным для обеих ролей.
- **Адаптивный дизайн**: Интерфейс должен корректно отображаться и функционировать на различных устройствах (десктопы, планшеты, смартфоны).

#### 5.6. Поддерживаемость и Расширяемость
- **Чистый код**: Код должен быть хорошо структурирован, читаем, с комментариями и соответствовать общепринятым стандартам кодирования.
- **Тестирование**: Код должен быть покрыт модульными и интеграционными тестами для обеспечения качества и упрощения рефакторинга.
- **Модульность**: Архитектура должна способствовать добавлению нового функционала без значительных изменений в существующих модулях.

### 6.2. Детализация Компонентов

- **LearningPlatform.Client (Blazor App):**
  - **Роль:** Пользовательский интерфейс.
  - **Технология:** Blazor Server (для упрощения и быстрой демонстрации).
  - **Функционал:** Отображение данных, взаимодействие с пользователем через формы и навигацию.
  - **Взаимодействие:** Общается с `LearningPlatform.API` по HTTP/S.
  - **Структура:**
    - `Pages/` — Blazor-компоненты для страниц.
    - `Shared/` — общие UI-компоненты.
    - `Services/` — клиентские сервисы для вызова API.

- **LearningPlatform.API (ASP.NET Core Web API):**
  - **Роль:** Точка входа для всех клиентских запросов, шлюз для бизнес-логики.
  - **Технология:** ASP.NET Core Web API.
  - **Функционал:** Прием HTTP/S запросов, аутентификация (JWT), авторизация (роли), валидация входных данных, делегирование запросов в `LearningPlatform.Core` через MediatR, генерация JSON-ответов, интеграция со Swagger/OpenAPI.
  - **Взаимодействие:**
    - Получает запросы от клиента.
    - Отправляет команды/запросы в `LearningPlatform.Core`.
    - Отправляет сообщения в RabbitMQ.
    - Логирует в Azure Application Insights.
    - Получает секреты из Azure Key Vault.
  - **Структура:**
    - `Controllers/`
    - `Program.cs`
    - `appsettings.json`

- **LearningPlatform.Core (Бизнес-логика):**
  - **Роль:** Содержит основную бизнес-логику приложения.
  - **Технология:** .NET Standard/.NET 8 Class Library.
  - **Функционал:** Реализует паттерн CQRS с MediatR (Команды, Запросы, Обработчики), содержит бизнес-правила, использует репозитории из `LearningPlatform.Data`.
  - **Взаимодействие:**
    - Получает команды/запросы от `LearningPlatform.API` (и потенциально от `LearningPlatform.BackgroundTasks`).
    - Использует репозитории из `LearningPlatform.Data`.
    - Может вызывать внешние сервисы (например, ML-модель).
    - Публикует события в RabbitMQ.
  - **Структура:**
    - `Commands/`
    - `Queries/`
    - `Handlers/`
    - `Interfaces/` — для репозиториев.
    - `Services/` — дополнительные сервисы бизнес-логики.

- **LearningPlatform.Data (Доступ к Данным):**
  - **Роль:** Отвечает за взаимодействие с базой данных.
  - **Технология:** .NET Standard/.NET 8 Class Library, Entity Framework Core.
  - **Функционал:** Содержит сущности предметной области (POCO-классы), реализует `DbContext` и паттерн Repository для абстрагирования доступа к данным. Управляет миграциями базы данных (Entity Framework Core Migrations).
  - **Взаимодействие:**
    - Используется обработчиками из `LearningPlatform.Core`.
    - Взаимодействует с выбранной базой данных.
  - **Структура:**
    - `Entities/`
    - `Repositories/`
    - `Context/` — `ApplicationDbContext`
    - `Migrations/`

- **LearningPlatform.Common (Общие Компоненты):**
  - **Роль:** Содержит общие модели и утилиты, используемые всеми слоями.
  - **Технология:** .NET Standard/.NET 8 Class Library.
  - **Функционал:** Определяет DTO (Data Transfer Objects) для передачи данных между API и клиентом, перечисления (Enums), общие константы.
  - **Взаимодействие:** Используется всеми остальными проектами.
  - **Структура:**
    - `DTOs/`
    - `Enums/`

- **LearningPlatform.BackgroundTasks (Фоновые Задачи):**
  - **Роль:** Выполнение асинхронных и запланированных задач.
  - **Технология:** .NET 8 Worker Service.
  - **Функционал:** Содержит Quartz.NET для планирования и выполнения периодических или отложенных задач. Содержит воркеры-консьюмеры RabbitMQ, которые слушают сообщения из очереди (например, `SubmissionGradedMessageConsumer`). Выполняет асинхронные операции, такие как отправка Email-уведомлений.
  - **Взаимодействие:**
    - Получает сообщения от RabbitMQ.
    - Может взаимодействовать с `LearningPlatform.Core` для выполнения бизнес-логики.
    - Логирует события в Azure Application Insights.
  - **Структура:**
    - `Jobs/` — Quartz.NET jobs.
    - `Consumers/` — RabbitMQ consumers.
    - `Services/` — Email sender service.

- **Планировщик задач:** Quartz.NET

### 8. Развертывание и Операции (DevOps)

- **Система контроля версий:** Использование Azure Repos для хранения исходного кода, с применением стратегий ветвления (например, GitFlow или GitHub Flow) и Pull Request'ов.
- **Управление проектом:** Все задачи, баги, фичи и спринты управляются в Azure Boards.
- **Непрерывная Интеграция (CI):** Автоматическая сборка, запуск тестов и создание Docker-образов при каждом изменении кода в Azure Pipelines.
- **Контейнеризация:** Приложение будет контейнеризировано с использованием Docker, а образы будут храниться в Azure Container Registry (ACR).
- **Инфраструктура как Код (IaC):** Все необходимые ресурсы Azure (App Service Plan, App Service, SQL Database/PostgreSQL, Key Vault, RabbitMQ, Azure ML Workspace) будут развертываться и управляться с помощью Azure Bicep.
- **Непрерывное Развертывание (CD):** Автоматическое развертывание контейнеризированного приложения в Azure App Service (для API и Blazor) и Azure Container Instances (для фоновых воркеров) через Azure Pipelines. Возможно использование стратегий Blue/Green или Canary развертываний.
- **Управление Секретами:** Все чувствительные данные (строки подключения, ключи API) будут храниться в Azure Key Vault и безопасно инжектироваться в приложение и пайплайны.


### 9. План Разработки (Детализированный по Этапам)

Проект будет реализован поэтапно, в соответствии с темами курса:

#### Этап 1: Основы Git и Управление Проектом (Azure Boards)
- **Цель:** Настройка базовой среды разработки и планирование проекта.
- **Действия:**
  - Инициализация Git-репозитория и его связывание с Azure Repos.
  - Настройка SSH-ключей для безопасного взаимодействия с репозиторием.
  - Освоение базовых операций Git (`clone`, `pull`, `push`, `branch`, `checkout`, `merge`) и стратегий ветвления (например, GitHub Flow).
  - Создание организации и проекта в Azure Boards.
  - Планирование проекта: создание Epic, Feature, User Story, Task в Azure Boards.
  - Создание первых спринтов и распределение задач.
- **Ожидаемые компоненты:** Настроенный репозиторий, заполненный бэклог в Azure Boards.

#### Этап 2: Разработка Core API и Аутентификации (ASP.NET Core)
- **Цель:** Создание основы Backend-приложения и системы аутентификации.
- **Действия:**
  - Создание проектов `LearningPlatform.API`, `LearningPlatform.Core`, `LearningPlatform.Data`, `LearningPlatform.Common`.
  - Определение сущностей (`User`, `Role`) в `LearningPlatform.Data/Entities`.
  - Настройка `ApplicationDbContext` и первой миграции для пользователей.
  - Реализация Repository для `User` (`IUserRepository`, `UserRepository`) в `LearningPlatform.Data/Repositories`.
  - Настройка Dependency Injection (DI) / Inversion of Control (IoC) контейнера в `LearningPlatform.API` для регистрации сервисов и репозиториев.
  - Разработка DTO для регистрации и входа (`RegisterRequest`, `LoginRequest`, `AuthResponse`) в `LearningPlatform.Common/DTOs`.
  - Реализация CQRS-команд (`RegisterUserCommand`) и CQRS-запросов (`LoginUserQuery`) в `LearningPlatform.Core/Commands` и `LearningPlatform.Core/Queries`.
  - Разработка соответствующих CQRS-обработчиков (`RegisterUserCommandHandler`, `LoginUserQueryHandler`) в `LearningPlatform.Core/Handlers`.
  - Создание `AuthController` в `LearningPlatform.API/Controllers` для эндпоинтов `/api/auth/register` и `/api/auth/login`.
  - Настройка JWT-аутентификации и ролевой авторизации в `LearningPlatform.API`.
  - Интеграция Swagger/OpenAPI для документирования API.
- **Ожидаемые компоненты:** Базовый API с аутентификацией, работающей с БД.

- **Цель:** Расширение API с использованием CQRS для управления курсами и заданиями, обеспечение качества кода через тестирование.
- **Действия:**
  - Определение сущностей (`Course`, `Assignment`, `Submission`) в `LearningPlatform.Data/Entities`.
  - Создание соответствующих Repository (`ICourseRepository`, `CourseRepository`, `IAssignmentRepository`, `AssignmentRepository`, `ISubmissionRepository`, `SubmissionRepository`) и реализация Unit of Work паттерна в `LearningPlatform.Data/Repositories`.
  - Настройка AutoMapper для преобразования между сущностями и DTO.
  - Разработка DTO (`CourseDto`, `AssignmentDto`, `SubmissionDto`) в `LearningPlatform.Common/DTOs`.
  - Реализация CQRS-команд (`CreateCourseCommand`, `UpdateCourseCommand`, `DeleteCourseCommand`, `JoinCourseCommand`, `CreateAssignmentCommand`, `SubmitAssignmentCommand`, `GradeSubmissionCommand`) в `LearningPlatform.Core/Commands`.
  - Реализация CQRS-запросов (`GetCoursesQuery`, `GetCourseByIdQuery`, `GetAssignmentsQuery`, `GetSubmissionsQuery`) в `LearningPlatform.Core/Queries`.
  - Разработка соответствующих CQRS-обработчиков в `LearningPlatform.Core/Handlers`.
  - Создание `CoursesController`, `AssignmentsController`, `SubmissionsController` в `LearningPlatform.API/Controllers`.
  - Написание Unit-тестов для обработчиков CQRS (в `LearningPlatform.Core.Tests`), используя Moq для зависимостей.
  - Написание интеграционных тестов для API-контроллеров (в `LearningPlatform.API.Tests`) для проверки полного потока запроса.
- **Ожидаемые компоненты:** Функциональный API для управления курсами, заданиями и сдачами работ, покрытый тестами.

#### Этап 6: Инфраструктура как Код (IaC) и Непрерывное Развертывание (CD)

    - Автоматического развертывания инфраструктуры с помощью Bicep.

#### Этап 7: Асинхронная Обработка и Мониторинг
- **Цель:** Внедрение системы уведомлений, фоновых задач и комплексного мониторинга.
